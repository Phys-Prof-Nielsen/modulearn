## D8.5.1 Motivation: What Does an Atom’s Light Reveal?

When gases are heated or excited by electricity, they emit light — but not a continuous rainbow. Instead, we see **discrete, sharp lines** of color. These **spectral lines** became one of the most precise tools in physics and chemistry.

> But why do atoms emit only specific colors of light? What do these lines tell us about the atom?

Studying atomic spectra not only revealed the **quantized nature of atomic behavior**, but also helped define the structure of the periodic table and identify elements across the universe.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.5.2 Historical Background

In the 1800s, physicists and chemists observed that:

- Each element emits its **own unique pattern** of spectral lines when heated.
- These patterns can be observed using a **spectroscope**, which separates light by wavelength.

This led to the development of **spectroscopy**, a powerful tool for probing atomic structure and composition.

### Major Discoveries:

- **Fraunhofer lines (1814)**: Dark absorption lines in the solar spectrum — first evidence of atomic fingerprints.
- **Bunsen and Kirchhoff (1850s)**: Showed that elements emit specific lines and that these could be used to **identify elements**.
- **Balmer (1885)**: Found a simple mathematical pattern for visible lines in hydrogen — known as the **Balmer series**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.5.3 What Are Atomic Spectra?

There are three main types of spectra:

1. **Continuous Spectrum**  
   - Produced by incandescent solids or dense gases.  
   - Contains all wavelengths without interruption (e.g. a heated filament).

2. **Emission Line Spectrum**  
   - Produced by hot, low-density gases.  
   - Contains **bright lines** at specific wavelengths.

3. **Absorption Line Spectrum**  
   - Produced when continuous light passes through a cooler gas.  
   - Contains **dark lines** where specific wavelengths are absorbed.

> These patterns are like **fingerprints** for each element.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.5.4 The Hydrogen Emission Spectrum

Hydrogen, the simplest atom, has a surprisingly rich and structured spectrum. When excited, hydrogen emits light only at certain discrete wavelengths — these form what are known as the **hydrogen spectral series**.

Each **series** corresponds to electron transitions ending in a particular energy level ($n_f$), with the electron falling from a higher energy level ($n_i > n_f$).

---

### The Spectral Series of Hydrogen

The general formula for all hydrogen spectral lines is:

$$
\frac{1}{\lambda} = R_H \left( \frac{1}{n_f^2} - \frac{1}{n_i^2} \right), \quad n_i > n_f
$$

Where:
- $\lambda$ is the emitted wavelength,
- $R_H \approx 1.097 \times 10^7 \ \text{m}^{-1}$ is the Rydberg constant,
- $n_i$ is the initial (higher) energy level,
- $n_f$ is the final (lower) energy level.

---

### Named Spectral Series

| Series     | Final Level ($n_f$) | Region of Spectrum | Example Transitions        |
|------------|----------------------|---------------------|-----------------------------|
| **Lyman**  | $n_f = 1$            | Ultraviolet         | $n=2 \to 1$, $n=3 \to 1$, … |
| **Balmer** | $n_f = 2$            | Visible             | $n=3 \to 2$, $n=4 \to 2$, … |
| **Paschen**| $n_f = 3$            | Infrared            | $n=4 \to 3$, $n=5 \to 3$, … |
| **Brackett** | $n_f = 4$          | Infrared            | $n=5 \to 4$, $n=6 \to 4$, … |
| **Pfund**  | $n_f = 5$            | Infrared (far)      | $n=6 \to 5$, $n=7 \to 5$, … |

---

### The Balmer Series (Visible Light)

The **Balmer series** is historically important because it includes lines in the visible spectrum. Johann Balmer (1885) found an empirical formula to fit these lines:

$$
\frac{1}{\lambda} = R_H \left( \frac{1}{2^2} - \frac{1}{n^2} \right), \quad n = 3, 4, 5, \ldots
$$

This formula predicted visible hydrogen lines such as:

- **H-alpha** (red) — $n=3 \to 2$ (~656 nm)
- **H-beta** (blue-green) — $n=4 \to 2$ (~486 nm)
- **H-gamma** (violet) — $n=5 \to 2$ (~434 nm)
- **H-delta** (deep violet) — $n=6 \to 2$ (~410 nm)

Balmer's formula matches the **observed spectral lines of hydrogen** with astonishing precision — but had **no classical explanation** at the time.

---

### What These Lines Represent

Each line in a hydrogen spectrum corresponds to a **quantized transition** between energy levels. The photon emitted has an energy:

$$
E = h\nu = \frac{hc}{\lambda}
$$

These transitions:
- Occur between discrete levels, not continuously
- Release photons in specific spectral regions (UV, visible, IR)
- Provide evidence of **quantized atomic energy states**

---

### Visualization Idea

If you'd like to visualize this:

- The **Lyman series** lies entirely in the **ultraviolet**
- The **Balmer series** is the only one in the **visible**
- All other series (Paschen, Brackett, Pfund) are in the **infrared**

These patterns were known **before** any theoretical model explained them — and were one of the key clues that led to the Bohr model of the atom.

---

### Summary of Spectral Series

| Series     | $n_f$ | Region     | Photon Energy | Historical Importance       |
|------------|-------|------------|----------------|------------------------------|
| Lyman      | 1     | Ultraviolet| High           | Detected via UV spectroscopy |
| Balmer     | 2     | Visible     | Medium         | First discovered (Balmer)    |
| Paschen    | 3     | Infrared    | Lower          | Found using IR detectors     |
| Brackett   | 4     | Infrared    | Lower          | Observed in atomic physics   |
| Pfund      | 5     | Infrared    | Lowest         | Rare, far IR                 |

## D8.5.5 Spectral Lines: Key Features

- **Discrete**: Only specific wavelengths are emitted or absorbed.
- **Sharp**: The lines are extremely narrow, indicating what we now know as **precise energy transitions**.
- **Element-Specific**: Each element has a unique set of spectral lines.
- **Energy-Wavelength Relationship**:

  $$
  E = \frac{hc}{\lambda}
  $$

This means that different lines correspond to **specific energy differences** within the atom — a hint at **quantized energy levels** (to be explored later).

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.5.6 Why Classical Physics Couldn’t Explain It

Under classical physics:
- Electrons should spiral into the nucleus, emitting a **continuous range** of energy.
- Atoms should radiate at **all frequencies**, not in discrete lines.
- There was **no mechanism** to explain fixed emission wavelengths.

These spectral patterns were among the **clearest signs that classical physics was incomplete** — and they laid the groundwork for quantum theory.

## D8.5.7 Applications of Atomic Spectra

### Element Identification (Spectroscopy)
- Used in chemistry, physics, and materials science to determine elemental composition.

### Astronomy and Cosmology
- Determine the composition of stars and galaxies by analyzing starlight.
- Identify elements across interstellar space (hydrogen, helium, etc.).

### Redshift and Motion
- Measure the **Doppler shift** of spectral lines to determine motion of stars/galaxies.

### Lighting Technologies
- Neon lights, sodium lamps, and fluorescent tubes work based on gas discharge and spectral emission.

## D8.5.8 Summary

- Atomic spectra consist of **discrete lines**, unique to each element.
- The **Balmer formula** for hydrogen’s visible lines was a major clue toward quantized energy.
- Classical physics could not explain why atoms emitted specific wavelengths.
- These spectra are central to fields from **quantum theory** to **astrophysics**.
