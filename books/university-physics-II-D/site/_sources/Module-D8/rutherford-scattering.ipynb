## D8.4.1 Motivation: What Is an Atom Made Of?

At the dawn of the 20th century, the atom was thought to be a simple, diffuse object — a kind of soft blob with positive and negative charge evenly distributed. But was that really true?

To find out, scientists began probing atoms with fast-moving particles. What they discovered would **revolutionize atomic physics** and lay the groundwork for quantum theory.

> The answer came from **shooting particles at gold** — and what they found changed physics forever.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.2 Historical Context: The Plum Pudding Model

After discovering the **electron** in 1897, J.J. Thomson proposed a new model of the atom to account for both negative and positive charge. Known as the **Plum Pudding Model**, it was the leading atomic model in the early 1900s.

### What the Model Proposed:

- Atoms are **spheres of diffuse positive charge**.
- Electrons are **embedded within** this sphere like **raisins in a pudding** (or plums in a cake).
- The positive charge was thought to **neutralize** the negative charge of the electrons.
- The mass of the atom was assumed to be **spread out uniformly**.

### Why It Made Sense (At the Time):

- There was **no experimental evidence** yet for a concentrated nucleus.
- Electron scattering experiments lacked the resolution to probe atomic structure.
- The model predicted that particles like alpha rays would pass through atoms with **only slight deflections** due to the spread-out positive charge.

### The Critical Assumption:

If an alpha particle (massive and positively charged, essentially a helium nuclei) passed through such an atom:
- It would experience **tiny deflections** from the weak electric field of the diffuse positive charge.
- **Large-angle scattering should be extremely rare — or impossible**.

This is exactly what Rutherford set out to test — and what the gold foil experiment would ultimately disprove.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.3 The Gold Foil Experiment (1909)

**Ernest Rutherford**, working with **Hans Geiger** and **Ernest Marsden**, devised an experiment to test atomic structure:

### Setup:
- A source of **alpha particles** directed at a **thin gold foil**.
- A fluorescent screen surrounded the foil to detect scattered particles.

### Expectations (if Thomson’s model were correct):
- Alpha particles should **pass through with minimal deflection**, as the positive charge was thought to be spread out.

### Shocking Observations

1. Most alpha particles **did pass straight through**, as expected.
2. Some were **deflected at large angles**.
3. A few were even **scattered backward** — almost straight back!

> “It was almost as incredible as if you fired a 15-inch shell at a piece of tissue paper and it came back and hit you.” — Rutherford

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.4 Rutherford’s Interpretation: A Nuclear Atom

To explain the data, Rutherford proposed a radically different model:

- The atom contains a **small, dense, positively charged nucleus**.
- The nucleus holds **nearly all the atom’s mass**.
- Electrons orbit this nucleus at relatively large distances.
- Most of the atom is **empty space**, allowing alpha particles to pass through.

This was the birth of the **nuclear model of the atom** — a direct precursor to the Bohr model and quantum atomic theory.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.5 Conceptual View of Scattering

When an alpha particle approaches the gold nucleus:
- It is **repelled** by the strong positive charge via the Coulomb force.
- If it passes close enough, it can be deflected — sometimes at large angles.
- The closer the encounter, the stronger the deflection.

Far from the nucleus, there’s little to no effect — hence most particles pass straight through.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.6 Rutherford Scattering Formula

Rutherford also derived a formula for the number of alpha particles scattered at different angles:

$$
\frac{d\sigma}{d\Omega} = \left( \frac{1}{4\pi\varepsilon_0} \cdot \frac{2Ze^2}{4E} \right)^2 \cdot \frac{1}{\sin^4(\theta/2)}
$$

Where:
- $\frac{d\sigma}{d\Omega}$: differential cross section (probability of scattering into angle $\theta$)
- $Z$: atomic number of target nucleus (e.g., $Z = 79$ for gold)
- $e$: elementary charge
- $E$: kinetic energy of the alpha particle
- $\theta$: scattering angle
- $\varepsilon_0$: vacuum permittivity

### Key Features:
- **Strong angular dependence**: scattering probability falls off as $\sin^{-4}(\theta/2)$
- Most particles scatter at small angles
- Rare but possible large-angle scattering → direct evidence of a **compact nucleus**

## D8.4.7 Significance of the Rutherford Experiment

✅ Disproved the plum pudding model  
✅ Demonstrated that atoms are **mostly empty space**  
✅ Revealed the existence of the **nucleus**  
✅ Laid the groundwork for **quantum atomic models**

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.8 Modern Applications of Rutherford Scattering

Although originally done with gold foil and alpha particles, Rutherford’s method evolved into techniques used in:

### Nuclear and Particle Physics
- Probing the internal structure of protons, neutrons, and nuclei.
- Using electron or muon scattering instead of alpha particles.

### Materials Science
- **Rutherford Backscattering Spectrometry (RBS)** is used to analyze thin films and surface layers with extreme precision.

### Astrophysics
- High-energy scattering techniques are used in cosmic ray and dark matter detection.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D8.4.9 Summary

- The gold foil experiment revealed that atoms have a **dense, charged nucleus**.
- Most of the atom is **empty space**, allowing alpha particles to pass through.
- The nuclear model replaced the outdated plum pudding model.
- The mathematical treatment of scattering matched experimental data — supporting the Coulomb force as the underlying interaction.
- Rutherford’s work directly led to the **Bohr model**, **quantum mechanics**, and **modern nuclear physics**.
