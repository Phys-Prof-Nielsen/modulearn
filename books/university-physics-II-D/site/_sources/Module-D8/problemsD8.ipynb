## Problem 8.3

**Context:**  
A star's total emitted power per unit area (its radiative flux) can be calculated by integrating **Planck's radiation law** across all frequencies.

This is how we derive the **Stefan–Boltzmann Law** from Planck's Law — and understand the *total energy* a star radiates per square meter of surface.

### Given:

Planck’s spectral energy density per unit frequency:

$$
u(\nu, T) = \frac{8 \pi h \nu^3}{c^3} \cdot \frac{1}{e^{\frac{h\nu}{k_B T}} - 1}
$$

We want to compute the **total power per unit area** emitted from the surface of a blackbody (like a star), which is given by:

$$
P(T) = \int_0^\infty \frac{c}{4} u(\nu, T) \, d\nu
$$

> The factor $\frac{c}{4}$ comes from averaging the energy flux over all directions.

### Tasks:

1. Perform the substitution $x = \frac{h \nu}{k_B T}$ and express the integral in terms of $x$.
2. Show that the integral becomes (feel free to use integral calculator or look up in a table):

$$
P(T) = \frac{2 \pi^5 k_B^4}{15 c^2 h^3} T^4
$$

3. Identify the constant in front as the **Stefan–Boltzmann constant** $\sigma$, and verify its value numerically.
4. For a star with surface temperature $T = 5800$ K (like the Sun), calculate the total power emitted per square meter.

## Problem 8.6

**Context:**  
A modern version of the Rutherford experiment is performed using a thin gold foil and a beam of 5.0 MeV alpha particles. The detector counts how many particles are scattered at large angles.

**Given:**

- Target: **Gold** ($Z = 79$)
- Alpha particle charge: $z = 2$
- Kinetic energy of alpha particles: $E = 5.0 \ \text{MeV}$
- Scattering angle of interest: $\theta = 60^\circ$
- Distance from foil to detector: $r = 10 \ \text{cm}$
- Detector solid angle coverage: $\Delta \Omega = 1.0 \times 10^{-3} \ \text{sr}$
- Number of incident alpha particles: $N_0 = 10^8$
- Number of gold atoms per unit area: $n = 5.9 \times 10^{22} \ \text{atoms/m}^2$

### Task:

**Estimate the number of alpha particles** scattered into the detector at $\theta = 60^\circ$ using the Rutherford differential cross section formula:

$$
\frac{d\sigma}{d\Omega} = \left( \frac{1}{4\pi\varepsilon_0} \cdot \frac{2Zze^2}{4E} \right)^2 \cdot \frac{1}{\sin^4(\theta/2)}
$$

Then compute:

$$
N(\theta) = N_0 \cdot n \cdot \frac{d\sigma}{d\Omega} \cdot \Delta\Omega
$$

### Conceptual Insight:

- Large-angle scattering events are **rare**, but still measurable with high beam intensity.
- The $\sin^{-4}(\theta/2)$ dependence means small changes in angle greatly affect probability.
- This analysis allows physicists to **map the nuclear charge distribution** and test the validity of Coulomb interaction models.
