# D6.7 Thin Lenses

## D6.7.1 Learning Goals
By the end of this lesson, students should be able to:
- Distinguish between converging and diverging lenses.
- Construct accurate ray diagrams for image formation.
- Apply the thin lens equation and magnification formula.
- Understand the role of focal length, object distance, and image distance.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.2 What Are Thin Lenses?

A **lens** is a transparent optical element that bends (refracts) light to form images.

A **thin lens** is one where the lens thickness is much smaller than the radii of curvature of its surfaces — so we can treat all refraction as occurring at a single plane.

There are two main types:
- **Converging lens (convex)**: brings parallel rays **to a point** (focus)
- **Diverging lens (concave)**: spreads parallel rays **away from a point**

The **focal point** is the point at which rays initially parallel to the optical axis converge (for a converging lens) or appear to diverge from (for a diverging lens) after passing through the lens. It represents where the lens focuses light and is located a distance equal to the **focal length** from the lens.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.3 Principal Rays and Ray Diagrams

To construct image diagrams, we use three **principal rays**:

### For a converging lens:
1. **Parallel ray**: enters parallel to the axis, exits through the far focal point.
2. **Central ray**: goes through the center of the lens, continues straight.
3. **Focal ray**: enters through the near focal point, exits parallel to the axis.

### For a diverging lens:
1. **Parallel ray**: enters parallel, exits as if it came from the near focal point.
2. **Central ray**: goes through the center, continues straight.
3. **Focal ray**: heads toward the far focal point, exits parallel to the axis.

Where the rays **intersect** (or appear to intersect), the **image** is formed.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.4 The Thin Lens Equation

The relationship between focal length, object distance, and image distance is:

<div class="alert alert-success">
    
$$
\frac{1}{f} = \frac{1}{d_o} + \frac{1}{d_i}
$$

</div>
    
Where:
- $f$: focal length (positive for converging, negative for diverging)
- $d_o$: object distance (positive if object is on the incoming side)
- $d_i$: image distance (positive if image is on the outgoing side)

Use **sign conventions** carefully:
- **Converging lens**: $f > 0$
- **Diverging lens**: $f < 0$
- **Real image**: $d_i > 0$
- **Virtual image**: $d_i < 0$

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.5 Magnification

The size and orientation of the image is given by:

<div class="alert alert-success">
    
$$
M = \frac{h_i}{h_o} = -\frac{d_i}{d_o}
$$
    
</div>

Where:
- $M$: magnification
- $h_i$, $h_o$: image and object heights
- $M > 0$: image is **upright**
- $M < 0$: image is **inverted**

If $|M| > 1$, the image is **larger** than the object.  
If $|M| < 1$, the image is **smaller**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.6 Real vs. Virtual Images

- **Real image**: formed when rays **actually converge**; can be projected on a screen.
- **Virtual image**: formed when rays **appear to diverge** from a point; cannot be projected.

Examples:
- Converging lens: real image when object is **outside** focal length; virtual if **inside**.
- Diverging lens: always forms **virtual**, **upright**, **smaller** image.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.7.7 Applications of Thin Lenses

- Cameras  
- Microscopes  
- Telescopes  
- Magnifying glasses  
- Eyeglasses and contact lenses

These devices use combinations of converging and diverging lenses to control image formation, size, and position.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D6.7.8 Summary

- **Converging lenses** focus light to a point; **diverging lenses** spread it out.
- The **thin lens equation** relates object distance, image distance, and focal length.
- Use sign conventions and ray diagrams to determine image type.
- Magnification tells us how **big and in what orientation** the image is.
