##  D6.5.2 What Is Total Internal Reflection?

**Total internal reflection (TIR)** occurs when a light ray attempts to pass from a **more optically dense** medium into a **less dense** one (e.g., glass to air) at an angle too steep to allow refraction.

- Instead of bending out, the light **reflects entirely back** into the original medium.
- TIR only occurs when:
  1. Light is traveling from a medium with **higher** refractive index ($n_1$) to one with **lower** index ($n_2$).
  2. The angle of incidence **exceeds the critical angle**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.5.3 Deriving the Critical Angle

To find the **critical angle** $\theta_c$, apply **Snell's Law**:

$$
n_1 \sin\theta_1 = n_2 \sin\theta_2
$$

The **critical angle** occurs at the exact angle of incidence for which the **refracted ray travels along the boundary** â€” that is, when $\theta_2 = 90^\circ$.

Substitute this into Snellâ€™s Law:

$$
n_1 \sin\theta_c = n_2 \sin 90^\circ
$$

Since $\sin 90^\circ = 1$, we get:

<div class="alert alert-warning">
    
$$
\theta_c = \sin^{-1}\left(\frac{n_2}{n_1}\right)
$$
    
</div>

> The critical angle $\theta_c$ is the **smallest angle of incidence** at which **no refracted ray escapes** â€” instead, all light reflects internally.

- If $\theta_i < \theta_c$, partial refraction occurs.
- If $\theta_i = \theta_c$, the refracted ray grazes the surface.
- If $\theta_i > \theta_c$, **total internal reflection** occurs.

Note: TIR can only happen when $n_1 > n_2$.

## D6.5.5 Real-World Applications of TIR

- **Fiber optics**: Light bounces along the core due to TIR.
- **Endoscopy**: Flexible light delivery in medical imaging.
- **Sparkle of diamonds**: Cut to maximize TIR.
- **Retroreflectors**: Used in road signs, bike reflectors.
- **Rain sensors**: Cars use TIR to detect water on windshields.

## D6.5.6 Summary

- **Total internal reflection** happens when light cannot exit a material â€” instead, it reflects entirely inside.
- It requires $n_1 > n_2$ and $\theta_i > \theta_c$.
- The **critical angle** occurs when the **refracted angle is $90^\circ$**, so the refracted ray skims the boundary.
- The critical angle is given by:

  $$
  \theta_c = \sin^{-1}\left(\frac{n_2}{n_1}\right)
  $$

- TIR is a **fully ray-based phenomenon** and explains many technologies involving light guidance.
