## D6.4.2 What Is Refraction?

**Refraction** is the bending of light as it passes from one medium into another with a different optical density (refractive index).

- Caused by a **change in light speed** across the boundary.
- Light bends **toward the normal** when entering a more optically dense medium.
- Light bends **away from the normal** when entering a less dense medium.

The direction of the light ray changes, but its frequency remains constant.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.4.3 Snellâ€™s Law

Snellâ€™s Law quantitatively describes how light bends at an interface:

<div class="alert alert-success">

$$
n_1 \sin\theta_1 = n_2 \sin\theta_2
$$
    
</div>

Where:
- $n_1$: refractive index of medium 1  
- $n_2$: refractive index of medium 2  
- $\theta_1$: angle of incidence (from the normal)  
- $\theta_2$: angle of refraction (from the normal)

The **refractive index** $n$ is defined as: 

<div class="alert alert-success">
    
$$
 n = \frac{c}{v}
$$
   
</div>

where $c$ is the speed of light in vacuum and $v$ is the speed in the medium.

## D6.4.4 Drawing Ray Diagrams for Refraction

Steps:
1. Draw the **boundary** between two media.
2. Draw the **normal line** perpendicular to the surface at the point of incidence.
3. Draw the **incident ray** approaching the boundary.
4. Use Snellâ€™s Law to estimate the angle of refraction.
5. Draw the **refracted ray**, bending appropriately:
   - **Toward the normal** if $n_2 > n_1$
   - **Away from the normal** if $n_2 < n_1$

## D6.4.5 Real-World Examples of Refraction
- A straw appears bent in a glass of water.
- Lenses focus light by refraction through curved surfaces.
- Mirage formation in deserts and on roads.
- Corrective eyeglasses and contact lenses.
- Optical fibers guiding light via total internal reflection (related topic).

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.4.6 Special Case: No Refraction

If $\theta_1 = 0^\circ$ (normal incidence), then $\theta_2 = 0^\circ$ â€” the light ray passes straight through with no change in direction.

Also, if $n_1 = n_2$, there is no bending regardless of angle.

## D6.4.7 Critical Angle and Total Internal Reflection (Preview)

When light travels from a more dense to a less dense medium ($n_1 > n_2$):

- There exists a **critical angle** $\theta_c$ beyond which **no refraction** occurs.
- Light is completely reflected back into the medium â€” this is **total internal reflection**.

## D6.4.8 Summary
- Refraction is the **bending of light** due to a change in speed across a boundary.
- Described by **Snellâ€™s Law**: $n_1 \sin\theta_1 = n_2 \sin\theta_2$.
- The direction of bending depends on the relative refractive indices.
- Refraction is foundational in understanding lenses, vision correction, and many optical technologies.
