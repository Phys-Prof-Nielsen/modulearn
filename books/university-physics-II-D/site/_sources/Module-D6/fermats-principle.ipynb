## D6.6.2 What Is Fermat’s Principle?

**Fermat’s Principle of Least Time** states:

> *The path taken by a ray of light between two points is the one that takes the least time.*

This is a **variational principle** — it does not prescribe a path, but says that **of all possible paths**, light chooses the one that minimizes travel time.

It unifies:
- **Straight-line motion** in uniform media
- **Reflection** at boundaries
- **Refraction** at interfaces

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.6.3 Application to Reflection

Imagine light traveling from point A to point B, reflecting off a flat mirror.

- Out of all the possible points where light could reflect, Fermat’s Principle tells us it picks the one that minimizes total time.
- Since the speed of light is the same before and after reflection, this becomes a geometric problem: **minimize total path length**.

When you do this minimization, you get:

$$
\theta_i = \theta_r
$$

That is, **the angle of incidence equals the angle of reflection**.

➡️ Fermat’s Principle explains why this rule exists, not just how to use it.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.6.4 Application to Refraction

Light goes from point A in medium 1 to point B in medium 2.

- Speeds are different: $v_1$ in medium 1, $v_2$ in medium 2.
- Fermat’s Principle: **minimize total travel time** across both media.

Since time = distance / speed:
- Time in medium 1: $t_1 = \frac{\ell_1}{v_1}$
- Time in medium 2: $t_2 = \frac{\ell_2}{v_2}$
- Total time: $T = t_1 + t_2$

Minimizing this total with respect to the boundary point leads to:

$$
n_1 \sin \theta_1 = n_2 \sin \theta_2
$$

That is, **Snell’s Law**.

➡️ Fermat’s Principle explains **why** light bends the way it does.

## D6.6.5 A Glimpse Ahead: Mechanics and the Least Action Principle

Fermat’s Principle is just one member of a larger family of **“least” or “extremal” principles** in physics.

In **classical mechanics**, there's a similar idea called the **Principle of Least Action**:

> A particle moves between two points in space and time in such a way that the **action** is minimized (or more precisely, extremized).

The **action** is defined as:

$$
S = \int L \, dt, \quad \text{where } L = T - V
$$

- $L$ is the **Lagrangian**, the difference between kinetic and potential energy.
- The path a particle takes is the one for which $S$ is minimized.

This is **analogous** to Fermat’s Principle — instead of minimizing time, particles minimize a quantity related to **energy**.

**Takeaway**: Nature often “chooses” paths that optimize something. In optics, it’s time. In mechanics, it’s action.

## D6.6.6 Deepest Connection: Quantum Physics and the Sum Over All Paths

Fermat’s Principle turns out to be not just an approximation — it's a **shadow** of something deeper.

In **quantum field theory**, light and particles do not follow a single path.

According to **Feynman’s Path Integral Formulation**:
> A photon explores **all possible paths** between two points — and the paths “add up” quantum mechanically.

- Paths that take much longer or shorter times interfere destructively.
- The paths **near the one of least time** interfere constructively and dominate the result.

So: what we observe as the **“least-time path”** in classical optics is the result of **quantum interference** of **infinitely many paths**.

➡️ Fermat’s Principle is an **emergent classical behavior** that arises from **quantum mechanics**.

## D6.6.7 Summary

- **Fermat’s Principle** says light takes the path that minimizes total travel time.
- From this single principle, we can derive:
  - Law of Reflection: $\theta_i = \theta_r$
  - Snell’s Law: $n_1 \sin \theta_1 = n_2 \sin \theta_2$
- Fermat’s Principle is part of a broader theme in physics: **optimization principles**.
- It connects to:
  - **Classical mechanics** via the principle of least action
  - **Quantum field theory** via Feynman’s sum-over-paths idea

This principle not only explains how light behaves — it gives us a preview of how **deeply unified** physical laws really are.
