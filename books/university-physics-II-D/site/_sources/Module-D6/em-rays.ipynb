## D6.2.2 What Is Light?

Light is an **electromagnetic (EM) wave** — a self-sustaining oscillation of **electric and magnetic fields** traveling through space.

- It propagates at the **speed of light**:

  $$ c = 3 \times 10^8 \, \text{m/s} $$


- The electric and magnetic fields are **perpendicular** to each other and to the direction of motion — a **transverse wave**.

- All visible colors, infrared, ultraviolet, X-rays, and radio waves are part of the **electromagnetic spectrum**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.2.3 Wavefronts and Rays

- A **wavefront** is a surface of constant phase (like the crest of a wave).  
  → Think of it as the "leading edge" of the wave.

- A **ray** is a line **perpendicular to the wavefront**, indicating the **direction of energy flow**.

- In uniform media, **rays travel in straight lines**. In non-uniform media, rays bend — this is the basis for **refraction**.

**Key idea**: Rays are not physical objects — they are **mathematical abstractions** that show how energy moves.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.2.4 When Can We Treat Light as a Ray?

We can simplify light as rays when:

- The **wavelength** is much smaller than the size of objects it interacts with:  

  $$ \lambda \ll L $$

- There is **no significant interference or diffraction**.

This is known as the **geometrical optics limit** or **eikonal approximation** — a high-frequency limit of wave optics.

**Valid scenarios:**
- Lenses and mirrors in imaging systems
- Laser beams in air
- Reflection/refraction at flat surfaces
- Even some **interference effects**, if we only track the ray paths and consider their phase shifts

> Thin film interference will be a **ray-based application** of wave principles — tracking the paths, angles, and phase changes of reflected rays.

## D6.2.5 Rays from Maxwell's Equations (Without Derivation)

While we won't dive into the full Maxwell equations yet, it's helpful to know this:

- Maxwell’s equations predict **wave solutions** for electric and magnetic fields.
- These fields satisfy the **wave equation**:

  $$ \nabla^2 \mathbf{E} - \frac{1}{c^2} \frac{\partial^2 \mathbf{E}}{\partial t^2} = 0 $$
  
- In the **short-wavelength limit**, these wavefronts behave like **geometrical surfaces** and rays emerge **perpendicular** to them.

So ray optics is not a separate theory — it's an **approximation** of full electromagnetic theory.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.2.6 Fermat’s Principle of Least Time

A foundational idea in ray optics is **Fermat’s Principle**:
> *Light takes the path that requires the least time.*

This leads to:
- **Straight-line paths** in uniform media
- **Snell's Law** in layered materials
- **Reflection and refraction rules** that we’ll explore in detail later

This principle is the **ray-optics analog** of how nature "optimizes" energy and time — and it's a stepping stone toward **variational principles** in physics.

## D6.2.8 Summary

- Light is an electromagnetic wave, but in many situations, we can treat it as traveling in straight **rays**.
- This approximation works when the **wavelength is small** compared to the system.
- Rays are **perpendicular to wavefronts** and represent the **direction of energy propagation**.
- This approach allows us to predict and understand **reflection**, **refraction**, **image formation**, and **optical design** — which are the focus of ray optics.
