## D6.3.1 Learning Goals
By the end of this lesson, students should be able to:
- State the law of reflection.

- Predict the path of a light ray reflecting off flat and curved surfaces.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.3.2 Law of Reflection

<div class="alert alert-success">
    
The angle of incidence equals the angle of reflection:

$$ \theta_i = \theta_r $$
    
</div>

Where:
- $ \theta_i $ is the angle between the **incident ray** and the **normal** (perpendicular to the surface).
- $ \theta_r $ is the angle between the **reflected ray** and the **normal**.

This law holds for:
- Flat (plane) mirrors
- Smooth curved surfaces (locally approximated as flat)

The **normal line** is always drawn **perpendicular to the surface** at the point of incidence.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.3.3 Drawing Ray Diagrams for Plane Mirrors

Steps:
1. Draw the incident ray approaching the surface.
2. At the point of incidence, draw the **normal**.
3. Measure the angle $ \theta_i $ from the incident ray to the normal.
4. Draw the reflected ray such that $ \theta_r = \theta_i $.

**Conventions:**
- Use arrows to indicate the direction of ray travel.
- Label all angles and rays clearly.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D6.3.4 Reflection from Curved Surfaces

We can apply the same law of reflection **locally** to curved mirrors:
- Treat each small part of the surface as if it were flat.
- At the point of incidence, draw the normal (perpendicular to the tangent).
- Use the law $ \theta_i = \theta_r $.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D6.3.7 Summary
- Light reflects from surfaces according to the rule: $ \theta_i = \theta_r $.
- The law of reflection applies to **all surfaces**, flat or curved, when defined locally.
- Reflection is a cornerstone of ray optics and forms the basis for more complex optical systems.

**Box 1: Predicting the Reflection Angle**

A floor-to-ceiling **vertical mirror** is mounted on a wall, and a laser beam strikes it at a point $35.0$ cm above the floor. The upward angle of incidence is $35^\circ$ with respect to the normal.

**Tasks:**
1. Sketch the situation: include the floor, the mirror, the laser ray, and the normal.
