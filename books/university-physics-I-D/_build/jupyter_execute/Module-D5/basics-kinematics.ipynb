## D5.1.1 What Kinematics Is (and Is Not)

Kinematics is the **description of motion**, not its **cause**. In this section, we are not asking *why* an object moves the way it does. Instead, we focus on **how motion is represented and quantified** using measurable quantities.

Kinematics provides the mathematical language needed to describe motion. Later modules will connect this description of motion to interactions.

All motion in this module is restricted to **one dimension**, with direction handled using **signed quantities**, not vectors.

We have briefly introduced the concept of speed and velocity when we introduced momentum. Here we will dig a little deeper into the formal definitions.

The concept that most people are familiar with is the speed. For example, the speedometer in a car shows how fast the car is moving. However, the speedometer does not provide *any* information about the direction the car is moving: **the speed is a scalar**.

In contrast, many people may use the the concept of velocity but not understanding the difference between speed and velocity. If we have information about how fast an object is moving *and* its direction, then we have the velocity: **the velocity is a vector**. We will discuss vectors in details in Phase C.

## D5.1.2 Position and Time

The **position** of a particle in one dimension is specified by a single coordinate $x$, measured relative to a chosen origin.

Position is:
- A signed quantity characterizing a quantity with direction.  
- Dependent on the choice of origin and axis direction. 
- A description of *where* an object is, not how it got there.

Time $t$ is treated as a continuously increasing parameter that labels when position is measured. Often we specifify that position is a function of time: $x(t)$.

```python
%reset -f

import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 4))

# Draw the reference axis (positive and negative directions)
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.006)
ax.quiver(0, 0, -1, 0, angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.006)

# Define two positions
x_positive = 0.8
x_negative = -0.6

# Plot the positions
ax.scatter(x_positive, 0, color='green', s=80, zorder=3)
ax.scatter(x_negative, 0, color='red', s=80, zorder=3)

# Label the positions with explicit values
ax.text(x_positive, 0.10, rf'$x = {x_positive} > 0$', ha='center', fontsize=12)
ax.text(x_negative, 0.10, rf'$x = {x_negative} < 0$', ha='center', fontsize=12)

# Origin
ax.scatter(0, 0, color='blue', s=50)
ax.text(-0.04, -0.12, 'O', fontsize=11)

# Add tick marks along the x-axis
ticks = np.arange(-1.0, 1.1, 0.5)
for t in ticks:
    ax.plot([t, t], [-0.03, 0.03], color='black', linewidth=1)
    if t != 0:
        ax.text(t, -0.10, f'{t}', ha='center', fontsize=10)

# Axis direction labels
ax.annotate('$+x$', (1.02, 0), fontsize=12)
ax.annotate('$-x$', (-1.18, 0), fontsize=12)

# Formatting
ax.set_xlim(-1.4, 1.4)
ax.set_ylim(-0.3, 0.3)
ax.set_aspect('equal', adjustable='box')
plt.axis('off')

# Title
plt.title("Positive and Negative Positions", fontsize=14)

plt.show()
```

## D5.1.3 Displacement

The **displacement** of a particle is defined as the change in position,

$$
\Delta x = x - x_0 .
$$

Displacement depends only on the initial and final positions and includes information about direction through its sign. 

```python
%reset -f

import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 4))

# Draw the reference axis (positive and negative directions)
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.006)
ax.quiver(0, 0, -1, 0, angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.006)

# Define initial and final positions
x_initial = -0.6
x_final = 0.8

# Vertical offset for clarity
y_offset = 0.15

# Plot initial and final positions on the axis
ax.scatter(x_initial, 0, color='purple', s=80, zorder=3)
ax.scatter(x_final, 0, color='green', s=80, zorder=3)

# Label initial and final positions
ax.text(x_initial, 0.08, rf'$x_0 = {x_initial}$', ha='center', fontsize=12)
ax.text(x_final, 0.08, rf'$x = {x_final}$', ha='center', fontsize=12)

# Draw displacement arrow ABOVE the axis
displacement = x_final - x_initial
ax.arrow(x_initial, y_offset, displacement, 0,
         length_includes_head=True,
         head_width=0.05, head_length=0.06,
         fc='black', ec='black', linewidth=2)

# Label displacement above the arrow
ax.text((x_initial + x_final)/2, y_offset + 0.08,
        rf'$\Delta x = x - x_0 = +{displacement}$',
        ha='center', fontsize=12)

# Origin
ax.scatter(0, 0, color='blue', s=50)
ax.text(-0.04, -0.18, 'O', fontsize=11)

# Add tick marks along the x-axis
ticks = np.arange(-1.0, 1.1, 0.5)
for t in ticks:
    ax.plot([t, t], [-0.03, 0.03], color='black', linewidth=1)
    if t != 0:
        ax.text(t, -0.24, f'{t}', ha='center', fontsize=10)

# Axis direction labels
ax.annotate('$+x$', (1.02, 0), fontsize=12)
ax.annotate('$-x$', (-1.18, 0), fontsize=12)

# Formatting
ax.set_xlim(-1.4, 1.4)
ax.set_ylim(-0.4, 0.45)
ax.set_aspect('equal', adjustable='box')
plt.axis('off')

# Title
plt.title("Displacement in One-Dimensional Motion", fontsize=14)

plt.show()
```

## D5.1.4 Speed and Velocity

**Speed** measures how fast something moves, without regard to direction.  
**Velocity** measures how fast and in which direction something moves.

In one dimension, velocity is represented as a signed quantity.

The **average speed** is defined as

$$
v_{\text{avg}} = \frac{\text{total distance traveled}}{\Delta t}.
$$

The **average velocity** is defined as

$$
v_{\text{avg},x} = \frac{\Delta x}{\Delta t}.
$$

where the subscript $x$ indicates a directed quantity for the velocity.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D5.1.5 Acceleration

Acceleration describes **how velocity changes in time**.

The **average acceleration** is defined as

$$
a_{\text{avg},x} = \frac{\Delta v_x}{\Delta t}.
$$

Acceleration can be positive or negative depending on how velocity changes, and it does not need to have the same sign as velocity.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D5.1.6 Reasoning About Motion in One Dimension

In one-dimensional motion, direction is handled using signs rather than vectors.

- A positive velocity indicates motion in the positive direction.  
- A negative velocity indicates motion in the negative direction.  
- The sign of acceleration indicates how velocity is changing over time.

Understanding motion in one dimension requires careful attention to sign conventions and physical interpretation rather than memorization of formulas.

## D5.1.7 Graphical Representations of Motion in One Dimension

In addition to describing motion with equations, we can represent motion **graphically**. Graphs provide a powerful way to visualize how position, velocity, and acceleration change in time, and they reveal relationships that are sometimes less obvious from equations alone.

We consider three fundamental graphs in one-dimensional motion:

- position vs. time  
- velocity vs. time  
- acceleration vs. time  

Throughout, time is plotted on the **horizontal axis**, and all quantities are treated as **signed**, consistent with our one-dimensional sign convention.

---

### Position vs. Time Graphs

A **positiontime graph** shows how an objects position $x$ changes as time progresses.

Key interpretations:

- The **slope** of a positiontime graph represents the objects **velocity**.
- A **steeper slope** means a larger magnitude of velocity.
- A **positive slope** corresponds to motion in the positive direction.
- A **negative slope** corresponds to motion in the negative direction.
- A **horizontal line** indicates zero velocity (the object is at rest).

If the slope is constant, the velocity is constant.  
If the slope changes with time, the velocity is changing and the object is accelerating.

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Example motion: position vs time (constant acceleration)
# --------------------------------------------------
# x(t) = x0 + v0 t + (1/2) a t^2
x0 = 2.0        # m
v0 = 1.5        # m/s
a  = 0.8        # m/s^2

t = np.linspace(0, 6, 400)
x = x0 + v0*t + 0.5*a*t**2

# --------------------------------------------------
# Choose a point to illustrate the slope (velocity)
# --------------------------------------------------
t0 = 3.0                               # s
x_t0 = x0 + v0*t0 + 0.5*a*t0**2        # position at t0
v_t0 = v0 + a*t0                       # slope = velocity at t0

# Tangent line at t0: x  x(t0) + v(t0)(t  t0)
t_tan = np.linspace(t0 - 1.2, t0 + 1.2, 100)
x_tan = x_t0 + v_t0*(t_tan - t0)

# --------------------------------------------------
# Plot
# --------------------------------------------------
plt.figure(figsize=(10, 6))

# Position vs time curve
plt.plot(t, x, linewidth=2, label="Position x(t)")

# Tangent line
plt.plot(t_tan, x_tan, linestyle="--", linewidth=2,
         label="Tangent line at t = 3.0 s")

# Mark the point of tangency
plt.scatter(t0, x_t0, zorder=5)

# Annotations
plt.text(t0 + 0.3, x_t0 + 0.5,
         f"Slope = v = {v_t0:.2f} m/s",
         fontsize=11)

plt.text(t0 , x_t0 - 1.2,
         "(t = 3.0 s)", fontsize=10)

# Labels and formatting
plt.xlabel("Time t (s)")
plt.ylabel("Position x (m)")
plt.title("Position vs Time with Tangent Line (Slope = Velocity)")
plt.legend()
plt.grid(True)

plt.show()
```

---

### Velocity vs. Time Graphs

A **velocitytime graph** shows how the velocity $v_x$ of an object changes with time.

Key interpretations:

- The **slope** of a velocitytime graph represents the objects **acceleration**.
- A **constant slope** indicates constant acceleration.
- A **horizontal line** indicates zero acceleration (constant velocity).
- The **sign of the velocity** tells us the direction of motion at each instant.

Equally important:

- The **area under a velocitytime graph** over a time interval represents the objects **displacement** during that interval.

This provides a graphical way to determine displacement without directly using positiontime equations.


```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Example motion: velocity vs time (constant acceleration)
# --------------------------------------------------
# v(t) = v0 + a t
v0 = 1.5        # m/s
a  = 0.8        # m/s^2

t = np.linspace(0, 6, 400)
v = v0 + a*t

# --------------------------------------------------
# Tangent line (slope = acceleration)
# --------------------------------------------------
t0 = 3.0                         # time where slope is illustrated
v_t0 = v0 + a*t0                 # velocity at t0

# Tangent line equation: v  v(t0) + a (t  t0)
t_tan = np.linspace(t0 - 1.2, t0 + 1.2, 100)
v_tan = v_t0 + a*(t_tan - t0)

# --------------------------------------------------
# Area under curve (displacement)
# --------------------------------------------------
t1 = 1.5
t2 = 4.5
mask = (t >= t1) & (t <= t2)

displacement = np.trapezoid(v[mask], t[mask])

# --------------------------------------------------
# Plot
# --------------------------------------------------
plt.figure(figsize=(10, 6))

# Velocity curve
plt.plot(t, v, linewidth=2, label="Velocity v(t)")

# Tangent line
plt.plot(t_tan, v_tan, linestyle="--", linewidth=2,
         label="Tangent line at t = 3.0 s")

# Point of tangency
plt.scatter(t0, v_t0, zorder=5)

# Shaded area (displacement)
plt.fill_between(t[mask], v[mask], alpha=0.3,
                 label="Area = displacement")

# Interval markers
plt.axvline(t1, linestyle=":")
plt.axvline(t2, linestyle=":")

# Annotations
plt.text(2, v_t0 + 0.7,
         f"Slope = a = {a:.2f} m/s",
         fontsize=11)

plt.text(2.0, 2.5,
         f"Area  x = {displacement:.2f} m",
         fontsize=11)

# Labels and formatting
plt.xlabel("Time t (s)")
plt.ylabel("Velocity v (m/s)")
plt.title("Velocity vs Time: Slope (Acceleration) and Area (Displacement)")
plt.legend()
plt.grid(True)

plt.show()
```

---

### Acceleration vs. Time Graphs

An **accelerationtime graph** shows how the acceleration $a_x$ changes with time.

Key interpretations:

- The **value** of the graph at any time gives the acceleration at that instant.
- A **horizontal line** indicates constant acceleration.
- A **changing curve** indicates variable acceleration.

Most importantly:

- The **area under an accelerationtime graph** over a time interval represents the **change in velocity** during that interval.

This connects acceleration directly to how velocity evolves in time.

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Example motion: acceleration vs time
# --------------------------------------------------
# Constant acceleration
a = 0.8   # m/s^2

t = np.linspace(0, 6, 400)
a_t = np.full_like(t, a)

# --------------------------------------------------
# Interval for area (change in velocity)
# --------------------------------------------------
t1 = 1.5
t2 = 4.5
mask = (t >= t1) & (t <= t2)

delta_v = np.trapezoid(a_t[mask], t[mask])

# --------------------------------------------------
# Plot
# --------------------------------------------------
plt.figure(figsize=(10, 6))

# Acceleration vs time
plt.plot(t, a_t, linewidth=2, label="Acceleration a(t)")

# Shaded area (change in velocity)
plt.fill_between(t[mask], a_t[mask], alpha=0.3,
                 label="Area = change in velocity")

# Interval markers
plt.axvline(t1, linestyle=":")
plt.axvline(t2, linestyle=":")

# Annotations
plt.text(2.0, 0.4,
         f"Area  v = {delta_v:.2f} m/s",
         fontsize=11)

plt.text(t1 - 0.35, 0.05, "t", fontsize=11)
plt.text(t2 - 0.35, 0.05, "t", fontsize=11)

# Labels and formatting
plt.xlabel("Time t (s)")
plt.ylabel("Acceleration a (m/s)")
plt.title("Acceleration vs Time with Shaded Area (Change in Velocity)")

plt.legend(loc="upper right")
plt.grid(True)

plt.show()
```

### Connecting the Three Graphs

These three graphical descriptions of motion are not independent; they are deeply connected:

- Velocity is the **slope** of the positiontime graph.
- Acceleration is the **slope** of the velocitytime graph.
- Displacement is the **area** under the velocitytime graph.
- Change in velocity is the **area** under the accelerationtime graph.

Understanding these relationships allows us to move fluently between graphical and algebraic descriptions of motion and prepares us for more general treatments introduced in later phases.

At this stage, all interpretations rely on **average slopes and areas over finite intervals**, consistent with our non-calculus approach in Phase D.

<div style="background-color:#fff8e1; border-left:5px solid #ffc20e; padding:12px; border-radius:4px;">

**Important Clarifications: Areas, Slopes, and Higher-Order Motion**

<strong>Area of a PositionTime Graph: Absement</strong>

The area under a positiontime graph <em>does</em> have a name: it is called <strong>absement</strong> (a contraction of <em>absolute displacement</em>). Absement is defined as the time integral of position. Its units are <strong>metersecond (ms)</strong>.

Conceptually, absement measures how far an object is displaced from a reference position <em>and for how long</em>.  
It weights displacement by duration, rather than describing motion itself.

Although absement is not part of standard introductory kinematics, it appears in several applied and engineering as well as biological systems context.

Examples of where absement is used include:
- analysis of hydraulic and pneumatic systems, where valve position held over time affects fluid delivery,
- vibration control and suspension systems,
- control theory, where accumulated position error over time matters,
- mechanical systems where sustained offset from equilibrium produces long-term effects.
- Fatigue in stretched systems: like metals, but also muscles and tendons.

In these applications, the question is not how fast is the object moving? but rather how long has the system been displaced from its desired position?

<br>

<strong>Why Absement Is Not Central in Kinematics</strong>

In kinematics, we focus on:
- how position changes,
- how velocity and acceleration describe motion,
- how forces and interactions alter motion.

For this purpose:
- the <strong>slope</strong> of the positiontime graph (velocity) is physically meaningful,
- but the <strong>area</strong> under the positiontime graph does not directly describe motion.

Absement does not tell us how an object moves; instead, it characterizes the cumulative effect of being displaced. This is why it is typically omitted from introductory treatments of motion, even though it is a legitimate physical quantity.

<br>

<strong>Slope of an AccelerationTime Graph</strong>

The slope of an accelerationtime graph represents the rate at which acceleration itself changes with time. This quantity is known as <strong>jerk</strong>.

- Jerk describes how smoothly (or abruptly) acceleration changes.
- Its units are meters per second cubed (m/s).

While jerk is not part of the core kinematic concepts in this module, it plays an important role in real-world systems.

Examples where jerk is relevant include:
- ride comfort in elevators, trains, and roller coasters,
- mechanical stress in robotic arms and machinery,
- vehicle suspension and road design,
- motion planning in aerospace and robotics.

In these contexts, large acceleration alone is not the main concern  rapid <em>changes</em> in acceleration can be more problematic.

<br>

<strong>Beyond Acceleration</strong>

In principle, motion can involve even higher-order time rates of change:
- change of jerk (sometimes called snap),
- followed by crackle and pop in playful terminology.

Although these higher-order quantities can be measured and modeled, their effects are often much smaller and are not treated as fundamental concepts in introductory physics. Instead, they become relevant in specialized engineering, biomechanics, and control-system contexts.

For this course, we focus on:
- position,
- velocity,
- acceleration,

because these quantities capture the essential structure of motion and connect directly to forces, energy, and momentum.

## D5.1.7 Summary

Kinematics provides a **descriptive framework** for motion using position, displacement, velocity, and acceleration.

These quantities allow us to describe how motion evolves in time, but they do not explain *why* motion changes.  
To make predictions about motion, kinematics must later be connected to interactions  a connection developed in Modules D5.2 and D5.3.
