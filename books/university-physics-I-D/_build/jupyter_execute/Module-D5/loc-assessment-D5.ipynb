## Problem 1: Kinematic Quantities in One Dimension

**LOC addressed:** A1, A2, A3, A5

A cyclist moves along a straight road that we model as the $x$-axis.

At time $t_0 = 0.00\ \text{s}$, the cyclist is at

$$
x_0 = -12.0\ \text{m}.
$$

At time $t_1 = 4.00\ \text{s}$, the cyclist is at

$$
x_1 = +6.0\ \text{m}.
$$

At time $t_2 = 7.00\ \text{s}$, the cyclist is at

$$
x_2 = -3.0\ \text{m}.
$$

1. Compute the displacement $\Delta x$ from $t_0$ to $t_2$.  
2. Compute the total distance traveled from $t_0$ to $t_2$.  
3. Compute the average velocity from $t_0$ to $t_2$.  
4. Compute the average speed from $t_0$ to $t_2$.  
5. In one or two sentences, explain why average speed and average velocity are not the same for this motion.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 1 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Displacement**

Displacement is the signed change in position:

$$
\Delta x = x_2 - x_0 = (-3.0) - (-12.0) = +9.0\ \text{m}.
$$

$$
\boxed{\Delta x = +9.0\ \text{m}}
$$

---

**Part 2 — Total distance traveled**

Distance traveled is the total path length:

Segment 1 (from $x_0$ to $x_1$):

$$
d_1 = |x_1-x_0| = |6.0-(-12.0)| = 18.0\ \text{m}.
$$

Segment 2 (from $x_1$ to $x_2$):

$$
d_2 = |x_2-x_1| = |-3.0-6.0| = 9.0\ \text{m}.
$$

Total distance:

$$
d_{\text{tot}} = d_1+d_2 = 18.0+9.0 = 27.0\ \text{m}.
$$

$$
\boxed{d_{\text{tot}} = 27.0\ \text{m}}
$$

---

**Part 3 — Average velocity**

Average velocity uses displacement divided by elapsed time:

$$
v_{\text{avg},x} = \frac{\Delta x}{\Delta t}
= \frac{+9.0}{7.00}
= +1.29\ \text{m/s}.
$$

$$
\boxed{v_{\text{avg},x} \approx +1.29\ \text{m/s}}
$$

---

**Part 4 — Average speed**

Average speed uses total distance divided by elapsed time:

$$
v_{\text{avg}} = \frac{d_{\text{tot}}}{\Delta t}
= \frac{27.0}{7.00}
= 3.86\ \text{m/s}.
$$

$$
\boxed{v_{\text{avg}} \approx 3.86\ \text{m/s}}
$$

---

**Part 5 — Conceptual explanation**

Average velocity depends on displacement (net change in position), while average speed depends on total distance traveled.  
Because the cyclist reverses direction, the distance traveled is larger than the magnitude of the displacement, so the average speed is larger than the magnitude of the average velocity.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
x0 = -12.0   # m
x1 =  6.0    # m
x2 = -3.0    # m

t0 = 0.00    # s
t1 = 4.00    # s
t2 = 7.00    # s

# Symbols (for clean expressions)
x0s, x1s, x2s, t0s, t2s = sp.symbols('x0 x1 x2 t0 t2', real=True)

# Displacement and elapsed time
Dx = x2s - x0s
Dt = t2s - t0s

# Distances on each segment
d1 = sp.Abs(x1 - x0)
d2 = sp.Abs(x2 - x1)
d_tot = d1 + d2

# Average velocity and speed
v_avg_x = Dx / Dt
s_avg = d_tot / Dt

# Substitute and evaluate
subs = {x0s: x0, x2s: x2, t0s: t0, t2s: t2}

Dx_num = sp.N(Dx.subs(subs), 6)
Dt_num = sp.N(Dt.subs(subs), 6)
vavg_num = sp.N(v_avg_x.subs(subs), 6)
savg_num = sp.N(s_avg.subs(subs), 6)

display(Math(r"\textbf{Problem 1}"))
display(Math(r"\Delta x = x_2 - x_0 = " + sp.latex(Dx_num) + r"\ \mathrm{m}"))
display(Math(r"\Delta t = t_2 - t_0 = " + sp.latex(Dt_num) + r"\ \mathrm{s}"))

display(Math(r"d_{\text{tot}} = |x_1-x_0| + |x_2-x_1| = " + sp.latex(sp.N(d_tot, 6)) + r"\ \mathrm{m}"))

display(Math(r"v_{\text{avg},x} = \frac{\Delta x}{\Delta t} = " + sp.latex(vavg_num) + r"\ \mathrm{m/s}"))
display(Math(r"v_{\text{avg}} = \frac{d_{\text{tot}}}{\Delta t} = " + sp.latex(savg_num) + r"\ \mathrm{m/s}"))
```

## Problem 2: Average Acceleration and Sign Conventions

**LOC addressed:** A4, A5

A cart moves along the $x$-axis. At time

$$
t_0 = 2.00\ \text{s},
$$

its velocity is

$$
v_{0,x} = -6.0\ \text{m/s}.
$$

At a later time

$$
t_f = 8.00\ \text{s},
$$

its velocity is

$$
v_{f,x} = +9.0\ \text{m/s}.
$$

1. Compute the change in velocity $\Delta v_x$.  
2. Compute the elapsed time $\Delta t$.  
3. Compute the average acceleration $a_x$ during this time interval.  
4. State the direction of the acceleration (positive or negative) and explain how you can tell from the signs.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 2 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant definition**

Average acceleration in 1D:

$$
a_x = \frac{\Delta v_x}{\Delta t}
= \frac{v_{f,x} - v_{0,x}}{t_f - t_0}.
$$

---

**Part 1 — Change in velocity**

$$
\Delta v_x = v_{f,x} - v_{0,x} = (+9.0) - (-6.0) = +15.0\ \text{m/s}.
$$

$$
\boxed{\Delta v_x = +15.0\ \text{m/s}}
$$

---

**Part 2 — Elapsed time**

$$
\Delta t = t_f - t_0 = 8.00 - 2.00 = 6.00\ \text{s}.
$$

$$
\boxed{\Delta t = 6.00\ \text{s}}
$$

---

**Part 3 — Average acceleration**

$$
a_x = \frac{\Delta v_x}{\Delta t}
= \frac{15.0}{6.00}
= 2.50\ \text{m/s}^2.
$$

$$
\boxed{a_x = +2.50\ \text{m/s}^2}
$$

---

**Part 4 — Direction and interpretation**

The acceleration is positive because the velocity increases from a negative value to a positive value, meaning $\Delta v_x$ is positive.  
A positive acceleration means the velocity is changing in the positive $x$-direction (becoming more positive over time).

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
t0 = 2.00     # s
tf = 8.00     # s
v0x = -6.0    # m/s
vfx = 9.0     # m/s

# Symbols
t0s, tfs, v0s, vfs = sp.symbols('t0 tf v0 vF', real=True)

# Definitions
Dv = vfs - v0s
Dt = tfs - t0s
ax = Dv / Dt

# Substitute values
subs = {t0s: t0, tfs: tf, v0s: v0x, vfs: vfx}

Dv_num = sp.N(Dv.subs(subs), 6)
Dt_num = sp.N(Dt.subs(subs), 6)
ax_num = sp.N(ax.subs(subs), 6)

# Display (MathJax-safe units)
display(Math(r"\textbf{Problem 2}"))
display(Math(r"\Delta v_x = v_{f,x}-v_{0,x} = " + sp.latex(Dv_num) + r"\ \mathrm{m/s}"))
display(Math(r"\Delta t = t_f-t_0 = " + sp.latex(Dt_num) + r"\ \mathrm{s}"))
display(Math(r"a_x = \frac{\Delta v_x}{\Delta t} = " + sp.latex(ax_num) + r"\ \mathrm{m/s^2}"))
```

## Problem 3: Constant-Acceleration Kinematics (Velocity–Time)

**LOC addressed:** B1, B2, C1

The train’s initial velocity is

$$
v_{0,x} = 0.0\ \text{m/s}.
$$

It accelerates at

$$
a_x = +1.10\ \text{m/s}^2
$$

for an elapsed time of

$$
t = 18.0\ \text{s}.
$$

1. State the key physical assumption that justifies using constant-acceleration kinematics in this problem.  
2. Compute the final velocity $v_{f,x}$ after 18.0 s.  
3. Using your result, state the direction of motion at $t = 18.0\ \text{s}$ (positive or negative).  

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 3 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Assumption**

The constant-acceleration equations apply only if the acceleration is approximately constant over the time interval being analyzed.  
Here we assume the train’s acceleration is constant from $t=0$ to $t=18.0\ \text{s}$.

---

**Part 2 — Final velocity**

Use the velocity–time relation for constant acceleration:

$$
v_{f,x} = v_{0,x} + a_x t.
$$

Substitute:

$$
v_{f,x} = 0.0 + (1.10)(18.0) = 19.8\ \text{m/s}.
$$

$$
\boxed{v_{f,x} = +19.8\ \text{m/s}}
$$

---

**Part 3 — Direction of motion**

The final velocity is positive, so the train is moving in the positive $x$-direction at $t=18.0\ \text{s}$.

</div>
</details>

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
v0x_val = 0.0     # m/s
ax_val  = 1.10    # m/s^2
t_val   = 18.0    # s

# Symbols
v0x, ax, t = sp.symbols('v0x ax t', real=True)

# Constant-acceleration velocity equation
vfx = v0x + ax*t

# Substitute and evaluate
vfx_num = sp.N(vfx.subs({v0x: v0x_val, ax: ax_val, t: t_val}), 6)

display(Math(r"\textbf{Problem 3}"))
display(Math(r"v_{f,x} = v_{0,x} + a_x t = " + sp.latex(vfx_num) + r"\ \mathrm{m/s}"))
```

## Problem 4: Constant-Acceleration Kinematics (Position–Time)

**LOC addressed:** B1, B3, C1

A test cart moves along the $x$-axis with approximately constant acceleration.

At time $t = 0$ s, the cart is at

$$
x_0 = -2.50\ \text{m}
$$

and has velocity

$$
v_{0,x} = +3.00\ \text{m/s}.
$$

The cart’s acceleration is

$$
a_x = -1.20\ \text{m/s}^2
$$

and is treated as constant for the next

$$
t = 5.00\ \text{s}.
$$

1. State the key physical assumption that justifies using the position–time kinematics equation.  
2. Compute the position $x$ of the cart at $t = 5.00\ \text{s}$.  
3. Based on the sign of $a_x$, explain whether the cart’s velocity becomes more positive or less positive during this interval.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 4 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Assumption**

The position–time kinematics equation applies only if the acceleration is approximately constant over the time interval.  
Here we assume $a_x$ is constant from $t=0$ to $t=5.00\ \text{s}$.

---

**Part 2 — Position at 5.00 s**

Use the position–time relation for constant acceleration:

$$
x = x_0 + v_{0,x}t + \tfrac{1}{2}a_x t^2.
$$

Substitute the values:

$$
x = (-2.50) + (3.00)(5.00) + \tfrac{1}{2}(-1.20)(5.00)^2.
$$

Compute each term:

- $v_{0,x}t = (3.00)(5.00) = 15.0\ \text{m}$
- $\tfrac{1}{2}a_x t^2 = 0.5(-1.20)(25.0) = -15.0\ \text{m}$

So,

$$
x = -2.50 + 15.0 - 15.0 = -2.50\ \text{m}.
$$

$$
\boxed{x(5.00\ \text{s}) = -2.50\ \text{m}}
$$

---

**Part 3 — Interpretation of the sign of acceleration**

Because $a_x$ is negative, it makes the velocity become less positive as time increases.  
In other words, the acceleration points in the negative $x$-direction, opposing the initial positive motion.

</div>
</details>

```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
x0_val  = -2.50   # m
v0x_val =  3.00   # m/s
ax_val  = -1.20   # m/s^2
t_val   =  5.00   # s

# Symbols
x0, v0x, ax, t = sp.symbols('x0 v0x ax t', real=True)

# Constant-acceleration position equation
x = x0 + v0x*t + sp.Rational(1, 2)*ax*t**2

# Substitute and evaluate
x_num = sp.N(x.subs({x0: x0_val, v0x: v0x_val, ax: ax_val, t: t_val}), 6)

display(Math(r"\textbf{Problem 4}"))
display(Math(r"x = x_0 + v_{0,x}t + \tfrac{1}{2}a_x t^2 = " + sp.latex(x_num) + r"\ \mathrm{m}"))
```

## Problem 5: Torricelli’s Equation (Time-Independent Kinematics)

**LOC addressed:** B1, B4, C1

A lab cart moves along the $x$-axis with approximately constant acceleration.

At position

$$
x_0 = 0.00\ \text{m},
$$

the cart has velocity

$$
v_{0,x} = +14.0\ \text{m/s}.
$$

It then experiences a constant acceleration

$$
a_x = -3.50\ \text{m/s}^2
$$

over a displacement of

$$
x - x_0 = +18.0\ \text{m}.
$$

1. State the key physical assumption that allows you to use Torricelli’s equation in this situation.  
2. Use Torricelli’s equation to compute the final velocity $v_{f,x}$ after the cart has moved 18.0 m.  
3. State whether the cart is still moving in the positive direction at that location, and justify your answer using the sign of $v_{f,x}$.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 5 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Assumption**

Torricelli’s equation is valid only if the acceleration is approximately constant over the displacement interval.  
Here we assume $a_x$ remains constant while the cart moves from $x_0$ to $x$.

---

**Part 2 — Final velocity**

Use Torricelli’s equation:

$$
v_{f,x}^2 = v_{0,x}^2 + 2a_x(x-x_0).
$$

Substitute:

$$
v_{f,x}^2 = (14.0)^2 + 2(-3.50)(18.0).
$$

Compute:

- $(14.0)^2 = 196$
- $2(-3.50)(18.0) = -126$

So,

$$
v_{f,x}^2 = 196 - 126 = 70.0.
$$

Take the square root:

$$
v_{f,x} = \pm\sqrt{70.0} = \pm 8.37\ \text{m/s}.
$$

Now determine the correct sign. The cart starts moving in the positive direction, and the acceleration is negative (slowing it down). Over this 18.0 m interval, it does not stop and reverse because stopping would require $v_{f,x}^2$ to reach zero at some point, but here $v_{f,x}^2$ is still positive (70.0). Therefore the cart is still moving in the positive direction.

$$
\boxed{v_{f,x} = +8.37\ \text{m/s}}
$$

---

**Part 3 — Direction of motion**

Because $v_{f,x}$ is positive, the cart is still moving in the positive $x$-direction at $x = 18.0\ \text{m}$.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
v0x_val = 14.0     # m/s
ax_val  = -3.50    # m/s^2
Dx_val  = 18.0     # m

# Symbols
v0x, vfx, ax, Dx = sp.symbols('v0x vfx ax Dx', real=True)

# Torricelli equation
eq = sp.Eq(vfx**2, v0x**2 + 2*ax*Dx)

# Solve for vfx
sol = sp.solve(eq.subs({v0x: v0x_val, ax: ax_val, Dx: Dx_val}), vfx)

# Choose physically consistent root (starts +, decelerates but does not reverse here)
vfx_num_pos = float([s for s in sol if s.evalf() > 0][0])

display(Math(r"\textbf{Problem 5}"))
display(Math(r"v_{f,x}^2 = v_{0,x}^2 + 2a_x(x-x_0)"))
display(Math(r"v_{f,x} = " + sp.latex(sp.N(vfx_num_pos, 6)) + r"\ \mathrm{m/s}"))
```

## Problem 6 (Expanded): Average vs. Piecewise Acceleration

**LOC addressed:** C1, C2, C3, B3

A student studies the motion of a cart along the $x$-axis. The cart is driven by a spring-like mechanism, so the net force (and acceleration) changes over time.

The cart’s velocity is measured at several instants:

- At $t = 0.0\ \text{s}$, $v_x = 0.0\ \text{m/s}$  
- At $t = 1.0\ \text{s}$, $v_x = +2.0\ \text{m/s}$  
- At $t = 2.0\ \text{s}$, $v_x = +3.0\ \text{m/s}$  
- At $t = 3.0\ \text{s}$, $v_x = +3.2\ \text{m/s}$  

Assume the cart starts at

$$
x_0 = 0.0\ \text{m}.
$$

1. Compute the **average acceleration** over the full interval from $t=0$ to $t=3.0\ \text{s}$.  
2. Using this single average acceleration, estimate the displacement of the cart at $t=3.0\ \text{s}$ using the position–time relation for constant acceleration.  
3. Compute the **average acceleration separately** over each 1.0 s interval.  
4. Using a **piecewise approach**, compute the cart’s displacement over each interval and determine the total displacement at $t=3.0\ \text{s}$.  
5. Compare the two displacement results and explain, in one or two sentences, which method is more physically reliable and why.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 6 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Average acceleration over 0–3 s**

$$
a_{\text{avg},x} = \frac{v_{3.0}-v_{0.0}}{3.0-0.0}
= \frac{3.2-0.0}{3.0}
= 1.07\ \text{m/s}^2.
$$

---

**Part 2 — Displacement using single average acceleration**

Using

$$
x = x_0 + v_{0,x}t + \tfrac{1}{2}a_x t^2,
$$

with $v_{0,x}=0$:

$$
x(3.0) = \tfrac{1}{2}(1.07)(3.0)^2
= 4.80\ \text{m}.
$$

---

**Part 3 — Average acceleration in each interval**

From 0–1 s:

$$
a_{01} = \frac{2.0-0.0}{1.0} = 2.0\ \text{m/s}^2
$$

From 1–2 s:

$$
a_{12} = \frac{3.0-2.0}{1.0} = 1.0\ \text{m/s}^2
$$

From 2–3 s:

$$
a_{23} = \frac{3.2-3.0}{1.0} = 0.20\ \text{m/s}^2
$$

---

**Part 4 — Piecewise displacement**

Interval 0–1 s:

$$
x_1 = 0 + (0)(1.0) + \tfrac{1}{2}(2.0)(1.0)^2 = 1.0\ \text{m}
$$

Interval 1–2 s (starting with $v=2.0$ m/s):

$$
x_2 = 1.0 + (2.0)(1.0) + \tfrac{1}{2}(1.0)(1.0)^2 = 3.5\ \text{m}
$$

Interval 2–3 s (starting with $v=3.0$ m/s):

$$
x_3 = 3.5 + (3.0)(1.0) + \tfrac{1}{2}(0.20)(1.0)^2 = 6.6\ \text{m}
$$

So,

$$
\boxed{x(3.0\ \text{s}) \approx 6.6\ \text{m}}
$$

---

**Part 5 — Interpretation**

The piecewise method is more physically reliable because it respects the changing acceleration.  
Using a single average acceleration masks important variations and significantly underestimates the displacement.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given data
t = [0.0, 1.0, 2.0, 3.0]       # s
v = [0.0, 2.0, 3.0, 3.2]       # m/s

# -------------------------
# Average acceleration over full interval
# -------------------------
a_avg = (v[-1] - v[0]) / (t[-1] - t[0])
x_avg = 0.5 * a_avg * (t[-1])**2

# -------------------------
# Piecewise accelerations
# -------------------------
a01 = (v[1] - v[0]) / 1.0
a12 = (v[2] - v[1]) / 1.0
a23 = (v[3] - v[2]) / 1.0

# Piecewise displacements
x1 = 0.5 * a01 * 1.0**2
x2 = x1 + v[1]*1.0 + 0.5 * a12 * 1.0**2
x3 = x2 + v[2]*1.0 + 0.5 * a23 * 1.0**2

display(Math(r"\textbf{Expanded Problem 6}"))

display(Math(r"a_{\text{avg}} = " + sp.latex(sp.N(a_avg, 6)) + r"\ \mathrm{m/s^2}"))
display(Math(r"x_{\text{avg}}(3.0\ \mathrm{s}) = " + sp.latex(sp.N(x_avg, 6)) + r"\ \mathrm{m}"))

display(Math(r"a_{01} = " + sp.latex(sp.N(a01, 6)) + r"\ \mathrm{m/s^2}"))
display(Math(r"a_{12} = " + sp.latex(sp.N(a12, 6)) + r"\ \mathrm{m/s^2}"))
display(Math(r"a_{23} = " + sp.latex(sp.N(a23, 6)) + r"\ \mathrm{m/s^2}"))

display(Math(r"x_{\text{piecewise}}(3.0\ \mathrm{s}) = " + sp.latex(sp.N(x3, 6)) + r"\ \mathrm{m}"))
```

## Problem 7: From Force to Motion — Newton’s Second Law and Prediction

**LOC addressed:** D1, D2, D3, E1, E3

A cart of mass

$$
m = 2.50\ \text{kg}
$$

moves along the $x$-axis on a low-friction track. Over a short time interval, two interactions act on the cart **simultaneously**:

Answer the following:

   $$

   $$

   $$
   t = 3.00\ \text{s}.
   $$

<details>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Net force**

Because both forces act over the same elapsed time, we may add them directly:

$$
\sum f_x = f_{1,x} + f_{2,x} = 18.0 - 6.0 = +12.0\ \text{N}.
$$

---

**Part 2 — Acceleration**

Using Newton’s second law,

$$
\sum f_x = m a_x,
$$

we find

$$
a_x = \frac{\sum f_x}{m}
= \frac{12.0}{2.50}
= 4.80\ \text{m/s}^2.
$$

$$
\boxed{a_x = +4.80\ \text{m/s}^2}
$$

---

**Part 3 — Velocity at 3.00 s**

Use the constant-acceleration velocity relation:

$$
v_{f,x} = v_{0,x} + a_x t.
$$

Substitute:

$$
v_{f,x} = 4.00 + (4.80)(3.00) = 18.4\ \text{m/s}.
$$

$$
\boxed{v_{f,x} = +18.4\ \text{m/s}}
$$

---

**Part 4 — Position at 3.00 s**

Use the position–time relation:

$$
x = x_0 + v_{0,x}t + \tfrac{1}{2}a_x t^2.
$$

$$
x = 0 + (4.00)(3.00) + \tfrac{1}{2}(4.80)(3.00)^2
= 33.6\ \text{m}.
$$

$$
\boxed{x(3.00\ \text{s}) = +33.6\ \text{m}}
$$

---

**Part 5 — Determinism**

Because the initial conditions ($x_0$, $v_{0,x}$) and the interaction (constant net force) are known, the cart’s motion is uniquely determined at all later times.  
This illustrates the deterministic nature of Newtonian motion within its domain of validity.

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# Given values
m_val   = 2.50     # kg
f1_val  = 18.0     # N
f2_val  = -6.0     # N
v0x_val = 4.00     # m/s
x0_val  = 0.00     # m
t_val   = 3.00     # s

# Symbols
m, f1, f2, v0x, x0, t = sp.symbols('m f1 f2 v0x x0 t', real=True)

# Net force, acceleration, velocity, position
Fx = f1 + f2
ax = Fx / m
vfx = v0x + ax*t
x = x0 + v0x*t + sp.Rational(1, 2)*ax*t**2

# Substitute values
subs = {
    m: m_val,
    f1: f1_val,
    f2: f2_val,
    v0x: v0x_val,
    x0: x0_val,
    t: t_val
}

Fx_num  = sp.N(Fx.subs(subs), 6)
ax_num  = sp.N(ax.subs(subs), 6)
vfx_num = sp.N(vfx.subs(subs), 6)
x_num   = sp.N(x.subs(subs), 6)

display(Math(r"\textbf{Problem 7}"))
display(Math(r"\sum f_x = " + sp.latex(Fx_num) + r"\ \mathrm{N}"))
display(Math(r"a_x = " + sp.latex(ax_num) + r"\ \mathrm{m/s^2}"))
display(Math(r"v_{f,x} = " + sp.latex(vfx_num) + r"\ \mathrm{m/s}"))
display(Math(r"x(3.00\ \mathrm{s}) = " + sp.latex(x_num) + r"\ \mathrm{m}"))
```

## Problem 8: Graph Interpretation — Constant Acceleration and Displacement

**LOC addressed:** A6, B1, B2, B3

A cart moves along the $x$-axis. Its velocity is recorded every 2.0 s and shown below:

| Time $t$ (s) | Velocity $v_{x}$ (m/s) |
|---:|---:|
| 0.0 | -4.0 |
| 2.0 | 0.0 |
| 4.0 | +4.0 |
| 6.0 | +8.0 |

Assume the motion between the listed points is well approximated by straight-line segments on a $v_x$ vs. $t$ graph.

1. Create a plot of $v_x$ vs. $t$.  
2. Use the slope of the $v_x$ vs. $t$ graph to determine the cart’s acceleration $a_x$.  
3. Use the area under the $v_x$ vs. $t$ graph from $t=0$ s to $t=6$ s to determine the displacement $\Delta x$.  
4. Using the displacement result and the given initial velocity, determine the cart’s position at $t=6$ s if $x_0 = +12.0\ \text{m}$.  
5. In one sentence, explain what it means physically that part of the area under the graph is negative.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 8 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Graph**

Plot the four points and connect them with straight segments.

---

**Part 2 — Acceleration from slope**

Use any two points, for example $(t,v_x)=(0,-4.0)$ and $(6.0,+8.0)$:

$$
a_x = \frac{\Delta v_x}{\Delta t}
= \frac{8.0-(-4.0)}{6.0-0.0}
= \frac{12.0}{6.0}
= 2.0\ \text{m/s}^2.
$$

$$
\boxed{a_x = +2.0\ \text{m/s}^2}
$$

This constant slope indicates constant acceleration.

---

**Part 3 — Displacement from area**

Displacement is the signed area under the $v_x$ vs. $t$ curve.

Break into three trapezoids (each 2.0 s wide).

From 0 to 2 s (from -4 to 0):

Average velocity is

$$
\bar v_x = \frac{-4.0+0.0}{2} = -2.0\ \text{m/s}.
$$

So,

$$
\Delta x_{0\to2} = \bar v_x\,\Delta t = (-2.0)(2.0) = -4.0\ \text{m}.
$$

From 2 to 4 s (from 0 to +4):

$$
\bar v_x = \frac{0.0+4.0}{2} = 2.0\ \text{m/s},
\quad
\Delta x_{2\to4} = (2.0)(2.0) = 4.0\ \text{m}.
$$

From 4 to 6 s (from +4 to +8):

$$
\bar v_x = \frac{4.0+8.0}{2} = 6.0\ \text{m/s},
\quad
\Delta x_{4\to6} = (6.0)(2.0) = 12.0\ \text{m}.
$$

Total displacement:

$$
\Delta x = (-4.0) + 4.0 + 12.0 = 12.0\ \text{m}.
$$

$$
\boxed{\Delta x_{0\to6} = +12.0\ \text{m}}
$$

---

**Part 4 — Final position**

$$
x(6.0) = x_0 + \Delta x = 12.0 + 12.0 = 24.0\ \text{m}.
$$

$$
\boxed{x(6.0\ \text{s}) = +24.0\ \text{m}}
$$

---

**Part 5 — Meaning of negative area**

Negative area corresponds to negative displacement, meaning the cart moves in the negative $x$-direction during that time interval.

</details>

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Data
t = np.array([0.0, 2.0, 4.0, 6.0])
v = np.array([-4.0, 0.0, 4.0, 8.0])

# Acceleration from slope (overall)
a = (v[-1] - v[0]) / (t[-1] - t[0])

# Displacement from area (trapezoid rule matches piecewise trapezoids)
dx = np.trapezoid(v, t)

# Given initial position
x0 = 12.0
xf = x0 + dx

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t, v, marker='o', linewidth=2, label="Velocity data")
plt.fill_between(t, v, alpha=0.3, label="Area = displacement")

plt.text(0.2, 7.0, f"Slope = a = {a:.2f} m/s²", fontsize=11)
plt.text(0.2, 6.1, f"Area ≈ Δx = {dx:.2f} m", fontsize=11)
plt.text(0.2, 5.2, f"x(6.0 s) = {xf:.2f} m", fontsize=11)

plt.xlabel("Time t (s)")
plt.ylabel("Velocity vₓ (m/s)")
plt.title("Assessment Problem: Velocity vs Time (Slope and Area)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
