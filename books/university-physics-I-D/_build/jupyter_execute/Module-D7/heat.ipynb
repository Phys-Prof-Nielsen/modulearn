## D7.3.1 Motivation
When two objects at different temperatures are placed in contact, energy flows from the hotter one to the colder one. This energy in transit is called **heat**.  
It is important to emphasize that **heat is not the same as temperature**:  
- Temperature is a measure of the thermal state of a system (related to particle motion).  
- Heat is the **energy transferred** because of a temperature difference.  

In thermodynamics, heat is treated in the same family as **work**.  
- **Mechanical work** is energy transferred by a force acting through a distance:
   
  $$
  W = f_x \cdot d_x
  $$
  
- **Heat** is energy transferred because of a **temperature difference**, often through microscopic collisions of particles.  

Both are forms of **energy transfer**, not stored substances. A system can have internal energy, but it does not "contain heat" any more than it "contains work." Instead, heat and work are ways energy moves in or out of a system.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.3.2 Heat and Specific Heat Capacity
When an object is heated and its temperature changes (but no phase change occurs), the amount of heat absorbed or released is proportional to its mass, its temperature change, and a material property called **specific heat capacity**:

$$
Q = m c \Delta T
$$

where  
- $Q$ = heat transferred (J)  
- $m$ = mass of the substance (kg)  
- $c$ = specific heat capacity ($\mathrm{J/(kg \cdot K)}$)  
- $\Delta T$ = temperature change (K or °C). Because temperature differences are independent of zero-point offsets, a change of 1 K is identical to a change of 1°C.


What happens to this energy?  
The energy goes primarily into the **microscopic motion of particles**:  
- Translational motion (particles moving faster)  
- Rotational and vibrational modes (in molecules)  
- Small contributions to electronic states at high temperatures  

This is why temperature rises — it reflects an increase in the average thermal energy associated with microscopic motion.

A material with a large $c$ (like water) requires more heat to achieve the same temperature change, because it can store more energy in these microscopic motions compared to a material with small $c$ (like metals).

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Heat Required to Warm Water (Specific Heat)</h3>

A camping pot contains

$$
m = 0.75\,\text{kg}
$$

of water initially at

$$
T_1 = 18^\circ\text{C}.
$$

It is heated to

$$
T_2 = 95^\circ\text{C}.
$$

Assume the specific heat capacity of water is

$$
c = 4186~\text{J/(kg$\cdot$ K)}.
$$

<br>

<strong>Calculation</strong><br>

Because no phase change occurs, the heat required is

$$
Q = mc\Delta T,
$$

with

$$
\Delta T = T_2 - T_1 = 95 - 18 = 77\,\text{K}.
$$

Substitute:

$$
Q = (0.75)(4186)(77) \approx 2.42\times 10^5\,\text{J}.
$$

<br>

<strong>Interpretation</strong><br>

The water absorbs about

$$
Q \approx 2.4\times 10^5\,\text{J} \approx 240\,\text{kJ}
$$

of energy to warm from $18^\circ$C to $95^\circ$C.  
Water has a relatively large specific heat, which is why it takes a substantial amount of energy to produce even moderate temperature increases.

<br>

<strong>Key takeaway</strong><br>

For a temperature change with no phase change, the required heat scales with:
- the mass $m$,
- the material property $c$,
- and the temperature change $\Delta T$:

$$
Q = mc\Delta T.
$$

## D7.3.3 Latent Heat
During a **phase change** (melting, freezing, boiling, condensing), temperature does **not** change even though heat is being transferred. The energy goes into rearranging molecular structure rather than changing kinetic energy.

The heat involved in a phase change is given by:

$$
Q = m L
$$

where  
- $Q$ = heat transferred (J)  
- $m$ = mass of the substance (kg)  
- $L$ = **latent heat** (J/kg), depending on the process  

Types of latent heat:  
- **Latent heat of fusion ($L_f$):** heat required to melt or freeze  
- **Latent heat of vaporization ($L_v$):** heat required to boil or condense  

---

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Latent Heat During a Phase Change</h3>

A mass of ice

$$
m = 0.30\,\text{kg}
$$

is at its melting temperature of

$$
0^\circ\text{C}.
$$

Heat is added to the ice until it completely melts, but the temperature remains constant throughout the process.

The latent heat of fusion for water is

$$
L_f = 3.34 \times 10^5\,\text{J/kg}.
$$

<br>

<strong>Calculation</strong><br>

During a phase change, the heat transferred is given by

$$
Q = mL_f.
$$

Substituting the values,

$$
Q = (0.30)(3.34 \times 10^5)
= 1.00 \times 10^5\,\text{J}.
$$

<br>

<strong>Interpretation</strong><br>

About

$$
Q \approx 100\,\text{kJ}
$$

of energy is required to melt the ice, even though its temperature does not increase.

This energy goes into **rearranging the molecular structure** of the water—from a solid lattice to a liquid—rather than increasing microscopic kinetic energy.

<br>

<strong>Key takeaway</strong><br>

Latent heat involves **large energy transfers without temperature change**.  
Temperature remains constant during a phase change, even as energy flows into or out of the system.

The **vaporization step** requires far more energy than any other stage.

This explains why boiling water takes so long and why steam carries enormous amounts of thermal energy.

<h3 style="margin-top:0; color:#000000;">Example — Latent Heat Release in a Growing Cumulonimbus Cloud</h3>

A cumulonimbus cloud can grow explosively because rising moist air cools and water vapor **condenses** into tiny liquid droplets.  
Condensation is the reverse of vaporization, so it releases **latent heat** into the surrounding air.

To estimate how large this energy release can be, consider a storm region with volume

$$
V = 1.0\,\text{km}^3 = 10^9\,\text{m}^3.
$$

A typical (order-of-magnitude) liquid water content inside an active cloud is about

$$
\text{LWC} \sim 1\,\text{g/m}^3 = 1\times10^{-3}\,\text{kg/m}^3.
$$

So the mass of condensed liquid water in that region is approximately

$$
m \approx (\text{LWC})V = (1\times10^{-3})(10^9) = 1\times10^6\,\text{kg}.
$$

Use the latent heat of vaporization for water,

$$
L_v = 2.26\times 10^6\,\text{J/kg}.
$$

<br>

<strong>Energy released by condensation</strong><br>

The magnitude of the latent heat released is

$$
|Q| = mL_v = (1\times10^6)(2.26\times10^6)
= 2.26\times10^{12}\,\text{J}.
$$

If we treat the condensing water vapor as the “system,” then heat flows out of that system during condensation, so

$$
Q \approx -2.26\times10^{12}\,\text{J}.
$$

(The atmosphere surrounding the droplets receives that energy.)

<br>

<strong>Interpretation</strong><br>

This is an enormous amount of energy for just a $1\,\text{km}^3$ region of cloud.  
For scale, using $1\,\text{ton TNT} \approx 4.184\times10^9\,\text{J}$,

$$
2.26\times10^{12}\,\text{J} \approx 540\,\text{tons of TNT}.
$$

This released energy warms the surrounding air, increases buoyancy, and helps drive strong updrafts and turbulence.  
A significant fraction of the storm’s energy is then redistributed through the atmosphere via **turbulence** and **atmospheric waves** (especially gravity waves), which is one reason flying near cumulonimbus clouds can be very unpleasant.

<br>

<strong>Key takeaway</strong><br>

Phase changes move *huge* amounts of energy.  
In thunderstorms, the latent heat released by condensation is a major energy source that powers intense motion and wave activity in the atmosphere.

## D7.3.5 Summary
- **Heat** is energy transferred due to temperature difference, just as **work** is energy transferred due to force acting through distance.  
- Neither heat nor work is “contained” in a system; they are **processes of energy transfer**.  
- **Specific heat** ($c$): energy needed to change temperature of a mass by $\Delta T$.  
- **Latent heat** ($L$): energy needed for a phase change, with no temperature change.  
- $Q = m c \Delta T$ for warming/cooling.  
- $Q = m L$ for phase changes.
