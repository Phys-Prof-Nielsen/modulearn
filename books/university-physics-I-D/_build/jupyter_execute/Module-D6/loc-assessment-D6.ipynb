## Problem 1: Spatial and Temporal Repetition in a Traveling Wave

**LOC addressed:** B4, B5, B6

A traveling wave on a string is described by the equation

$$
y(x,t) = 0.15\cos(4\pi x - 2\pi t),
$$

where $x$ is measured in meters and $t$ is measured in seconds.

Answer the following:

1. Identify the **wavenumber** $k$ and the **angular frequency** $\omega$.  
2. Compute the **wavelength** $\lambda$ of the wave.  
3. Compute the **period** $T$ of the wave.  
4. Compute the **frequency** $f$.  
5. Compute the **phase speed** $v$.  
6. In one or two sentences, explain how the spatial term and temporal term in the wave equation encode different physical information.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 1 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Identify $k$ and $\omega$**

Compare the given equation to the standard form

$$
y(x,t) = A\cos(kx - \omega t).
$$

From direct comparison:

$$
k = 4\pi\ \text{rad/m}, \qquad \omega = 2\pi\ \text{rad/s}.
$$

---

**Part 2 — Wavelength**

$$
\lambda = \frac{2\pi}{k}
= \frac{2\pi}{4\pi}
= 0.50\ \text{m}.
$$

$$
\boxed{\lambda = 0.50\ \text{m}}
$$

---

**Part 3 — Period**

$$
T = \frac{2\pi}{\omega}
= \frac{2\pi}{2\pi}
= 1.00\ \text{s}.
$$

$$
\boxed{T = 1.00\ \text{s}}
$$

---

**Part 4 — Frequency**

$$
f = \frac{1}{T}
= 1.00\ \text{Hz}.
$$

$$
\boxed{f = 1.00\ \text{Hz}}
$$

---

**Part 5 — Phase speed**

$$
v = f\lambda
= (1.00)(0.50)
= 0.50\ \text{m/s}.
$$

$$
\boxed{v = 0.50\ \text{m/s}}
$$

---

**Part 6 — Conceptual explanation**

The spatial term $kx$ determines how the wave pattern repeats in space (it sets $\lambda$), while the temporal term $\omega t$ determines how the motion repeats in time (it sets $T$ and $f$). Together they describe a repeating pattern that propagates through space.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
from IPython.display import display, Math

# Identified parameters from the wave equation
k = 4*np.pi          # rad/m
omega = 2*np.pi      # rad/s

# Derived quantities
lam = 2*np.pi / k
T = 2*np.pi / omega
f = 1 / T
v = f * lam

display(Math(r"\textbf{Identified parameters}"))
display(Math(rf"k = {k:.2f}\ \text{{rad/m}},\quad \omega = {omega:.2f}\ \text{{rad/s}}"))

display(Math(r"\textbf{Derived wave quantities}"))
display(Math(rf"\lambda = {lam:.2f}\ \text{{m}}"))
display(Math(rf"T = {T:.2f}\ \text{{s}}"))
display(Math(rf"f = {f:.2f}\ \text{{Hz}}"))
display(Math(rf"v = {v:.2f}\ \text{{m/s}}"))
```

## Problem 2: Phase Differences in a Traveling Wave

**LOC addressed:** C1, C2, B2

A traveling wave on a string is described by

$$
y(x,t) = A\cos(kx - \omega t),
$$

where $A$, $k$, and $\omega$ are positive constants.

Consider two fixed positions along the string:

$$
x_1 = 0.20~\text{m}, \qquad x_2 = 0.45~\text{m}.
$$

The wavelength of the wave is

$$
\lambda = 0.50~\text{m}.
$$

Answer the following:

1. Determine the **phase difference** $\Delta\phi$ between the oscillations at $x_1$ and $x_2$.  
2. State whether the two points are **in phase**, **out of phase**, or **neither**.  
3. Determine what fraction of a cycle the phase difference represents.  
4. In one or two sentences, explain what this phase difference means physically for the motion of the string at the two positions.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 2 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Phase difference**

For a traveling wave, the phase is

$$
\phi(x,t) = kx - \omega t.
$$

At a fixed time, the phase difference between two positions depends only on their separation:

$$
\Delta\phi = k(x_2 - x_1).
$$

First compute the wavenumber:

$$
k = \frac{2\pi}{\lambda}
= \frac{2\pi}{0.50}
= 4\pi~\text{rad/m}.
$$

Now compute the separation:

$$
\Delta x = x_2 - x_1 = 0.45 - 0.20 = 0.25~\text{m}.
$$

So the phase difference is

$$
\Delta\phi = (4\pi)(0.25) = \pi~\text{rad}.
$$

$$
\boxed{\Delta\phi = \pi~\text{rad}}
$$

---

**Part 2 — In phase or out of phase?**

A phase difference of $\pi$ radians corresponds to half of a full cycle.

Therefore, the two points are **out of phase**.

---

**Part 3 — Fraction of a cycle**

A full cycle corresponds to $2\pi$ radians. Thus,

$$
\frac{\Delta\phi}{2\pi} = \frac{\pi}{2\pi} = \frac{1}{2}.
$$

The two points differ by **half a cycle**.

---

**Part 4 — Physical interpretation**

When one point on the string is at a maximum positive displacement, the other point is at a maximum negative displacement. The two points oscillate with the same frequency but are always in opposite states of motion.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
from IPython.display import display, Math

# Given values
x1 = 0.20          # m
x2 = 0.45          # m
lam = 0.50         # m

# Compute wavenumber and phase difference
k = 2*np.pi / lam
delta_x = x2 - x1
delta_phi = k * delta_x

display(Math(r"\textbf{Given:}"))
display(Math(rf"x_1 = {x1:.2f}\ \text{{m}},\quad x_2 = {x2:.2f}\ \text{{m}}"))
display(Math(rf"\lambda = {lam:.2f}\ \text{{m}}"))

display(Math(r"\textbf{Computed quantities:}"))
display(Math(rf"k = \frac{{2\pi}}{{\lambda}} = {k:.2f}\ \text{{rad/m}}"))
display(Math(rf"\Delta\phi = k\Delta x = {delta_phi:.2f}\ \text{{rad}}"))

display(Math(r"\textbf{Fraction of a cycle:}"))
display(Math(rf"\frac{{\Delta\phi}}{{2\pi}} = {delta_phi/(2*np.pi):.2f}"))
```

## Problem 3: Interpreting Wave Graphs — Space vs. Time

**LOC addressed:** D1, D2, D3, D4

Two graphs of the *same traveling wave* are shown below:

- **Graph A:** Displacement $y$ versus position $x$ at a fixed time $t=t_0$.  
- **Graph B:** Displacement $y$ versus time $t$ at a fixed position $x=x_0$.

Answer the following questions.

1. For **Graph A**, identify what physical quantity is shown on each axis and state what is being held fixed.  
2. From **Graph A**, determine the **wavelength** $\lambda$ of the wave.  
3. For **Graph B**, identify what physical quantity is shown on each axis and state what is being held fixed.  
4. From **Graph B**, determine the **period** $T$ of the wave.  
5. Using your values of $\lambda$ and $T$, compute the **phase speed** $v$.  
6. In one or two sentences, explain why Graph A does *not* show particles moving along the string.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 3 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Interpreting Graph A**

Graph A shows **displacement $y$ versus position $x$**.

- The horizontal axis represents **position** along the string.
- The vertical axis represents **displacement** of the string.
- **Time is held fixed** at $t=t_0$.

This graph is a **spatial snapshot** of the wave.

---

**Part 2 — Wavelength**

The wavelength is the distance between repeating features (such as crests).

From Graph A, successive crests are separated by approximately

$$
\lambda \approx 0.60~\text{m}.
$$

$$
\boxed{\lambda \approx 0.60~\text{m}}
$$

---

**Part 3 — Interpreting Graph B**

Graph B shows **displacement $y$ versus time $t$**.

- The horizontal axis represents **time**.
- The vertical axis represents **displacement** at a fixed position $x=x_0$.
- **Position is held fixed**.

This graph is a **time trace** of the wave.

---

**Part 4 — Period**

The period is the time between repeating features in time.

From Graph B, successive crests occur approximately every

$$
T \approx 0.30~\text{s}.
$$

$$
\boxed{T \approx 0.30~\text{s}}
$$

---

**Part 5 — Phase speed**

The phase speed is

$$
v = \frac{\lambda}{T}
= \frac{0.60}{0.30}
= 2.0~\text{m/s}.
$$

$$
\boxed{v = 2.0~\text{m/s}}
$$

---

**Part 6 — Conceptual explanation**

Graph A does not show particles moving along the string. Instead, it shows how the *displacement of the string varies with position at a single instant in time*. The wave describes the motion of a repeating pattern, not the transport of matter.

</div>
</details>


```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Wave parameters (chosen to match Problem 3 values)
# -------------------------------------------------
A = 0.20              # amplitude (m)
lam = 0.60            # wavelength (m)
T = 0.30              # period (s)

k = 2*np.pi / lam     # wavenumber (rad/m)
omega = 2*np.pi / T  # angular frequency (rad/s)

# -------------------------------------------------
# Graph A: Spatial snapshot (time fixed)
# -------------------------------------------------
t0 = 0.0
x = np.linspace(0, 2.0, 1000)
y_snapshot = A * np.cos(k*x - omega*t0)

plt.figure(figsize=(8,4))
plt.plot(x, y_snapshot, label=r"$y(x,t_0)$ with $t_0=0$")
plt.axhline(0, linewidth=0.7)

# Mark one wavelength
plt.axvspan(0, lam, alpha=0.25, label="One wavelength")
plt.annotate("", xy=(0, -0.28), xytext=(lam, -0.28),
             arrowprops=dict(arrowstyle="<->"))
plt.text(lam/2, -0.33, r"$\lambda$", ha="center")

plt.xlabel("Position $x$ (m)")
plt.ylabel("Displacement $y$ (m)")
plt.title("Graph A — Spatial Snapshot (Time Fixed)")
plt.legend()
plt.grid(True)
plt.ylim(-0.35, 0.35)
plt.show()

# -------------------------------------------------
# Graph B: Time trace (position fixed)
# -------------------------------------------------
x0 = 0.30
t = np.linspace(0, 1.2, 1000)
y_time = A * np.cos(k*x0 - omega*t)

plt.figure(figsize=(8,4))
plt.plot(t, y_time, label=r"$y(x_0,t)$ with $x_0=0.30$ m")
plt.axhline(0, linewidth=0.7)

# Mark one period
plt.axvspan(0, T, alpha=0.25, label="One period")
plt.annotate("", xy=(0, -0.28), xytext=(T, -0.28),
             arrowprops=dict(arrowstyle="<->"))
plt.text(T/2, -0.33, r"$T$", ha="center")

plt.xlabel("Time $t$ (s)")
plt.ylabel("Displacement $y$ (m)")
plt.title("Graph B — Time Trace (Position Fixed)")
plt.legend()
plt.grid(True)
plt.ylim(-0.35, 0.35)
plt.show()
```

## Problem 4: Phase Constant and Initial Conditions in Oscillatory Motion

**LOC addressed:** A2, A3, A5

A particle undergoes oscillatory motion along the $x$-axis described by

$$
x(t) = A\cos(\omega t + \phi),
$$

where $A$, $\omega$, and $\phi$ are constants.

At time $t=0$, the particle is observed to be at its **maximum positive displacement**.

Answer the following:

1. What does it mean physically for the particle to be at its maximum positive displacement at $t=0$?  
2. What value of the **phase constant** $\phi$ is consistent with this observation?  
3. Suppose instead that at $t=0$ the particle is at **zero displacement and moving toward positive $x$**. What value of $\phi$ would describe this situation?  
4. In one or two sentences, explain the role of the phase constant in oscillatory motion.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 4 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Physical meaning of maximum displacement**

Maximum positive displacement means the particle is located at

$$
x = +A
$$

and is momentarily **not changing direction** at that instant. This corresponds to the crest of the oscillation.

---

**Part 2 — Phase constant for maximum displacement**

At $t=0$, the position is

$$
x(0) = A\cos(\phi).
$$

To have $x(0) = +A$, we require

$$
\cos(\phi) = 1.
$$

This occurs when

$$
\phi = 0.
$$

$$
\boxed{\phi = 0}
$$

---

**Part 3 — Zero displacement, moving toward positive $x$**

Zero displacement requires

$$
\cos(\phi) = 0.
$$

This occurs when

$$
\phi = -\frac{\pi}{2} \quad \text{or} \quad \phi = \frac{3\pi}{2}.
$$

These values correspond to the particle passing through equilibrium in the positive direction.

A valid choice is

$$
\boxed{\phi = -\frac{\pi}{2}}.
$$

---

**Part 4 — Conceptual explanation**

The phase constant determines the **state of the motion at $t=0$**. Different values of $\phi$ describe the same oscillation shifted in time, corresponding to different initial positions and directions of motion. See figure below for illustration.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0
omega = 2*np.pi     # rad/s
t = np.linspace(0, 1.0, 1000)

# Two different phase constants
phi_max = 0.0
phi_zero = -np.pi/2

# Oscillations
x_max = A * np.cos(omega*t + phi_max)
x_zero = A * np.cos(omega*t + phi_zero)

plt.figure(figsize=(8,4))
plt.plot(t, x_max, label=r"$\phi = 0$ (starts at $+A$)")
plt.plot(t, x_zero, label=r"$\phi = -\pi/2$ (starts at $x=0$, moving +)")
plt.axhline(0, linewidth=0.7)

plt.xlabel("Time $t$ (s)")
plt.ylabel("Displacement $x$")
plt.title("Effect of Phase Constant on Oscillatory Motion")
plt.legend()
plt.grid(True)
plt.show()
```

## Problem 5: Motion of Matter vs. Motion of a Pattern

**LOC addressed:** C3, E1

A traveling wave moves along a stretched string from left to right.

A student makes the following statement:

> “Since the wave travels to the right, points on the string must also move to the right as the wave passes.”

Answer the following:

1. Is this statement **correct or incorrect**?  
2. In one or two sentences, explain the **difference between motion of the wave pattern** and **motion of the string itself**.  
3. Describe the actual motion of a small marked point on the string as the wave passes.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 5 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Correct or incorrect**

The statement is **incorrect**.

---

**Part 2 — Pattern vs. matter**

The wave describes the motion of a **repeating pattern**, not the transport of matter.  
While the *pattern* (crests and troughs) moves to the right, the material points of the string do not travel with the wave.

---

**Part 3 — Motion of a point on the string**

A small marked point on the string oscillates **back and forth about its equilibrium position** as the wave passes. After the wave has passed, the point returns to its original location rather than being carried along with the wave.

---

**Key takeaway**

A traveling wave represents the **propagation of a disturbance**, not the bulk motion of matter. This distinction is essential for correctly interpreting wave motion.

## Problem 6: Interpreting and Plotting an Oscillation

**LOC addressed:** A2, A3, A4, A5, D1

An oscillatory motion along the $x$-axis is described by the expression

$$
x(t) = 0.25\cos(6\pi t - \tfrac{\pi}{3}),
$$

where $x$ is measured in meters and $t$ is measured in seconds.

Answer the following:

1. Identify the **amplitude** $A$.  
2. Identify the **angular frequency** $\omega$.  
3. Determine the **frequency** $f$ and **period** $T$.  
4. Identify the **phase constant** $\phi$ and describe what it tells you about the motion at $t=0$.  
5. Using Python, plot $x(t)$ for the time interval  

   $$
   0 \le t \le 1.0~\text{s}.
   $$
   Clearly label the axes and include units.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 6 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Part 1 — Amplitude**

From the coefficient in front of the cosine,

$$
\boxed{A = 0.25~\text{m}}.
$$

---

**Part 2 — Angular frequency**

Comparing to the standard form

$$
x(t) = A\cos(\omega t + \phi),
$$

we identify

$$
\boxed{\omega = 6\pi~\text{rad/s}}.
$$

---

**Part 3 — Frequency and period**

The frequency is

$$
f = \frac{\omega}{2\pi} = \frac{6\pi}{2\pi} = 3.0~\text{Hz}.
$$

The period is

$$
T = \frac{1}{f} = \frac{1}{3.0} = 0.333~\text{s}.
$$

$$
\boxed{f = 3.0~\text{Hz}}, \qquad \boxed{T = 0.333~\text{s}}.
$$

---

**Part 4 — Phase constant**

The phase constant is

$$
\boxed{\phi = -\tfrac{\pi}{3}}.
$$

This value specifies the **state of the oscillation at $t=0$**, meaning the particle does not start at maximum displacement but partway through its cycle.

---

**Part 5 — Plot interpretation**

The plot of $x(t)$ shows oscillatory motion repeating every $T = 0.333~\text{s}$ with maximum displacement $\pm 0.25~\text{m}$, consistent with the identified parameters.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Given parameters from the expression
A = 0.25
omega = 6*np.pi
phi = -np.pi/3

# Derived quantities
f = omega / (2*np.pi)
T = 1 / f

# Time array
t = np.linspace(0, 1.0, 1000)

# Oscillation
x = A * np.cos(omega*t + phi)

# Display identified parameters
display(Math(r"\textbf{Identified parameters}"))
display(Math(rf"A = {A:.2f}\ \text{{m}}"))
display(Math(rf"\omega = {omega:.2f}\ \text{{rad/s}}"))
display(Math(rf"f = {f:.2f}\ \text{{Hz}}"))
display(Math(rf"T = {T:.3f}\ \text{{s}}"))
display(Math(rf"\phi = -\frac{{\pi}}{{3}}"))

# Plot
plt.figure(figsize=(8,4))
plt.plot(t, x, label=r"$x(t)=0.25\cos(6\pi t-\pi/3)$")
plt.axhline(0, linewidth=0.7)

plt.xlabel("Time $t$ (s)")
plt.ylabel("Displacement $x$ (m)")
plt.title("Oscillatory Motion from a Given Expression")
plt.legend()
plt.grid(True)
plt.show()
```
