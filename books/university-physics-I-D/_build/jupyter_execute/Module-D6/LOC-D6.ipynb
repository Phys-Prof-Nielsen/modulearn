## Why this module matters

Many physical systems—ranging from mechanical vibrations and sound to light and quantum phenomena—exhibit **repetitive behavior**. This repetition can occur **in time**, as in oscillatory motion, or **in both space and time**, as in traveling waves.

In this module, we introduce **oscillations and traveling waves** as **patterns of motion**, characterized by a small set of descriptive parameters. Rather than explaining *why* these motions occur, we focus on **how they are described**, **recognized**, and **interpreted**.

We begin with **oscillatory motion**, introducing frequency-based descriptions that quantify repetition in time. We then extend these ideas to **traveling waves**, where repetition occurs across space as well as time. Throughout the module, motion is described using **given mathematical forms**, without derivation, emphasizing **physical meaning**, **parameter interpretation**, and **graphical understanding**.

This module prepares the conceptual ground for later phases, where calculus-based tools will be introduced to explain how these motions arise from underlying physical laws.

By the end of this module, you will be able to:

---

### **A. Oscillatory Motion**
*Describes repetitive motion in time.*

- **A1.** Identify **oscillatory motion** as motion that repeats in time about an equilibrium position.  
- **A2.** Define the **period** $T$ and **frequency** $f$, and explain their physical meanings.  
- **A3.** Define **angular frequency** $\omega$ and relate it to frequency using

  $$
  \omega = 2\pi f.
  $$

- **A4.** Define **amplitude** as the maximum displacement from equilibrium and describe its physical significance.  
- **A5.** Interpret the phase constant $\phi$ in a given oscillatory expression,

  $$
  x(t) = A\cos(\omega t + \phi),
  $$

  as specifying the state of the motion at $t = 0$, without deriving the equation.

---

### **B. Traveling Waves**
*Extends repetition from time to space and time.*

- **B1.** Identify a **traveling wave** as a pattern that propagates through space while maintaining its shape.  

- **B2.** Describe a traveling wave as being composed of **successive oscillations at different positions**, where each point in the medium undergoes oscillatory motion in time.  

- **B3.** Define **wavelength** $\lambda$ as the spatial period of a wave.  

- **B4.** Define **wavenumber** $k$ and relate it to wavelength using

  $$
  k = \frac{2\pi}{\lambda}.
  $$

- **B5.** Define **phase speed** $v$ as the speed at which a point of constant phase propagates through space, and relate it to wavelength and frequency using

  $$
  v = f\lambda.
  $$

- **B6.** Interpret a given traveling-wave expression,

  $$
  y(x,t) = A\sin(kx - \omega t),
  $$

  by identifying the physical meaning of each parameter and explaining how the wave represents oscillatory motion occurring at many positions, without deriving the equation.


---

### **C. Phase and Pattern Interpretation**
*Connects oscillations and waves through a shared conceptual language.*

- **C1.** Define **phase** as the argument of a sinusoidal function that specifies the state of motion.  
- **C2.** Explain what it means for two points in space or time to be **in phase** or **out of phase**.  
- **C3.** Distinguish between **motion of a pattern** (wave propagation) and **motion of matter**, recognizing that the wave describes a repeating disturbance rather than bulk transport.

---

### **D. Graphical Representations**
*Develops literacy in time- and space-based motion graphs.*

- **D1.** Interpret **displacement–time graphs** for oscillatory motion by identifying amplitude, period, and phase.  
- **D2.** Interpret **spatial snapshots** of traveling waves by identifying wavelength and amplitude.  
- **D3.** Distinguish between graphs that represent **motion in time** and those that represent **structure in space**.  
- **D4.** Explain how the same wave can be represented differently depending on whether time or space is held fixed.

---

### **E. Scope and Preparation for Later Phases**
*Clarifies what this description does—and does not—explain.*

- **E1.** Recognize that oscillation and wave equations in this module are **descriptive**, not derived.  
- **E2.** Identify the limitations of this approach and explain why understanding *why* these motions occur requires calculus-based tools introduced in Phase B.  
- **E3.** Describe this module as a **conceptual foundation** for later studies of dynamics, wave mechanics, and field theories.

---

### **Performance Criteria**

- Oscillatory and wave motion are described using **clear physical definitions**, not memorized formulas.  
- Parameters such as $A$, $f$, $\omega$, $\lambda$, $k$, and $v$ are interpreted **physically and graphically**.  
- Given mathematical expressions are treated as **models to be understood**, not results to be derived.  
- Explanations emphasize **patterns, repetition, and meaning**, rather than causes or forces.  
- Graphs are interpreted correctly with respect to **what varies** and **what is held fixed**.
