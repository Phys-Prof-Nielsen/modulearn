## Problem D6.1 — Mass–Spring SHM (Undamped)

A mass $m = 0.50~\mathrm{kg}$ is attached to a spring with stiffness  

$$
k = 200.0~\mathrm{N/m}.
$$

The mass is released from rest at a displacement  

$$
x_0 = 0.10~\mathrm{m},
$$

so the initial velocity is zero.  

The angular frequency for a mass–spring system undergoing simple harmonic motion is given by

$$
\omega = \sqrt{\frac{k}{m}}.
$$

Answer the following:

1. What is the **angular frequency** $\omega$?  
2. What is the **period** $T$ of the motion?  
3. What is the **frequency** $f$?

Be careful with **units** and **significant figures**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Angular frequency for a mass–spring oscillator:

$$
\omega = \sqrt{\frac{k}{m}}
$$

Relationships between angular frequency, period, and frequency:

$$
\omega = 2\pi f, \qquad T = \frac{1}{f}.
$$

---

**Part 1 — Angular frequency**

**Given**
- $k = 200.0~\mathrm{N/m}$
- $m = 0.50~\mathrm{kg}$

**Compute $\omega$**

$$
\omega = \sqrt{\frac{200.0}{0.50}}
= \sqrt{400}
= 20.0~\mathrm{rad/s}.
$$

With proper significant figures:

$$
\boxed{\omega = 20.0~\mathrm{rad/s}}
$$

---

**Part 2 — Period**

Use the relationship between period and angular frequency:

$$
T = \frac{2\pi}{\omega}.
$$

Substitute the value of $\omega$:

$$
T = \frac{2\pi}{20.0}
= 0.314~\mathrm{s}.
$$

With proper significant figures:

$$
\boxed{T = 0.314~\mathrm{s}}
$$

---

**Part 3 — Frequency**

The frequency is the inverse of the period:

$$
f = \frac{1}{T}.
$$

Substitute the value of $T$:

$$
f = \frac{1}{0.314}
= 3.18~\mathrm{Hz}.
$$

Equivalently, using $\omega = 2\pi f$:

$$
f = \frac{\omega}{2\pi} = \frac{20.0}{2\pi} = 3.18~\mathrm{Hz}.
$$

With proper significant figures:

$$
\boxed{f = 3.18~\mathrm{Hz}}
$$

---

**Final Summary**

- The **angular frequency** depends only on the system parameters $k$ and $m$, not on amplitude.  
- The **period** tells how long one complete oscillation takes.  
- The **frequency** tells how many oscillations occur each second.

These quantities describe the **timing of the oscillation**, independent of how far the mass is displaced.
</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
import numpy as np
from IPython.display import Markdown, display, Math

def fmt(x, sig=2):
    try:
        return f"{float(x):.{sig}g}"
    except Exception:
        return str(x)

# Symbols and expressions
m, k, t = sp.symbols('m k t', positive=True, real=True)
omega = sp.sqrt(k/m)
T = 2*sp.pi/omega
f = omega/(2*sp.pi)

# Numbers
vals = {m:0.50, k:200.0}
omega_num = omega.subs(vals).evalf()
T_num = T.subs(vals).evalf()
f_num = f.subs(vals).evalf()

display(Math(r"$$\omega=\sqrt{\frac{k}{m}},\quad T=\frac{2\pi}{\omega},\quad f=\frac{1}{T}=\frac{\omega}{2\pi}.$$"))
print(f"Numeric: ω ≈ {fmt(omega_num)} rad/s,  T ≈ {fmt(T_num)} s,  f ≈ {fmt(f_num)} Hz")
```

## Problem D6.3 — Traveling Wave Parameters

A periodic disturbance travels along a string with

$$
\lambda = 0.80~\mathrm{m}, \qquad f = 5.0~\mathrm{Hz}.
$$

Answer the following:

1. Find the **wavenumber** $k$.  
2. Find the **period** $T$.  
3. Find the **phase speed** $v_p$.

Be careful with **units** and **significant figures**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Wavenumber:

$$
k = \frac{2\pi}{\lambda}
$$

Period–frequency relationship:

$$
T = \frac{1}{f}
$$

Phase speed:

$$
v_p = f\lambda
$$

---

**Part 1 — Wavenumber**

**Given**
- $\lambda = 0.80~\mathrm{m}$

**Compute $k$**

$$
k = \frac{2\pi}{0.80}
= 7.85~\mathrm{rad/m}.
$$

With proper significant figures:

$$
\boxed{k = 7.9~\mathrm{rad/m}}
$$

---

**Part 2 — Period**

**Given**
- $f = 5.0~\mathrm{Hz}$

**Compute $T$**

$$
T = \frac{1}{5.0}
= 0.20~\mathrm{s}.
$$

With proper significant figures:

$$
\boxed{T = 0.20~\mathrm{s}}
$$

---

**Part 3 — Phase speed**

**Compute $v_p$**

$$
v_p = (5.0)(0.80)
= 4.0~\mathrm{m/s}.
$$

With proper significant figures:

$$
\boxed{v_p = 4.0~\mathrm{m/s}}
$$

---

**Final Summary**

- The **wavenumber** describes how rapidly the wave pattern repeats in space.  
- The **period** describes how often the oscillation repeats in time.  
- The **phase speed** tells how fast the **wave pattern** moves along the string.

These quantities together fully characterize the basic timing and spatial structure of a traveling wave.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp
import numpy as np
from IPython.display import Markdown, display, Math

def fmt(x, sig=2):
    try:
        return f"{float(x):.{sig}g}"
    except Exception:
        return str(x)

lam, f = sp.symbols('lam f', positive=True, real=True)

# Definitions
k = 2*sp.pi/lam
T = 1/f
vp = f*lam  # also vp = (2*pi/T)/(2*pi/lam) = lam/T

# Numbers
vals = {lam:0.80, f:5.0}

k_num  = sp.N(k.subs(vals))
T_num  = sp.N(T.subs(vals))
vp_num = sp.N(vp.subs(vals))

display(Math(r"$$k=\frac{2\pi}{\lambda},\qquad T=\frac{1}{f},\qquad v_p=f\,\lambda.$$"))
print(f"k ≈ {fmt(k_num)} rad/m")
print(f"T ≈ {fmt(T_num)} s")
print(f"v_p ≈ {fmt(vp_num)} m/s")
```

```python
%reset -f 

import sympy as sp
import numpy as np
from IPython.display import Markdown, display, Math

def fmt(x, sig=2):
    try:
        return f"{float(x):.{sig}g}"
    except Exception:
        return str(x)

k, T = sp.symbols('k T', positive=True, real=True)

# Definitions
lam = 2*sp.pi/k
f   = 1/T
vp  = lam/T     # also vp = (2*pi/T)/k

# Numbers
vals = {k:12.0, T:0.40}

lam_num = sp.N(lam.subs(vals))
f_num   = sp.N(f.subs(vals))
vp_num  = sp.N(vp.subs(vals))

display(Math(r"$$\lambda=\frac{2\pi}{k},\qquad f=\frac{1}{T},\qquad v_p=\frac{\lambda}{T}=\frac{2\pi}{k\,T}.$$"))
print(f"λ ≈ {fmt(lam_num)} m")
print(f"f ≈ {fmt(f_num)} Hz")
print(f"v_p ≈ {fmt(vp_num)} m/s")
```

```python
%reset -f
import numpy as np
from IPython.display import Markdown, display, Math

A = 0.8             # amplitude (units of x)
f_true = 1.50       # Hz (do not reveal to students; they should read from the graph)
phi = 0.35          # phase offset (rad)

# Time base
t = np.linspace(0, 3.2, 1200)  # a bit over 3 seconds for multiple cycles
x = A * np.cos(2*np.pi*f_true*t + phi)

# --- Peak detection: simple local-maximum finder ---
# A point i is a peak if x[i-1] < x[i] > x[i+1]
idx = np.where((x[1:-1] > x[:-2]) & (x[1:-1] > x[2:]))[0] + 1
t_peaks = t[idx]

# Use interior peaks to avoid edge effects, then compute average spacing
if len(t_peaks) >= 3:
    T_estimates = np.diff(t_peaks)         # successive peak spacings
    T_est = T_estimates.mean()
    f_est = 1.0 / T_est
else:
    T_est = np.nan
    f_est = np.nan

# Nicely formatted output
display(Markdown(r"""Read the period *T* as the time between **successive peaks** on the plot (average over several cycles).  
Then compute the frequency using
"""))
display(Math(r"$$f=\frac{1}{T}.$$"))

print(f"Estimated period from graph (avg peak spacing): T ≈ {T_est:.1e} s")
print(f"Estimated frequency: f ≈ {f_est:.1e} Hz")
```

## Problem D6.6 — Read $\lambda$ and $T$ from Traveling-Wave Graphs

Two figures are provided:

1) A **spatial snapshot** $y(x,t_0)$ at a fixed time $t_0$ (displacement vs. position).  
2) A **time trace** $y(x_0,t)$ at a fixed position $x_0$ (displacement vs. time).

**Tasks:**

1) From the spatial snapshot, estimate the **wavelength** $\lambda$ by measuring the horizontal distance between successive crests (average across several).  
2) From the time trace, estimate the **period** $T$ by measuring the time between successive crests (average across several).  
3) Compute the **frequency**:

$$
f=\frac{1}{T}.
$$

4) Compute the **phase speed**:

$$
v_p=\frac{\lambda}{T}.
$$

*Report your answers with units. Show your reading strategy (which peaks you chose) and your averaging.*

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Key ideas**

- In a **spatial snapshot** $y(x,t_0)$, time is fixed and the pattern repeats in **space**. The horizontal spacing between repeating features (such as crests) is the **wavelength** $\lambda$.  
- In a **time trace** $y(x_0,t)$, position is fixed and the motion repeats in **time**. The time spacing between repeating features is the **period** $T$.  
- Once $\lambda$ and $T$ are measured, the frequency and phase speed follow from

$$
f=\frac{1}{T},\qquad v_p=\frac{\lambda}{T}.
$$

---

**Part 1 — Wavelength from the spatial snapshot**

Choose several crests on the $y(x,t_0)$ graph. For example, suppose crests occur at approximately

$$
x = 0.40~\text{m},\ 1.20~\text{m},\ 2.00~\text{m},\ 2.80~\text{m}.
$$

Compute successive crest-to-crest spacings:

$$
\Delta x_1 = 1.20-0.40 = 0.80~\text{m}
$$
$$
\Delta x_2 = 2.00-1.20 = 0.80~\text{m}
$$
$$
\Delta x_3 = 2.80-2.00 = 0.80~\text{m}
$$

Average these values to estimate $\lambda$:

$$
\lambda \approx \frac{\Delta x_1+\Delta x_2+\Delta x_3}{3} = 0.80~\text{m}.
$$

So,

$$
\boxed{\lambda \approx 0.80~\text{m}}.
$$

---

**Part 2 — Period from the time trace**

Now choose several crests on the $y(x_0,t)$ graph. For example, suppose crests occur at approximately

$$
t = 0.30~\text{s},\ 0.70~\text{s},\ 1.10~\text{s},\ 1.50~\text{s}.
$$

Compute successive time spacings:

$$
\Delta t_1 = 0.70-0.30 = 0.40~\text{s}
$$
$$
\Delta t_2 = 1.10-0.70 = 0.40~\text{s}
$$
$$
\Delta t_3 = 1.50-1.10 = 0.40~\text{s}
$$

Average these values to estimate $T$:

$$
T \approx \frac{\Delta t_1+\Delta t_2+\Delta t_3}{3} = 0.40~\text{s}.
$$

So,

$$
\boxed{T \approx 0.40~\text{s}}.
$$

---

**Part 3 — Frequency**

Compute

$$
f=\frac{1}{T}=\frac{1}{0.40}=2.5~\text{Hz}.
$$

So,

$$
\boxed{f \approx 2.5~\text{Hz}}.
$$

---

**Part 4 — Phase speed**

Compute

$$
v_p=\frac{\lambda}{T}=\frac{0.80}{0.40}=2.0~\text{m/s}.
$$

So,

$$
\boxed{v_p \approx 2.0~\text{m/s}}.
$$

---

**Final Summary**

- The **wavelength** is read from the *spatial repetition* in $y(x,t_0)$.  
- The **period** is read from the *temporal repetition* in $y(x_0,t)$.  
- Frequency and phase speed follow directly from $f=1/T$ and $v_p=\lambda/T$.  

Because the measurements are read from a graph, results should reflect reasonable precision based on the axis scales.

</div>
</details>


```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

A = 1.0            # amplitude (arb. units)
lam_true = 0.75    # wavelength (m)
f_true = 2.0       # frequency (Hz)
T_true = 1.0 / f_true
vp_true = lam_true * f_true


# Spatial snapshot y(x, t0)
t0 = 0.10
x = np.linspace(0, 3*lam_true, 1200)
y_space = A * np.cos(2*np.pi*(x/lam_true) - 2*np.pi*f_true*t0)

plt.figure(figsize=(8,4))
plt.plot(x, y_space, linewidth=2)
plt.xlabel("Position $x$ (m)")
plt.ylabel("$y(x,t_0)$ (arb. units)")
plt.title("Spatial Snapshot $y(x,t_0)$ — Estimate Wavelength $\\lambda$")
plt.grid(True, alpha=0.4)
plt.xlim(x.min(), x.max())
plt.show()

# Time trace y(x0, t)
x0 = 0.60
t = np.linspace(0, 2.2, 1500)  # ~4+ cycles
y_time = A * np.cos(2*np.pi*(x0/lam_true) - 2*np.pi*f_true*t)

plt.figure(figsize=(8,4))
plt.plot(t, y_time, linewidth=2)
plt.xlabel("Time $t$ (s)")
plt.ylabel("$y(x_0,t)$ (arb. units)")
plt.title("Time Trace $y(x_0,t)$ — Estimate Period $T$")
plt.grid(True, alpha=0.4)
plt.xlim(t.min(), t.max())
plt.show()
```

```python
%reset -f

import numpy as np
from IPython.display import Markdown, display, Math

A = 1.0            # amplitude (arb. units)
lam_true = 0.75    # wavelength (m)
f_true = 2.0       # frequency (Hz)
T_true = 1.0 / f_true
vp_true = lam_true * f_true


# Spatial snapshot y(x, t0)
t0 = 0.10
x = np.linspace(0, 3*lam_true, 1200)
y_space = A * np.cos(2*np.pi*(x/lam_true) - 2*np.pi*f_true*t0)

# Time trace y(x0, t)
x0 = 0.60
t = np.linspace(0, 2.2, 1500)  # ~4+ cycles
y_time = A * np.cos(2*np.pi*(x0/lam_true) - 2*np.pi*f_true*t)

def find_peak_positions(xvals, yvals):
    """Return x-positions of simple local maxima (exclude endpoints)."""
    idx = np.where((yvals[1:-1] > yvals[:-2]) & (yvals[1:-1] > yvals[2:]))[0] + 1
    return xvals[idx]

# 1) Wavelength from spatial snapshot
x_peaks = find_peak_positions(x, y_space)
# Use interior peaks and average spacing
lam_estimates = np.diff(x_peaks)
lam_est = np.mean(lam_estimates) if len(lam_estimates) >= 1 else np.nan

# 2) Period from time trace
t_peaks = find_peak_positions(t, y_time)
T_estimates = np.diff(t_peaks)
T_est = np.mean(T_estimates) if len(T_estimates) >= 1 else np.nan

# 3) Frequency and 4) Phase speed
f_est = 1.0 / T_est if np.isfinite(T_est) and T_est > 0 else np.nan
vp_est = lam_est / T_est if np.isfinite(lam_est) and np.isfinite(T_est) and T_est > 0 else np.nan

display(Markdown(r""" 
- From the **spatial** plot, measure several crest-to-crest spacings and average to get wavelength. 
- From the **time** plot, measure several peak-to-peak spacings and average to get period.  
- Then compute
"""))
display(Math(r"$$f=\frac{1}{T},\qquad v_p=\frac{\lambda}{T}.$$"))

print(f"Estimated wavelength  λ ≈ {lam_est:.1e} m")
print(f"Estimated period     T ≈ {T_est:.1e} s")
print(f"Estimated frequency  f ≈ {f_est:.1e} Hz")
print(f"Estimated phase speed v_p ≈ {vp_est:.1e} m/s")
```

## Problem D6.7 🐍💻 — Plotting a Traveling Wave: Snapshot vs. Time Trace

A traveling wave along a string is described by

$$
y(x,t) = 0.20\cos(2\pi x - 4\pi t),
$$

where $x$ is measured in meters and $t$ is measured in seconds.

You will use Python to generate two different plots of this same wave and interpret what each plot represents.

---

**Tasks:**

1) Create a **spatial snapshot** $y(x,t_0)$ by holding time fixed at  

   $$
   t_0 = 0.
   $$
 
   Plot displacement versus position for  

   $$
   0 \le x \le 2.0~\text{m}.
   $$

2) Create a **time trace** $y(x_0,t)$ by holding position fixed at  

   $$
   x_0 = 0.25~\text{m}.
   $$

   Plot displacement versus time for  

   $$
   0 \le t \le 1.0~\text{s}.
   $$

3) From your **spatial snapshot**, estimate the **wavelength** $\lambda$.  
4) From your **time trace**, estimate the **period** $T$.  
5) Use your measured values to compute the **phase speed**  

$$
v_p=\frac{\lambda}{T}.
$$

Label your axes clearly and include units.

**Key idea**

The same wave can appear very different depending on what is held fixed:
- Holding **time fixed** shows the wave’s *spatial structure*.
- Holding **position fixed** shows the wave’s *oscillatory motion in time*.

---

**Part 1 — Spatial snapshot**

Set $t_0=0$. The wave becomes

$$
y(x,0)=0.20\cos(2\pi x).
$$

Plot this expression versus $x$.  
From the plot, successive crests are separated by approximately

$$
\lambda \approx 1.0~\text{m}.
$$

---

**Part 2 — Time trace**

Set $x_0=0.25~\text{m}$. The wave becomes

$$
y(0.25,t)=0.20\cos\!\big(2\pi(0.25)-4\pi t\big).
$$

Plot this expression versus $t$.  
From the plot, successive crests occur every

$$
T \approx 0.50~\text{s}.
$$

---

**Part 3 — Phase speed**

Using the measured values,

$$
v_p=\frac{\lambda}{T}
=\frac{1.0}{0.50}
=2.0~\text{m/s}.
$$

So,

$$
\boxed{v_p \approx 2.0~\text{m/s}}.
$$

---

**Final Summary**

- A **snapshot** reveals spatial repetition and wavelength.  
- A **time trace** reveals temporal repetition and period.  
- The **phase speed** connects these two perspectives and describes how fast the wave pattern moves.

This problem emphasizes that graphs of waves show **patterns**, not particles moving through space.

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# ----------------------------
# Wave parameters (given)
# ----------------------------
A = 0.20
k = 2*np.pi        # rad/m
omega = 4*np.pi    # rad/s

# For exact (model-based) results:
lam_exact = 2*np.pi / k
T_exact = 2*np.pi / omega
f_exact = 1 / T_exact
vp_exact = lam_exact / T_exact

display(Math(r"\textbf{Given wave:}\quad y(x,t)=0.20\cos(2\pi x-4\pi t)"))
display(Math(rf"A = {A}\ \text{{m}},\quad k = {k:.3g}\ \text{{rad/m}},\quad \omega = {omega:.3g}\ \text{{rad/s}}"))

display(Math(r"\textbf{Exact (from parameters):}"))
display(Math(rf"\lambda=\frac{{2\pi}}{{k}}={lam_exact:.3g}\ \text{{m}}"))
display(Math(rf"T=\frac{{2\pi}}{{\omega}}={T_exact:.3g}\ \text{{s}}"))
display(Math(rf"f=\frac{{1}}{{T}}={f_exact:.3g}\ \text{{Hz}}"))
display(Math(rf"v_p=\frac{{\lambda}}{{T}}={vp_exact:.3g}\ \text{{m/s}}"))

# ----------------------------
# (1) Spatial snapshot: hold t fixed
# ----------------------------
t0 = 0.0
x = np.linspace(0, 2.0, 1200)
y_snapshot = A * np.cos(k*x - omega*t0)

plt.figure(figsize=(8,4))
plt.plot(x, y_snapshot, label=rf"$y(x,t_0)=A\cos(kx-\omega t_0),\ t_0={t0:g}\,\mathrm{{s}}$")
plt.axhline(0, linewidth=0.7)

# Shade one wavelength from x=0 to x=lambda
plt.axvspan(0, lam_exact, alpha=0.25, label="One wavelength")

# Mark wavelength with a double arrow
y_arrow = -0.28
plt.annotate("", xy=(0, y_arrow), xytext=(lam_exact, y_arrow),
             arrowprops=dict(arrowstyle="<->"))
plt.text(lam_exact/2, y_arrow-0.05, r"$\lambda$", ha="center")

plt.xlabel("Position $x$ (m)")
plt.ylabel("Displacement $y$ (m)")
plt.title("Spatial Snapshot (time fixed)")
plt.legend()
plt.ylim(-0.35, 0.35)
plt.grid(True)
plt.show()

# ----------------------------
# (2) Time trace: hold x fixed
# ----------------------------
x0 = 0.25
t = np.linspace(0, 1.0, 1200)
y_time = A * np.cos(k*x0 - omega*t)

plt.figure(figsize=(8,4))
plt.plot(t, y_time, label=rf"$y(x_0,t)=A\cos(kx_0-\omega t),\ x_0={x0:g}\,\mathrm{{m}}$")
plt.axhline(0, linewidth=0.7)

# Shade one period from t=0 to t=T
plt.axvspan(0, T_exact, alpha=0.25, label="One period")

# Mark period with a double arrow
y_arrow = -0.28
plt.annotate("", xy=(0, y_arrow), xytext=(T_exact, y_arrow),
             arrowprops=dict(arrowstyle="<->"))
plt.text(T_exact/2, y_arrow-0.05, r"$T$", ha="center")

plt.xlabel("Time $t$ (s)")
plt.ylabel("Displacement $y$ (m) at $x_0$")
plt.title("Time Trace (position fixed)")
plt.legend()
plt.ylim(-0.35, 0.35)
plt.grid(True)
plt.show()

# ----------------------------
# Pretty "final answers" (display math)
# ----------------------------
display(Math(r"\textbf{Final results:}"))
display(Math(rf"\boxed{{\lambda = {lam_exact:.3g}\ \text{{m}}}}"))
display(Math(rf"\boxed{{T = {T_exact:.3g}\ \text{{s}}}}"))
display(Math(rf"\boxed{{f = {f_exact:.3g}\ \text{{Hz}}}}"))
display(Math(rf"\boxed{{v_p = {vp_exact:.3g}\ \text{{m/s}}}}"))
```
