## Motivation — Newton’s First Law Revisited

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Newton’s First Law</h3>

An object continues in its state of motion unless acted upon by an unbalanced interaction.
</div>

This statement already contains the essence of what we now call the **impulse–momentum theorem**.

* If motion changes, something must have happened over **time** to transfer momentum.  
* If motion does not change, then the **net momentum transfer must be zero**.

Seen this way, Newton’s First Law is not fundamentally a statement about force. It is a statement about **momentum change**—or more precisely, about the *absence* of momentum change.

This perspective will guide how we now connect force to impulse.


<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D3.4.1 Impulse Due to a Single Force

So far, we have defined impulse purely as momentum transfer:

$$
j_x = \Delta p_x.
$$

We now ask a deeper question: **what physical mechanism causes this transfer?**

When a single force acts on a point particle over a time interval $\Delta t$, the momentum transfer depends on:
- how strong the force is, and
- how long it acts.

For a **constant force**, or when a force can be reasonably replaced by its **average value**, the impulse delivered by that force is

$$
j_x = f_x\,\Delta t.
$$

This expression emphasizes an important idea:  
**a large force acting briefly can produce the same impulse as a smaller force acting over a longer time**.

---

### Units of Force

From the impulse relation,

$$
j_x = f_x\,\Delta t,
$$

and knowing that impulse has units of momentum, we can identify the unit of force:

$$
[f_x] = \frac{\text{kg·m/s}}{\text{s}} = \text{kg·m/s}^2.
$$

This unit is defined as the **newton (N)**:

$$
1\ \text{N} = 1\ \text{kg·m/s}^2.
$$

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D3.4.2 Net Impulse from Multiple Forces

## Net Impulse from Multiple Forces

In many physical situations, a point particle experiences **more than one force** during an interaction.  
Each force contributes to the particle’s momentum transfer.

The most general and physically correct way to describe this is to start with **impulse itself**, not force.

### Net Impulse — General Statement

The **net impulse** acting on a point particle is the **sum of the impulses** delivered by each individual force:

$$
j_{\text{net},x} = \sum j_x.
$$

Each individual impulse depends on:
- the force involved, and  
- the **time interval during which that force acts**.

For a single force acting over a time interval $\Delta t$, the impulse is

$$
j_x = f_x\,\Delta t \quad \text{(constant or average force)}.
$$

Putting this together, the net impulse can be written as

$$
j_{\text{net},x} = \sum (f_x\,\Delta t)_i,
$$

where the subscript $i$ labels the different forces.

<div style="background-color:#ffe6e6; border-left:6px solid #990000; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Important Caution</h3>

The time interval over which each force acts **does not have to be the same**.

Some forces may act very briefly (such as contact forces during a collision),  
while others may act over much longer times (such as gravity or friction).

You must therefore **never assume** that forces can be added first and multiplied by a single time interval unless you have explicitly checked that they act over the same duration.
</div>

---

### Special Case — Forces Acting Over the Same Time Interval

In many simplified situations, all forces acting on the particle act over the **same time interval** $\Delta t$.  
When this is true, the expression for net impulse simplifies significantly.

Factoring out the common time interval gives:

$$
j_{\text{net},x} = \left( \sum f_x \right)\Delta t.
$$

We then define the **net force** as

$$
f_{\text{net},x} = \sum f_x,
$$

so that the net impulse becomes

$$
j_{\text{net},x} = f_{\text{net},x}\,\Delta t.
$$

This form is convenient and widely used, but it is valid **only** when all contributing forces act over the same time interval.

---

### Why This Distinction Matters

Failing to account for different force durations can lead to serious conceptual and numerical errors, especially in:
- collisions,
- propulsion and thrust problems,
- contact interactions, and
- problems involving intermittent forces.

By starting from **impulse as momentum transfer**, and only later introducing net force as a special case, we preserve the correct physical logic of Newtonian mechanics.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Impulse–Momentum Theorem (1D)</h3>

The **net impulse** acting on a point particle equals the **change in its momentum**:

$$
j_{\text{net},x} = \Delta p_x.
$$

This relation connects **forces acting over time** to **changes in motion**.
</div>

---

### Why This Matters

Impulse is fundamentally different from many other quantities you have encountered so far because it explicitly involves **time**.  
When we describe momentum transfer through impulse, we are asking how interactions unfold *over a duration*.

By contrast, when we later study **work and energy**, we will see that work involves **force acting through position**, not time. In other words:

- **Impulse** connects force to **time**
- **Work** connects force to **displacement**

This distinction is not just a bookkeeping choice—it reflects a deep structural feature of physics.

Quantities that depend on time naturally emphasize **how interactions evolve**, while quantities that depend on position emphasize **where interactions occur**. These two perspectives lead to very different—but complementary—ways of understanding motion.

Far beyond this course, this distinction has profound consequences. For example:
- In signal analysis and physics, time-based and position-based descriptions lead to different mathematical tools, such as **Fourier transforms**, which relate how localized something is in time to how spread out it is in frequency.
- In quantum mechanics, similar ideas appear in the form of **uncertainty principles**, where pairs of quantities tied to time-like and position-like descriptions cannot both be specified arbitrarily precisely.

Even in classical mechanics, these ideas are already present. Momentum naturally pairs with position, while energy naturally pairs with time. These pairings—called **canonical relationships**—hint at a deeper organizational structure underlying Newton’s laws.

You do not need to understand any of this for this class. But it is worth knowing that when we choose to describe motion using impulse rather than force alone, we are not just solving problems more efficiently—we are aligning ourselves with a framework that remains meaningful across all of physics.

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example 1 – Constant Force Acting Over Time</h3>

**Problem**  
A $0.50\ \text{kg}$ cart initially moves at $+2.0\ \text{m/s}$.  
A constant force of $f_x = -4.0\ \text{N}$ acts for $0.50\ \text{s}$.

Find the final velocity of the cart.

---

**Solution**

Impulse:

$$
j_x = f_x\,\Delta t = (-4.0)(0.50) = -2.0\ \text{kg·m/s}.
$$

Initial momentum:

$$
p_{x,i} = m v_{x,i} = (0.50)(2.0) = +1.0\ \text{kg·m/s}.
$$

Final momentum:

$$
p_{x,f} = p_{x,i} + j_x = 1.0 - 2.0 = -1.0\ \text{kg·m/s}.
$$

Final velocity:

$$
v_{x,f} = \frac{p_{x,f}}{m} = \frac{-1.0}{0.50} = -2.0\ \text{m/s}.
$$

The cart reverses direction due to the impulse delivered by the force.

## D3.4.4 Impulse and the Force–Time Graph

So far, we have described impulse in two equivalent ways.

First, impulse is defined as **momentum transfer**,

$$
j_x = \Delta p_x.
$$

Second, when a force is constant (or can be well approximated by its average value), impulse can be written as

$$
j_x = f_x\,\Delta t.
$$

### Constant Force → Rectangle

If a force is **constant**, a force–time graph is simply a **horizontal line**.  
The impulse is then given by the **area under that line**, which forms a rectangle:

- height = force magnitude $f_x$  
- width = interaction time $\Delta t$

The area of this rectangle is

$$
\text{area} = f_x\,\Delta t = j_x,
$$

which shows that impulse is literally the **area under a force–time graph**.

---

### Variable Force → Sum of Small Rectangles

In real situations, forces often change with time. The force–time graph may be curved or irregular.  
Even in these cases, the idea of impulse as *area* still applies.

The key idea is this:

> Any curve can be approximated by a large number of **small rectangles**.

Each rectangle represents a small impulse delivered over a short time interval.  
Adding up the areas of all these rectangles gives an approximation to the total impulse.

This viewpoint allows us to estimate impulse **numerically**, even when the force is not constant.

---

### Connection to Calculus (Optional Insight)

If you have studied calculus, you may recognize this process as the idea behind **Riemann sums**.

Depending on how the rectangles are chosen, we obtain different approximations:

- **Left Riemann sum**: rectangles use the force value at the beginning of each interval  
- **Right Riemann sum**: rectangles use the force value at the end of each interval  
- **Midpoint (center) sum**: rectangles use the force value at the center of each interval  

As the rectangles become narrower, these approximations improve.  
More advanced methods—such as the **trapezoidal rule** and **Simpson’s rule**—combine rectangles and trapezoids to achieve higher accuracy.

In calculus, taking the limit as the rectangles become infinitesimally thin leads to the exact impulse.

---

### Why This Matters (Even Without Calculus)

You do **not** need calculus to understand the physical idea:

- Impulse depends on **force acting over time**
- The total impulse is the **accumulated effect** of that force
- Force–time graphs provide a direct visual and numerical way to determine momentum change

Whether using simple rectangles or advanced mathematics, the physical meaning remains the same:  
**impulse is the total momentum transferred during an interaction**.


---

### Force Acting Over Time

When a force acts on a point particle, its value may change during the interaction.  
Rather than focusing on the force at a single instant, impulse asks a different question:

> **How much total momentum is transferred over the entire time interval of interaction?**

This is where the **force–time graph** becomes useful.

---

### Impulse from a Force–Time Graph

A **force–time graph** shows the value of the force $f_x$ acting on a particle as a function of time $t$.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Impulse from a Force–Time Graph</h3>

The impulse delivered to a point particle equals the **area under the force–time curve**:

$$
j_x = \text{area under } f_x \text{ vs. } t.
$$

</div>

This statement holds regardless of whether the force is constant, increasing, decreasing, or changing in a more complicated way.

---

### Common Force–Time Shapes

In introductory problems, force–time graphs often take simple geometric forms:

- **Rectangle** (constant force):  

  $$
  j_x = f_x\,\Delta t
  $$

- **Triangle** (linearly changing force):  

  $$
  j_x = \frac{1}{2}(\text{base})(\text{height})
  $$

- **Trapezoid** (force changes between two values):  

  $$
  j_x = \frac{1}{2}(f_{\text{initial}} + f_{\text{final}})\,\Delta t
  $$

In each case, the impulse is found by calculating the **geometric area** under the curve.

---

### Physical Interpretation

- A **larger area** means a larger momentum transfer.  
- A **positive area** corresponds to impulse in the positive $x$-direction.  
- A **negative area** corresponds to impulse in the negative $x$-direction.

Importantly, very different force–time graphs can produce the **same impulse**.  
This explains why a soft catch (small force over long time) and a hard catch (large force over short time) can bring an object to rest with the same change in momentum.

---

### Why Force–Time Graphs Matter

Force–time graphs emphasize that:
- impulse depends on **time-integrated interaction**, not peak force,
- momentum change can be determined **without knowing acceleration**, and
- the details of how a force varies often matter less than the **total area** under the curve.

This perspective is essential for understanding collisions, safety devices, propulsion, and many real-world interactions.

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example 2 – Impulse from a Force–Time Graph</h3>

A particle initially moves at $v_{x,i} = +3.0\ \text{m/s}$ with mass $m = 1.0\ \text{kg}$.  
A force acts in the $x$-direction according to the force–time graph below.

The force increases linearly from $0$ to $6.0\ \text{N}$ over $2.0\ \text{s}$ and then drops to zero.

**Impulse from a force–time graph** is equal to the **area under the curve**.

For this triangular graph:

$$
j_x = \frac{1}{2}(2.0)(6.0) = 6.0\ \text{kg·m/s}.
$$

Initial momentum:

$$
p_{x,i} = (1.0)(3.0) = 3.0\ \text{kg·m/s}.
$$

Final momentum:

$$
p_{x,f} = 3.0 + 6.0 = 9.0\ \text{kg·m/s}.
$$

Final velocity:

$$
v_{x,f} = \frac{9.0}{1.0} = 9.0\ \text{m/s}.
$$
</div>

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Given parameters
m = 1.0          # mass (kg)
v_x_i = 3.0      # initial velocity (m/s)
T = 2.0          # duration of force (s)
F_max = 6.0      # maximum force (N)

# Time array
t = np.linspace(0, T, 200)

# Force as a function of time (linear ramp)
f_x = (F_max / T) * t

# Impulse calculations
j_x_geom = 0.5 * T * F_max           # geometric area (triangle)
j_x_num = np.trapezoid(f_x, t)           # numerical integration (trapezoidal rule)

# Momentum and final velocity
p_x_i = m * v_x_i
p_x_f = p_x_i + j_x_geom
v_x_f = p_x_f / m

# Output (scientific notation, 2 significant figures)
print(f"Impulse (geometry)  = {j_x_geom:.2e} kg·m/s")
print(f"Impulse (numerical) = {j_x_num:.2e} kg·m/s")
print(f"Final velocity      = {v_x_f:.2e} m/s")

# Plot force–time graph
plt.figure()
plt.plot(t, f_x)
plt.fill_between(t, 0, f_x, alpha=0.3)
plt.xlabel("Time t (s)")
plt.ylabel("Force f_x (N)")
plt.title("Force–Time Graph (Triangular Pulse)")
plt.show()
```
