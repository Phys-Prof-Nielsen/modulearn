## Motivation and Concept of Motion

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Newton’s First Law (Law of Inertia)</h3>

An object continues in its state of motion unless acted upon by an unbalanced interaction.
</div>

Newton’s First Law is fundamentally a conceptual statement about **motion**. 

It tells us that motion is not something that automatically changes on its own—something must *interact* with an object in order to change how it moves.

In everyday language, we often use the word **motion** without worrying about what it actually means. We typically associate motion with *speed*—how fast something is going. While this intuition feels natural, it turns out to be incomplete from a physics point of view.

To reveal this limitation, consider the following question:

**Which of the following has more motion: a sedan traveling at 70 mph or a semi-truck traveling at 70 mph?**

Many students initially answer that they have the same motion, since they are moving at the same speed. This response highlights a common assumption: that motion depends only on how fast an object moves.

Newton’s First Law invites us to think differently. If motion can only change when an unbalanced interaction occurs, then the *amount* of interaction required to stop an object tells us something meaningful about its motion.

Now reconsider the scenario from this perspective:

**A sedan and a semi-truck are both traveling at 70 mph. Which one requires a greater interaction to come to a complete stop?**

Without much hesitation, most people correctly answer that the semi-truck requires a greater interaction. This intuitive response points us toward a more precise physical understanding of motion.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Insight</h3>

A more massive object has **more motion** than a less massive object moving at the same speed.  
This shows that motion depends not only on how fast an object is moving, but also on how much matter it contains.

In physics, motion is therefore linked to both **speed** and **mass**, not speed alone.
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D3.2.1 Reference Frame
Building on our first discussion of motion, we used the example of a sedan and a semi-truck to argue that motion is **more than just speed**—it also depends on the mass of the moving object:

**A sedan and a semi-truck are both traveling at 70 mph. Which one requires a greater interaction to come to a complete stop?**

The natural conclusion was that the semi-truck requires a greater interaction. This told us that motion depends on both **how fast** an object moves and **how much matter** it contains.

Let us now take this idea one step further.

Imagine a sailboat moving across Utah Lake with a certain motion. Suppose strong winds begin to act on the boat. Whether the boat speeds up or slows down depends not only on the strength of the interaction, but also on the **relative direction** of the wind and the boat’s motion. A tailwind increases the boat’s speed and motion, while a headwind reduces both.

This example reveals an important limitation in how we have described motion so far: to fully describe changes in motion, we must also account for **direction**, not just magnitude.

At this point, it is helpful to distinguish between two closely related ideas. Speed tells us *how fast* something is moving, but it says nothing about *which way* it is moving. To include directional information, we introduce a new quantity.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Velocity</h3>

**Velocity** describes how fast an object is moving **and** the direction of its motion.

In other words, velocity combines **speed** with **direction**.
</div>

Mass does not depend on direction, and speed alone does not either (a car’s speedometer, for example, does not tell you which way you are traveling). Directional information enters our description of motion through velocity.

To describe direction in a clear and consistent way, physics requires the use of a **reference frame**, also called a **coordinate system**. A reference frame provides:
- a chosen direction (positive and negative), and  
- a reference point relative to which positions and motion are described.

That reference point is called the **origin**.

In this course, we will encounter several coordinate systems. In **Phase D**, we restrict ourselves to a **one-dimensional Cartesian system**, where direction is represented using positive and negative signs along a single line. Two- and three-dimensional Cartesian systems will be introduced in **Phase C**, and polar coordinates will appear later in **Phase B**.

While we may think of these reference frames being simple, they are indeed more complex than we may think and relates to the algebra they are formulated in. Fortunately, we do not have to worry about that in this course. I suggest you read the wikipedia page on [Cartesian Coordinate System](https://en.wikipedia.org/wiki/Cartesian_coordinate_system)

## D3.2.2 1D Cartesian System

The 1D cartesian system is represented just like a number line in elementary school. The difference here is that the positive and negative numbers now represents a sense of direction to us. For example, in the sketch below, positive numbers will represent objects that directed to the right, whereas negative numbers indicate objects directed to the left. 

```python
%reset -f

import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Draw the first reference frame
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
ax.quiver(0, 0, -1, 0, angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)

# Set axis limits
ax.set_xlim(-1.5,1.5)


# Add labels
ax.text(-0.035,-0.12, 'O', fontsize=10, color='k')


# Set the grid and aspect ratio
#ax.grid(True)
ax.set_aspect('equal', adjustable='box')

# Display the plot
plt.title("Reference Frame")

ax.annotate('$+x$', (1,0))
ax.annotate('$-x$', (-1.1,0))

# Turn off the axes
plt.axis('off')
plt.scatter(0,0, color='blue')

plt.show()
```

The 1D system could also be directed in the vertical direction on the page (or screen), and we then often label it $y$ or $z$ instead of $x$.

## D3.2.3 Linear Momentum 
From our discussion of the sedan and the semi-truck, we were led to the idea that motion depends on both **mass** and **speed**. This motivates an initial definition of momentum—without yet accounting for direction—as

$$
p = mv \tag{1}
$$

where $m$ is the mass of the particle and $v$ is its speed.

However, the sailboat example showed that motion—and therefore momentum—must also account for **direction**. In one dimension, we represent this directional information explicitly by attaching a subscript to the momentum and velocity.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Linear Momentum for a Point Particle (1D)</h3>

$$
p_x = m v_x \tag{2}
$$
</div>

Here, the subscript $x$ indicates that the momentum and velocity include directional information along a chosen one-dimensional axis. Positive and negative values correspond to motion in opposite directions.

The quantity defined in Equation (1), which does not include direction, is called the **magnitude** of the momentum.

We are now ready to explore momentum in one-dimensional motion.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example 1 – Momentum of a Point Particle in 1D</h3>

A Subaru Outback (modeled as a point particle) has a mass of $m = 1648\ \text{kg}$ and is traveling with a speed of $v = 32\ \text{m/s}$ in the $+x$ direction.

Determine the momentum of the Subaru.

**Methodology**

We are given the mass, speed, and direction of motion.  
Since direction is specified, the speed can be written as a **velocity** with a sign.  
Once the velocity is known, the momentum is found using the one-dimensional momentum definition,

$$
p_x = m v_x.
$$

**Solution**

The magnitude of the momentum is

$$
p = mv = (1648)(32) = 5.3\times10^4\ \text{kg·m/s}.
$$

Because the Subaru is traveling in the **positive** $x$-direction, its momentum must also be positive:

$$
p_x = +5.3\times10^4\ \text{kg·m/s}.
$$

</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example 2 – Momentum with Negative Direction</h3>

**Problem**  
A hockey puck (modeled as a point particle) has a mass of  
$m = 0.17\ \text{kg}$ and is sliding with a speed of  
$v = 12\ \text{m/s}$ in the **negative** $x$-direction.

Determine the momentum of the puck.

---

**Methodology**  
We are given the mass, speed, and direction of motion.  
Because the motion is along the negative $x$-direction, the velocity must be assigned a negative sign.  
The momentum is then computed using the one-dimensional definition

$$
p_x = m v_x.
$$

---

**Solution**  
The magnitude of the momentum is

$$
p = mv = (0.17)(12) = 2.0\ \text{kg·m/s}.
$$

Since the puck is moving in the **negative** $x$-direction, its momentum is negative:

$$
p_x = -2.0\ \text{kg·m/s}.
$$

</div>

**Method 2 — Solving symbolically with SymPy**

`import sympy as sp`  

`m, p_x, v_x = sp.symbols('m p_x v_x')`  

`equation = sp.Eq(p_x, m * v_x)`  

`equation_sub = equation.subs({m: 2.50, p_x: -15.0})`  

`solution = sp.solve(equation_sub, v_x)[0]`  

`print(f"{solution:.2e}")`

---
