## D3.1.1 Definition — Point Particle

Many core ideas in Newtonian physics—such as force, momentum, and energy—can become quite complex when applied to real objects and interactions. To build intuition and develop foundational reasoning skills, we begin with an idealized model known as a **point particle**.

This model allows us to focus on *how motion changes* without being distracted by details that are not yet important.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Point Particle — Definition</h3>

A **point particle** is an idealized object that:
- has **mass**, and  
- whose **size and shape do not affect the motion being studied**.
</div>

The phrase often used in textbooks—*“the particle has no spatial extension”*—can be misleading. Taken literally, it sounds unphysical and abstract. A more useful interpretation is this:

> An object may be treated as a point particle **as long as its size and shape play no role in the dynamics of interest**.

This perspective allows familiar, real-world objects to be modeled as point particles in appropriate situations. For example, a car, a basketball, or a puck can all be treated as point particles when we are interested only in their overall motion and not in rotation, deformation, or internal structure.

Later in the course, we will relax this assumption and explicitly account for size, shape, and internal motion. For now, the point-particle model provides a clean and powerful starting point for understanding motion and interactions.

## D3.1.2 Point Mass

### About Mass

**Mass** is one of those concepts that feels familiar in everyday life, yet turns out to be surprisingly subtle in physics. Depending on the physical framework being used, mass can be understood in different ways:

1. As a measure of resistance to changes in **linear motion**  
   (the Newtonian concept of **inertial mass**).
2. As a measure of the **amount of matter** in an object  
   (common in chemistry and materials science).
3. Through mass–energy equivalence in relativity, where mass and energy influence the curvature of spacetime.
4. As a parameter arising from interactions between quantum fields (e.g., the Higgs mechanism in particle physics).

In this course, we restrict our discussion to the **first two interpretations**, which are sufficient for introductory mechanics. The remaining perspectives belong to more advanced physics, but they are worth mentioning to emphasize that mass is a deeply fundamental—and fascinating—concept.

---

### What Do We Mean by a *Point Mass*?

At first glance, the idea of a *point mass* may seem unphysical. Objects in our everyday world have size and shape, and even fundamental particles are not literally mathematical points.

In mechanics, however, a **point mass** is not meant to describe physical size. Instead, it is a **modeling assumption** that builds directly on the idea of a point particle introduced earlier.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Point Mass — Definition</h3>

A **point mass** is a point particle to which a mass is assigned, under the assumption that the object’s size and shape do not affect the motion being analyzed.
</div>

This means that many real objects—cars, balls, pucks, or even planets—can be treated as point masses **as long as their size, shape, and internal structure do not influence the dynamics of interest**.

### Definition of Mass — Newtonian Approach

In Newtonian mechanics, mass is defined operationally through how an object responds to interactions. From Newton’s second law (under appropriate assumptions), mass is given by

$$
m = \frac{\sum F}{a}
$$

where $\sum F$ is the magnitude of the net force acting on the object and $a$ is the magnitude of the resulting acceleration.

This definition highlights an important physical interpretation:

> **Mass measures an object’s resistance to changes in motion.**

For this reason, mass defined in this way is called **inertial mass**.

To date, no experiment has shown any difference between inertial mass and other notions of mass (such as gravitational mass). While classical physics allows for the possibility of distinct definitions, modern physics—particularly general relativity—shows that these concepts are fundamentally equivalent.

The standard symbol for the mass of a point particle is the lowercase letter $m$.
