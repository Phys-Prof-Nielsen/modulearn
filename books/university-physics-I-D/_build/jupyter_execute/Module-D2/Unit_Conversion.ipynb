## D2.2.1 Why Units Are Important

Units are the language we use to communicate physical measurements. A number by itself is not enough — we must also
express *what the number refers to*. Without units, a measurement has no physical meaning.

Consider the following statements:

- The car drove **50**.  
- The car drove **50 mph**.  
- The car drove **50 km**.

The first statement is grammatically correct, but physically meaningless.  
Fifty *what*? Miles? Kilometers? Feet? Seconds? Without a unit, we cannot interpret the magnitude of the motion, compare it to anything else, or use it in a calculation.

The next two statements, however, are meaningful because the unit clarifies the physical quantity being described.  
Whether the car traveled a **distance** of 50 km or had a **speed** of 50 mph, the unit tells us exactly what the number represents.

Units also allow us to:

- relate one quantity to another  
- apply formulas consistently (e.g., using SI units in $F = ma$)  
- check our work for errors  
- convert between measurement systems  
- communicate results unambiguously with other scientists

---

### Why This Matters — Units Are Part of the Measurement

A physical measurement has **two inseparable components**:

1. A **number**, which tells us the magnitude  
2. A **unit**, which tells us *what the number refers to*

Together, they form a complete physical quantity.

Units act as the **grammar of physics**, allowing us to communicate clearly, compare results, and perform correct calculations.  
Without units, quantities cannot be interpreted or used meaningfully in equations such as  

$$F = ma,\qquad v = \frac{\Delta x}{\Delta t},\qquad E = \frac{1}{2}mv^2.$$

Clear units ensure clear physics.

<div style="background-color:#ffe6e6; border-left:6px solid #cc0000; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Danger — Mixing or Omitting Units</h3>

One of the most common and serious errors in physics is **forgetting units** or **mixing incompatible ones**.

Examples of mistakes:
- Adding meters to centimeters without converting  
- Using seconds in one part of a formula and minutes in another  
- Reporting a measurement like “The object traveled 12” with no unit at all  
- Using inconsistent systems (e.g., feet in one term, meters in another)

These mistakes can completely invalidate a calculation.  
In real engineering and scientific work, unit errors have caused failed experiments, structural failures, and even destroyed spacecraft (e.g., NASA’s Mars Climate Orbiter).

Always **write units**, **check units**, and **keep units consistent** throughout every calculation.

## D2.2.2 Standard Metric Units
An almost unlimited number of units exists, and we can even make up our own units. In the US it is custom to use the <b>British Imperical Units</b>. The majority of the rest of the world is using a standard-based set of units known as the <b>International System of Units</b> or <b>SI</b> units for short (abbreviated from the French word). This metric system consists of 7 units from which all our physical units can be expressed or derived.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/SI_base_units.svg/1200px-SI_base_units.svg.png"  width="200"/>

For the interested student, more details can be found [here](https://en.wikipedia.org/wiki/SI_base_unit) and [here](https://en.wikipedia.org/wiki/Metric_system)

In mechanics, we use three fundamental units: kilogram (kg), meter (m), and second (s).

Other units can be derived from these. For example: Unit of Force:: Newton (N) = kgm/s$^2$ and Unit of Energy :: Joules (J) = kgm$^2$/s$^2$

### Metric vs. Standard Metric Units
We often see units such as millimeter (mm), centimeter (cm), and kilometer (km) as units of length measurements. These are all <b>metric</b> units but not standard metric unit. The <b>standard metric unit</b> for length is the meter (m). Similarly, units of mass can be in grams (g) which is a metric unit, but the standard metric unit is the kilogram (kg).

## D2.2.3 Derived Units and Dimensional Analysis

Most quantities in physics are **derived** from three base SI units used throughout this module:

- **mass:** kilogram (kg)  
- **length:** meter (m)  
- **time:** second (s)

A *derived unit* is just a combination of these base units. **Dimensional analysis** is the practice of checking (and building) equations by making sure the units match on both sides.

---

### Dimensions vs. Units

We often separate the *type* of quantity from the *unit* we measure it in:

- Dimensions use symbols like $M, L, T$ (mass, length, time).
- SI units are kg, m, s.

Examples:
- Mass: $[m] = M$ and unit is kg  
- Position: $[x] = L$ and unit is m  
- Time: $[t] = T$ and unit is s  

When we write **“the dimensions of something”**, we mean how it depends on $M, L, T$.

---

### Common Derived Quantities in Mechanics (Built from kg, m, s)

Below are some of the most important derived quantities we use in introductory physics. The key skill is to express each one in base units.

#### Velocity

$$v = \frac{\Delta x}{\Delta t}$$

Units:

$$[v] = \frac{\text{m}}{\text{s}} = \text{m s}^{-1}$$

Dimensions:

$$[v] = L T^{-1}$$

#### Acceleration

$$a = \frac{\Delta v}{\Delta t}$$

Units:

$$[a] = \frac{\text{m s}^{-1}}{\text{s}} = \text{m s}^{-2}$$

Dimensions:

$$[a] = L T^{-2}$$

#### Force (Newton’s Second Law)

$$F = ma$$

Units:

$$[F] = \text{kg}\cdot\text{m s}^{-2}$$

This combination is so common it has a special name:

$$1\ \text{N} = 1\ \text{kg}\cdot\text{m s}^{-2}$$

Dimensions:

$$[F] = M L T^{-2}$$

#### Momentum

$$p = mv$$

Units:

$$[p] = \text{kg}\cdot\text{m s}^{-1}$$

Dimensions:

$$[p] = M L T^{-1}$$

#### Work / Energy

$$W = Fd$$

Units:

$$[W] = (\text{kg}\cdot\text{m s}^{-2})\cdot \text{m} = \text{kg}\cdot\text{m}^2\text{s}^{-2}$$

Special name:

$$1\ \text{J} = 1\ \text{kg}\cdot\text{m}^2\text{s}^{-2}$$

Dimensions:

$$[W] = M L^2 T^{-2}$$

#### Power

$$P = \frac{W}{t}$$

Units:

$$[P] = \frac{\text{kg}\cdot\text{m}^2\text{s}^{-2}}{\text{s}} = \text{kg}\cdot\text{m}^2\text{s}^{-3}$$

Special name:

$$1\ \text{W} = 1\ \text{kg}\cdot\text{m}^2\text{s}^{-3}$$

Dimensions:

$$[P] = M L^2 T^{-3}$$

---

### Dimensional Consistency from Equations

A physical equation must be **dimensionally consistent**:

- Every term added or subtracted must have the **same dimensions**.
- Both sides of the equation must have the **same dimensions**.

Example: constant-acceleration kinematics

$$x = x_0 + v_0 t + \frac{1}{2}at^2$$

Check each term has units of meters:

- $x$ and $x_0$ are in m  
- $v_0 t$ has units $(\text{m s}^{-1})(\text{s})=\text{m}$  
- $at^2$ has units $(\text{m s}^{-2})(\text{s}^2)=\text{m}$  

So the equation is dimensionally consistent.

---

### Finding the Units (and Meaning) of a Constant in an Equation

Dimensional analysis can tell you the units of unknown constants.

Suppose a model says

$$F = kx$$

where $F$ is force and $x$ is displacement.

Solve for the units of $k$:

$$k = \frac{F}{x}$$

$$[k] = \frac{\text{kg}\cdot\text{m s}^{-2}}{\text{m}} = \text{kg}\cdot\text{s}^{-2}$$

That means $k$ is *not* unitless—it carries units, and those units matter.

---

### Forming an Equation Using Only Dimensions (Building Physics from kg, m, s)

A powerful use of dimensional analysis is to guess the form of an equation when you know which variables matter.

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example: Period of a pendulum (shape-only argument)</h3>

Example: Period of a pendulum (shape-only argument)
Assume the period $T$ depends only on length $\ell$ and gravitational acceleration $g$:
- $[\ell]=L$
- $[g]=L T^{-2}$

Assume

$$T \propto \ell^a g^b$$

Match dimensions. Left side has $[T]=T$.

Right side dimensions:

$$[\ell^a g^b] = L^a (L T^{-2})^b = L^{a+b} T^{-2b}$$

Equate exponents of $L$ and $T$:

For time:

$$-2b = 1 \quad \Rightarrow \quad b = -\frac{1}{2}$$

For length:

$$a + b = 0 \quad \Rightarrow \quad a = \frac{1}{2}$$

So

$$T \propto \sqrt{\frac{\ell}{g}}$$

Dimensional analysis cannot determine the numerical factor (the true result is $T = 2\pi\sqrt{\ell/g}$), but it gets the *structure* right.

</div>

---

### Two Quick Rules to Remember

1. **You can always rewrite any mechanical unit in terms of kg, m, s.**  
2. **If an equation is wrong in units, it is wrong in physics.**  

Use dimensional analysis as a fast “sanity check” while solving problems—and as a tool to build formulas when you’re not sure what the equation should look like.

### Number 1

In Linear Algebra, numbers (or more generally, *elements*) live inside a particular **set**, **group**, or **algebraic structure**.  
Each algebra comes with a collection of **axioms**—rules that all elements within that system must obey.

For example:

- The **commutative property** for addition states that  

  $$2 + 3 = 5 \quad \Rightarrow \quad 3 + 2 = 5.$$

  
  This seems obvious because we have grown up with it, but there are algebras where addition is **not** commutative.

- The **distributive property** states that  

  $$(A + B)C = AC + BC.$$

These rules are not arbitrary; they define what operations are allowed and how the algebra behaves.

---

One of the fundamental axioms we rely on is the existence of a **multiplicative identity**, a special number that leaves other numbers unchanged when it multiplies them. In everyday arithmetic, this number is **1**.

In our algebra, we assume:

> **There exists a number $1$ such that multiplying any element $x$ by 1 leaves it unchanged.**

Expressed mathematically:

$$
x \cdot 1 = x
$$

This simple property will be essential as we manipulate algebraic expressions throughout this section in terms of unit conversions.

---

### How to Write “1”

The number $1$ may look simple, but it can take many forms. Any expression whose value is exactly one can act as a **multiplicative identity**, meaning it does not change another quantity when multiplied. For example:

$$1$$

$$\frac{2}{2}$$

$$\frac{\sqrt{2x}}{\sqrt{2x}}$$

Each of these equals one, so multiplying by them leaves a number or expression unchanged.

---

### Using Units to Write “1”

A powerful idea in physics is that **equivalent quantities can be used to construct the number 1**.  
For instance, if we know that

$$1~\text{m} = 3.281~\text{ft},$$

then both of the following ratios equal one:

$$\frac{3.281~\text{ft}}{1~\text{m}} = 1,$$

$$\frac{1~\text{m}}{3.281~\text{ft}} = 1.$$

These are simply different ways of writing the number 1, but expressed using **units**.

>This is the foundation of **unit conversion**: multiplying a quantity by a ratio equal to 1 does not change its physical meaning, but allows us to express it in different units. Understanding how and why these “ones” work is essential for consistent and accurate calculations in physics. This method is known as the **factor-label method**.
>

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Factor-label Method</h3>
    
The factor–label method converts units by multiplying by ratios equal to 1, arranged so that unwanted units cancel and desired units remain.
</div>

---

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example — Converting Meters to Feet Using the Number “1”</h3>

**Problem:** Convert $2.6~\text{m}$ to feet.

---

**Methodology**  
To convert units, we multiply by a ratio that equals the number **1**, but is written using equivalent units.  
From conversion tables:

$$1~\text{m} = 3.281~\text{ft}.$$

This allows us to construct the identity:

$$1 = \frac{3.281~\text{ft}}{1~\text{m}}.$$

Multiplying by this form of 1 will not change the physical quantity—it simply expresses it in new units.

**Solution**  
Multiply the original measurement by the conversion factor:

$$
(2.6~\text{m})\left(\frac{3.281~\text{ft}}{1~\text{m}}\right).
$$

Handle numericals and units separately:

- Units: $\text{m}$ cancels, leaving $\text{ft}$.  
- Numbers: $2.6 \times 3.281 = 8.5306$.

Rounded to two significant figures (matching the original $2.6$):

$$8.5~\text{ft}.$$

**Final Answer:** 

$$2.6~\text{m} = 8.5~\text{ft}.$$

This method works for all unit conversions by choosing an appropriate “1” written as a ratio of equivalent units.

</div>
<br>

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example — Converting Multiple Units</h3>


When converting more than one unit, we simply multiply by **a separate version of “1”** for each unit we want to change.  
Because each ratio equals 1, the physical quantity stays the same — only its units change.

---

**Problem:** Convert $55~\text{m/s}$ to $\text{km/h}$.

We must convert:

- meters → kilometers  
- seconds → hours  

Each requires its own “1,” written as a ratio of equivalent units:

$$1 = \frac{1~\text{km}}{1000~\text{m}}, \qquad 
1 = \frac{3600~\text{s}}{1~\text{h}}.$$

Now multiply by both:

$$
55~\frac{\text{m}}{\text{s}}
\left(\frac{1~\text{km}}{1000~\text{m}}\right)
\left(\frac{3600~\text{s}}{1~\text{h}}\right)
$$

Units cancel:

- $\text{m}$ with $\text{m}$  
- $\text{s}$ with $\text{s}$  

Leaving:

$$
55 \times \frac{1}{1000} \times 3600~\frac{\text{km}}{\text{h}}
= 198~\frac{\text{km}}{\text{h}}.
$$

Rounded appropriately:

**$55~\text{m/s} \approx 200~\text{km/h}$.**

This method works for any multi-unit conversion—just apply one “1” for each unit you need to convert.

<h3 style="margin-top:0; color:#000000;">Box Activity 2 — Unit Conversion in Python (Miles → Meters)</h3>

In this activity, you will convert a distance from **miles** to **meters** using Python.  
Just like algebraic unit conversions, Python uses multiplication `*` and division `/` to apply the proper “1” factor.

Your final answer should include **units** and use **scientific notation** with **proper significant figures**.

Follow the guided steps below.

---

**1. Declare the given information**

Create a variable holding the distance in miles:

`d_mile = 1.42`

---

**2. Construct a conversion factor written as “1”**

From unit tables:

$$1~\text{mile} = 1609.34~\text{m}.$$

We represent this in Python with a descriptive variable name:

`mile_to_m = 1609.34/1      # meters per mile`

This is our “multiply-by-1” factor.

---

**3. Convert miles to meters**

Now multiply the given distance by the conversion factor:

`d_meter = d_mile * mile_to_m`

This produces the distance in **meters**.

---

**4. Format the final answer**

Use Python formatting to display:

- scientific notation  
- three significant figures (in this example)  
- the unit “m”

Example formatting:

`answer = "{:.2e}".format(d_meter)`

---

**5. Print the result professionally**

Your output should read something like:

`The distance in meters is: <value> m`

Use strings for the unit:

`print('The distance in meters is: ' + answer + ' m')`

or

`print(f"The distance in meters is: {answer} m")`

or we could even condense it by formatting it inside the print statement:

`print(f"The distance in meters is: {d_meter:.2e} m")`
