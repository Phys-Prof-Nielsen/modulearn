## D2.7.1 Introduction

Graphs are an essential tool in physics for:
- Visualizing relationships between variables
- Identifying trends and patterns
- Communicating results clearly and concisely

A **well-made scientific graph** can convey complex data at a glance,  
while a poorly made graph can obscure or mislead.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D2.7.2 Types of Scientific Graphs

In introductory physics, the most common graph types are:

1. **Linear (Cartesian) plots** — $y$ vs $x$
2. **Logarithmic plots** — $\log y$ vs $\log x$
3. **Semi-log plots** — log scale on one axis
4. **Histograms** — distribution of data
5. **Bar graphs** — comparing discrete categories (less common in physics experiments)

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D2.7.3 Essential Components of a Good Graph

A proper scientific graph should include:

1. **Descriptive title**  
   - Indicates what is being shown (e.g., “Velocity vs Time for a Rolling Ball”)

2. **Labeled axes with units**  
   - Example: *Time (s)*, *Velocity (m/s)*

3. **Proper scales and tick marks**  
   - Avoid cramped or excessively sparse scales  
   - Choose ranges that show trends without excessive empty space

4. **Data points with uncertainty** (Phase C) 
   - Use error bars when appropriate  
   - Vertical error bars: uncertainty in $y$  
   - Horizontal error bars: uncertainty in $x$

5. **Legend** (if multiple datasets or fits are shown)

6. **Consistent symbols and colors**  
   - Avoid overusing colors; keep it readable in black and white

## D2.7.4 Plotting Data: An Example

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example — Uniform Motion Plot</h3> 
    
**Experiment:** Measure displacement $x$ at different times $t$ for an object in uniform motion.

| Time $t$ (s) | Displacement $x$ (m) | 
|--------------|----------------------| 
| 0.0 | 0.00 | 
| 1.0 | 2.05 | 
| 2.0 | 4.10 | 
| 3.0 | 6.05 | 
| 4.0 | 8.10 |


---

**Step 1 — Choose axes**
- $t$ is the **independent variable** → horizontal axis
- $x$ is the **dependent variable** → vertical axis

**Step 2 — Label with units**
- $t$ in seconds (s)
- $x$ in meters (m)

**Step 3 — Fit the data (Phase B, optional for now)**
A **linear fit** should yield:

$$
x(t) \approx v t + x_0
$$

where $v$ is the slope (velocity) and $x_0$ is the intercept.

</div>

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Data
t = np.array([0.0, 1.0, 2.0, 3.0, 4.0])  # seconds
x = np.array([0.00, 2.05, 4.10, 6.05, 8.10])  # meters

# Linear fit: x = v*t + x0
coeffs = np.polyfit(t, x, 1)  # slope, intercept
v, x0 = coeffs
fit_line = v * t + x0

# Plot
plt.scatter(t, x, marker='o')
plt.plot(t, fit_line, 'r-', label=f'Fit: v={v:.2f} m/s, x0={x0:.2f} m')


# Labels and title
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Displacement vs Time for Uniform Motion')
plt.legend()
plt.grid(True)

plt.show()
```

In static equilibrium, a hanging mass stretches the spring until the upward spring force balances the downward
gravitational force. This condition is expressed by

$$mg = kx,$$

where $m$ is the mass, $g$ is the gravitational acceleration, $k$ is the spring constant, and $x$ is the displacement
from the spring’s natural length.

Solving for $x$ gives

$$x = \frac{g}{k}\, m.$$

This equation shows that displacement is **proportional to mass**, and the constant of proportionality is

$$\frac{g}{k}.$$

Therefore, when we plot $x$ versus $m$ and fit a straight line, the **slope of that line is equal to $g/k$**.  
Once the slope is known, we can determine the spring constant:

$$k = \frac{g}{\text{slope}}.$$

In other words, the straight-line fit allows us to extract a physical property of the spring directly from the graph.

## D2.7.5 Linearizing Data

Not all relationships are linear.  
If your data follow:

$$
y \propto x^n
$$

then plotting $\log y$ vs $\log x$ will produce a straight line  
with slope equal to $n$.

**Example:** Period of a pendulum

$$
T = 2 \pi \sqrt{\frac{L}{g}}
$$

Taking logs:

$$
\log T = \frac{1}{2} \log L + \text{constant}
$$

A $\log T$ vs $\log L$ plot will have slope $0.5$.

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Example pendulum data
L = np.array([0.20, 0.40, 0.60, 0.80, 1.00])  # length in meters
T = np.array([0.90, 1.27, 1.55, 1.80, 2.02])  # period in seconds

# Transform data to logs (base 10)
log_L = np.log10(L)
log_T = np.log10(T)

# Linear fit in log-log space: log(T) = n*log(L) + c
coeffs = np.polyfit(log_L, log_T, 1)
n, c = coeffs
fit_line = n * log_L + c

# Plot log-log data and fit
plt.scatter(log_L, log_T, label='Data', color='blue')
plt.plot(log_L, fit_line, 'r-', label=f'Fit: slope={n:.3f}, intercept={c:.3f}')

# Labels and title
plt.xlabel(r'$\log_{10}(L \ \mathrm{[m]})$')
plt.ylabel(r'$\log_{10}(T \ \mathrm{[s]})$')
plt.title('Log–Log Plot: Pendulum Period vs Length')
plt.legend()
plt.grid(True)

plt.show()
```

## D2.7.6 Common Graphing Mistakes

- **Missing units** on axes  
- **Connecting scattered data points** without a reason  
- **Using decorative but non-informative elements** (3D effects, excessive grid lines)  
- **Poor scale choice** — e.g., cutting off zero when it matters for interpretation  
- **Not showing uncertainty** when it’s significant

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D2.7.7 Summary

- Graphs are a **core communication tool** in physics.
- Always label axes with **quantity and units**.
- Choose scales and formats that reveal trends clearly.
- Show uncertainties when known.
- Use transformations (e.g., log plots) to reveal underlying relationships.
