## D2.1.1 Steps to Conduct Science
The scientific method is not a theory but a set it steps that naturally follow human curiosity for understanding the physical universe or universes &#128526;. The steps are:

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">A More Concise Programming Approach</h3>
    
* Make an **observation**.
* Explain the observation through a **hypothesis*.
* Design and **experiment** to test the hypothesis.
* If the experiment verifies the hypothesis, establish a **theory**.
</div>

where the buzz words are in bold face.

Sometimes, there is listed a fifth step: **law**, but this one should be omitted as science is an ever evolving subject. While the theories we have today can explain all our current observations and may suggest labeling that theory as a law, we may find a contradictory observation tomorrow. 

<div style="background-color:#ffe6e6; border-left:6px solid #990000; padding:14px; border-radius:4px;">
NOTE: Any theory is wrong if not in agreement with observation(s).
</div><br>

This fact was best put in words by physicist extraordinaire [Richard Feynman](https://en.wikipedia.org/wiki/Richard_Feynman): "We can only prove we are not wrong according to our current understanding, but never prove we are right."

<h3 style="margin-top:0; color:#000000;">Box Activity 1 – The Mysterious Cooling Cup</h3>

You notice that cups of hot coffee in the student lounge cool down to room temperature in about an hour.  
A friend claims that **stirring the coffee every few minutes makes it cool faster** because it “lets the heat escape,” while another insists **stirring makes no difference**.

Use the **scientific method** to outline how you would study this question:

1. **Observation**: State the initial observation and why it is interesting.
2. **Question**: Frame a clear, testable question.
3. **Hypothesis**: Propose at least two competing hypotheses (e.g., stirring speeds cooling vs. stirring does nothing).
4. **Experiment**: Describe a simple experiment using everyday materials. Identify what you control and what you measure.
5. **Analysis**: Explain what data to collect and how you would compare results.
6. **Conclusion**: State how you would decide which hypothesis is supported.

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**1. Observation: Coffee cools over time; people disagree on whether stirring affects the cooling rate.**

**2. Question: Does stirring a liquid change its cooling rate?**

**3. Hypotheses:**

* H1: Stirring increases the cooling rate.
* H2: Stirring has no effect on cooling rate.
* H3: Stirring slows cooling by mixing hotter fluid from below.

**4. Experiment**:
Prepare two cups of hot coffee with the same initial temperature, same cup type, same volume, and same room environment.  
Stir one cup every 2 minutes for 30 seconds; leave the other untouched. Measure temperature every 5 minutes with a thermometer.

**5. Analysis**:
Plot temperature vs. time for both cups and compare the curves to see if cooling rates differ.<br><br>

**6. Conclusion**:
Determine whether the stirred cup cools faster, the same, or slower.  
Select the hypothesis most strongly supported by the data.

<h3 style="margin-top:0; color:#000000;">Box Activity 2 – Do Masks Reduce the Spread of Cough Droplets?</h3>


You are shadowing at a clinic and notice two patients coughing while wearing surgical masks.  
A classmate wonders aloud: <i>“Do masks actually reduce how far cough droplets travel?”</i>  
Another student replies: <i>“I think masks don’t matter — the droplets are so small they go right through.”</i><br><br>

**Task:** Use the scientific method to approach this question:

1. State the **observation**.
2. Frame a clear **research question**.
3. Propose **competing hypotheses**.
4. Design an **experiment** (using safe, simple classroom materials).
5. Describe how you would **analyze the data**.
6. State what kind of **conclusion** you could draw.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**1. Observation**
Patients wear masks while coughing in the clinic. Some people believe this reduces how far droplets travel; others doubt that masks make a difference.

**2. Research Question**
Does wearing a surgical mask reduce the distance that cough droplets travel?

**3. Hypotheses**

* H1: Masks significantly reduce the distance and number of droplets expelled.
* H2: Masks have no effect on droplet distance.
* H3: Masks reduce only large droplets but not small aerosol-like ones.

**4. Experiment Design**

1. Use a spray bottle with colored water to simulate a cough.
2. Place sheets of paper at set distances (10 cm, 20 cm, 30 cm, etc.).
3. Spray once without a mask over the nozzle, once with a surgical mask secured over the nozzle.
4. Repeat each condition at least 3 times to reduce random variability.
5. Record the number of droplets on each sheet and the farthest distance reached.

**Controlled Variables**: Keep spray angle, height, number of sprays, and room conditions (no fans or wind) the same for all trials.

**5. Data Analysis**

* Compare the mean droplet counts at each distance for “mask” vs. “no mask.”
* Compare the maximum distance at which droplets are observed.
* Compute the percent reduction in droplets at each distance:

$$\% \text{ reduction} = \frac{N_{\text{no mask}} - N_{\text{mask}}}{N_{\text{no mask}}} \times 100$$


**6. Conclusion**

* If droplet counts and maximum distance are consistently lower with a mask, the data support H1.
* If results overlap within experimental uncertainty, H2 is supported.
* If large droplet spread is reduced but a fine mist still reaches far, H3 may be supported.

**7. Broader Relevance**
These results connect directly to infection-control practices in clinics and public spaces.  
However, you should note limitations: a spray bottle is only an approximation of a real cough, and real-world mask fit, humidity, and breathing patterns may differ from your classroom setup.

## D2.1.2 A More Rigorous Approach
We are often presented with a problem where there is no known solution, in the sense that no one else has previosly solved the problem. Hence, the list above can present itself in several ways. 

When we attempt to solve the problem or explain the observation, we must carefully design an experiment that is capable of addressing the issue at hand. Here are two examples of conducting the same experiment:

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Applying the Scientific Method to a 100-m Dash</h3>

**Observation**
I recently ran a 100-m dash and recorded a time of **14.2 seconds**. This observation raises a natural question about whether I could run faster under better, controlled conditions.

**Question**

*Can I run a 100-m dash in under 14 seconds when measured with a reliable, controlled timing system?*

**Hypothesis**
* **H₁:** Under proper conditions (warm-up, focus, controlled timing), I can run a sub-14-second 100-m dash.

This is specific, measurable, and testable — exactly what a scientific hypothesis should be.

**Experiment – First Attempt (Flawed Design)**
I recruit a friend to time me with a handheld stopwatch at the Springville High School track.  
The measured time: **13.9 seconds**.

This *appears* to support the hypothesis, but the scientific method requires us to examine the *quality* of the measurement.

Problems with this design:
* Human reaction time introduces 0.1–0.3 s error at both start and stop.
* This creates **observational bias**.
* Uncertainty is large enough that “13.9 s” might not actually be below 14.0 s.

This demonstrates an essential idea in science:  
**The reliability of a conclusion depends on the reliability of the measurement.**

**Improved Experiment – Reduced Bias & Higher Precision**
To improve the test:

* Replace the human stopwatch with **Arduino motion sensors**.  
* Use **GPS-synchronized timing** (millisecond accuracy).  
* Use break-beam sensors at the start and finish lines.  
* Eliminate all reaction-time delay.

With this upgraded method, a time of **13.9 seconds** is trustworthy and reflects the true performance within a known uncertainty.

This aligns with the scientific method:
1. Control variables  
2. Reduce bias  
3. Improve measurement precision  
4. Use appropriate instrumentation  

**Analysis**
The improved design produces data that are:

* Precise  
* Repeatable  
* Free of human-timing errors  

This allows a fair and unbiased evaluation of the hypothesis.

**Conclusion**
If repeated automated measurements consistently show times below 14.0 seconds, the hypothesis is supported.  
If not, the hypothesis must be rejected or revised.

## D2.1.3 Scientific Message
The 100-m dash example highlights a central theme of scientific work: a measurement is only meaningful if we understand its limitations. Even when an experiment appears simple, issues such as reaction time, bias, and instrument precision can dramatically change the conclusion. This naturally motivates the next topics in our scientific toolkit: significant figures and the basic handling of data, including the mean, median, standard deviation, and standard error of the mean. These tools allow us to quantify uncertainty, summarize variability, and communicate results honestly and transparently. Before we can trust or interpret any measurement, we must know not just the number itself—but how reliable that number truly is.

As scientists, we should avoid relying on human observations as our senses are not scientifically qualified! 

>Human senses are not scientific instruments.  
To follow the scientific method properly, we must:
>* minimize bias  
>* increase precision  
>* control conditions  
>* rely on calibrated measurement systems
>   
>Better measurement → better science.
