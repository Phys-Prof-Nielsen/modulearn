## Assessment Problem A1 — Scientific Notation

A physicist measures the intensity of a faint light source and records a value of  

$$ I = 0.000\,004\,52~\text{W/m}^2. $$

**Task:**  
Write the intensity $I$ in proper **scientific notation** with **three significant figures**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A1
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">

We begin by identifying the first nonzero digit in  
$0.00000452$ → the `4` in the millionths place.

Move the decimal so the coefficient is between 1 and 10:

$$
0.00000452 = 4.52 \times 10^{-6}.
$$

The exponent is $-6$ because we moved the decimal **six places to the right**.

- The coefficient $4.52$ has **three significant figures**.
- The unit remains $\text{W/m}^2$.

**Final answer:**

$$
I = 4.52 \times 10^{-6}~\text{W/m}^2.
$$

## Assessment Problem A2 — Orders of Magnitude

A small drone has a battery that stores  

$$ E = 72\,000~\text{J}. $$

A typical lightning strike releases approximately 

$$ E_{\text{lightning}} \approx 1\times10^{9}~\text{J}. $$

**Task:**  
Determine the **order of magnitude** of the ratio

$$ \dfrac{E}{E_{\text{lightning}}}, $$  
and briefly state what this comparison means physically.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A2
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">

First convert the drone battery energy into scientific notation:

$$
E = 7.2\times10^{4}~\text{J}.
$$

Compute the ratio:

$$
\frac{E}{E_{\text{lightning}}}
= \frac{7.2\times10^{4}}{1\times10^{9}}
= 7.2\times10^{-5}.
$$

---

**Order-of-magnitude rounding**
Order-of-magnitude rounding uses a *logarithmic midpoint* at 

$$3.16 \approx \sqrt{10}.$$

- If the coefficient is **below 3.16**, we round down.  
- If the coefficient is **above 3.16**, we round up.

Because:

$$
7.2 > 3.16,
$$

the ratio is closer to 

$$
10^{-4} \quad\text{than}\quad 10^{-5}.
$$

---

**Final order of magnitude**

$$
\boxed{10^{-4}}
$$

---

**Interpretation**
A lightning strike releases roughly  

$$10^{4} = 10{,}000$$  

times more energy than the drone battery.  
Equivalently, the drone battery holds only about one ten-thousandth of the energy in a single lightning strike.

## Assessment Problem A3 — Significant Figures (Multiplication & Division)

A metal rod has a measured diameter  

$$ d = 0.86~\text{cm}. $$

You want to compute the **cross-sectional area** assuming the rod is cylindrical.
**Task:**  
Compute the area $A$ and report the final answer with the **correct number of significant figures**.

Use $\pi = 3.14159$ (exact for sig-fig purposes).


<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A3
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">


We begin by identifying significant figures:

- Diameter $d = 0.86$ has **2 significant figures**, and it will determine the sig figs of the final result.  
- $\pi$ is treated as an **exact** mathematical constant (infinite sig figs).

---

**Step 1: Compute the radius**

$$
r = \frac{d}{2} = \frac{0.86}{2} = 0.43~\text{cm}.
$$

The intermediate value 0.43 has **2 significant figures**, consistent with the input.

---

**Step 2: Compute the area**

$$
A = \pi r^2 = 3.14159 \times (0.43)^2.
$$

and we get
$$
A = 0.5808~\text{cm}^2.
$$

---

**Step 3: Apply significant-figure rules**

Since the diameter had **2 significant figures**, the final answer must also have **2 significant figures**.

Round $0.5808$ to **2 sig figs**:

- First two sig figs: $0.58$  
- Next digit: $0$ (so we do not round up)

Final rounded area:

$$
A = 0.58~\text{cm}^2.
$$

---

**Final Answer**

$$
\boxed{A = 0.58~\text{cm}^2}
$$

## Assessment Problem A4 — Significant Figures (Addition & Subtraction)

In a timing experiment, a student measures how long a cart takes to travel three segments of a track:

- Segment 1: $t_1 = 1.204~\text{s}$
- Segment 2: $t_2 = 0.83~\text{s}$
- Segment 3: $t_3 = 2.1150~\text{s}$

The **total travel time** is

$$
T = t_1 + t_2 + t_3.
$$

**Task:**  
Compute $T$ and report the final answer with the **correct number of digits after the decimal point** (addition/subtraction rule).

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A4
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">


**1. Identify decimal precision (not just sig figs)**

For **addition and subtraction**, the rule is:

> The final answer must be rounded to the **least precise decimal place** among the terms.

Check each time:

- $t_1 = 1.204~\text{s}$ → thousandths place (3 digits after decimal)  
- $t_2 = 0.83~\text{s}$ → hundredths place (2 digits after decimal)  
- $t_3 = 2.1150~\text{s}$ → ten-thousandths place (4 digits after decimal)

The **least precise** is $t_2$ (hundredths place),  
so the **final answer must be rounded to the hundredths place**.

---

**2. Add the times**

First, add them normally (keep extra digits during the calculation):

$$
T = t_1 + t_2 + t_3
= 1.204 + 0.83 + 2.1150.
$$

Compute step by step:

$$
1.204 + 0.83 = 2.034
$$

$$
2.034 + 2.1150 = 4.1490
$$

So the raw sum is:

$$
T = 4.1490~\text{s}.
$$

---

**3. Apply the decimal-place rule**

We must report $T$ to the **hundredths place**, i.e., two digits after the decimal.

- $4.1490 \rightarrow 4.15$ (because the third decimal digit is 9, so we round up)

---

**Final Answer**

$$
\boxed{T = 4.15~\text{s}}
$$

This follows the **addition/subtraction rule**: match the **least precise decimal place** (here, the hundredths from $t_2$).

## Assessment Problem A5 — Basic Statistics (Mean, Standard Deviation, SEM)

A student times one full oscillation of a pendulum **five times** and records:

- $T_1 = 2.01~\text{s}$
- $T_2 = 1.98~\text{s}$
- $T_3 = 2.05~\text{s}$
- $T_4 = 2.00~\text{s}$
- $T_5 = 1.99~\text{s}$

**Tasks:**

1. Compute the **mean period** $\bar{T}$.  
2. Compute the **sample standard deviation** $s$ of the measurements.  
3. Compute the **standard error of the mean (SEM)**, $s_{\bar{T}} = s / \sqrt{N}$, where $N$ is the number of measurements.  
4. Report the period in the two common formats:
   - As a **single measurement with spread**: $\bar{T} \pm s$  
   - As a **mean with precision**: $\bar{T} \pm s_{\bar{T}}$  
   Use a sensible number of significant figures in each.


<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A5
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">


**1. Mean period $\bar{T}$**

Add the times:

$$
2.01 + 1.98 + 2.05 + 2.00 + 1.99 = 10.03~\text{s}.
$$

Number of measurements: $N = 5$.

Mean:

$$
\bar{T} = \frac{10.03}{5} = 2.006~\text{s}.
$$

The raw mean is $2.006~\text{s}$. Since the original data are given to the **hundredths place**, it is reasonable to report the mean to the **hundredths place** as well:

$$
\bar{T} \approx 2.01~\text{s}.
$$

---

**2. Sample standard deviation $s$**

Formula for the **sample** standard deviation:

$$
s = \sqrt{\frac{1}{N-1}\sum_{i=1}^{N} (T_i - \bar{T})^2}.
$$

Using the raw mean $2.006~\text{s}$, one finds (by calculator or Python):

$$
s \approx 0.0270~\text{s}.
$$

We typically keep **1–2 significant figures** in the standard deviation. Here we’ll report:

$$
s \approx 0.03~\text{s}.
$$

---

**3. Standard error of the mean (SEM)**

The standard error of the mean is

$$
s_{\bar{T}} = \frac{s}{\sqrt{N}}.
$$

With $s \approx 0.0270~\text{s}$ and $N = 5$:

$$
s_{\bar{T}} = \frac{0.0270}{\sqrt{5}}
\approx \frac{0.0270}{2.24}
\approx 0.012~\text{s}.
$$

Again, 1–2 significant figures is typical. Since we report the period the hundredths:

$$
s_{\bar{T}} \approx 0.01~\text{s}.
$$

---

**4. Reporting the result**

**As a single measurement with spread (shows scatter of the data):**

$$
\bar{T} \pm s \approx (2.01 \pm 0.03)~\text{s}.
$$

This says: *individual* measurements typically vary by about $0.03~\text{s}$ around the mean.

---

**As a mean with precision (shows how well the mean is known):**

$$
\bar{T} \pm s_{\bar{T}} \approx (2.01 \pm 0.01)~\text{s}.
$$

This says: based on 5 trials, the **best estimate of the true period** is $2.01~\text{s}$ with an **uncertainty of about $0.01~\text{s}$ in the mean**.

---

**(Optional) Python check**

You can verify your calculations with a short Python snippet:

`T_mean = np.mean(T)`  
`s = np.std(T, ddof=1)          # sample standard deviation`  
`sem = s / np.sqrt(len(T))`  

`print(f"Mean T: {T_mean:.3f} s")`  
`print(f"Sample std dev s: {s:.3f} s")`  
`print(f"SEM: {sem:.3f} s")`  

Expected output:

`Mean T: 2.006 s`  
`Sample std dev s: 0.027 s`  
`SEM: 0.012 s`

## Assessment Problem A6 — Unit Conversions (Factor–Label Method)

A student weighs **150. lb** on a bathroom scale (in pounds-force, lb).  
We want to find their **weight in newtons** (SI unit of force).

Use the following information:

- $1~\text{lb} = 4.448~\text{N}$  
- $1~\text{N} = 1~\text{kg·m/s}^2$

**Task:**  
Convert the student’s weight from pounds to newtons using the **factor–label method**.  
Show the intermediate step with the conversion factor written as a “number 1,” and report the final answer with the **correct number of significant figures**.

Optionally, verify your result with a short Python calculation.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Solution to Problem A6
</summary>
<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:8px;">


**1. Identify given and needed quantities**

Given:

- Measured weight: $W = 150.~\text{lb}$  

We want:

- $W$ in newtons (N).

Conversion factor (our “number 1”):

$$
1 = \frac{4.448~\text{N}}{1~\text{lb}}
$$

This equals 1 because $4.448~\text{N}$ and $1~\text{lb}$ represent the **same physical force**.

---

**2. Set up the factor–label conversion**

Start with the given value and multiply by the conversion factor:

$$
W = 150.~\text{lb} \times \frac{4.448~\text{N}}{1~\text{lb}}.
$$

Units:

- $\text{lb}$ in numerator and denominator cancel, leaving newtons.

Numerically:

$$
W = 150. \times 4.448~\text{N}.
$$

Compute the product:

$$
W = 667.2~\text{N}.
$$

---

**3. Significant figures**

The given weight is $150.$ with a decimal point, which we interpret as **3 significant figures**.

- Conversion factor $4.448$ has **4 significant figures**.
- The result is limited by the **fewest** significant figures: **3 sig figs**.

So we round $667.2$ to **3 significant figures**:

- $667.2 \rightarrow 667~\text{N}$ (since the next digit is 2, we do not round up).

---

**Final Answer**

$$
\boxed{W \approx 667~\text{N}}
$$

The student’s weight is about $667~\text{N}$ in SI units.

---

**4. Optional Python check**

You can verify with a simple Python snippet:

`conv = 4.448      # N per lb`  

`W_N = W_lb * conv`  

`print(f"Weight in newtons (raw): {W_N:.4f} N")`  
`print(f"Weight in newtons (3 sig figs): {W_N:.3g} N")`  

Expected output:

`Weight in newtons (3 sig figs): 667 N`
