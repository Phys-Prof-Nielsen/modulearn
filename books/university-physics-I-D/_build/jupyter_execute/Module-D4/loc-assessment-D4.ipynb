## Problem 1: Kinetic Energy as a Property of Motion

**LOC addressed:** A1, A2, A3, A4

A point particle has mass

$$
m = 1.50\ \text{kg}
$$

and velocity

$$
v_x = -12.0\ \text{m/s}.
$$

1. Compute the particle’s **kinetic energy**.  
2. Explain why the kinetic energy is **positive**, even though the velocity is negative.  
3. Explain what a **change** in kinetic energy would indicate physically for this particle.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 1 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

Kinetic energy in one dimension is defined as

$$
k = \frac{1}{2} m v^2.
$$

Substitute the given values:

$$
k = \frac{1}{2}(1.50)(12.0^2)
= 108\ \text{J}.
$$

---

**Interpretation**

- Kinetic energy depends on the **square of the speed**, not the direction of motion.  
- The velocity is negative because the particle moves in the negative $x$-direction, but squaring the speed makes the kinetic energy **positive**.  
- A **change in kinetic energy** would indicate that **energy has been transferred** to or from the particle through work done by an interaction.

$$
\boxed{k = 1.08\times10^{2}\ \text{J}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 1 — Kinetic Energy as a Property of Motion

import sympy as sp
from IPython.display import display, Math

# Define symbols
m, v = sp.symbols('m v', real=True)

# Kinetic energy definition
k = 0.5 * m * v**2

# Numerical values
values = {
    m: 1.50,     # kg
    v: -12.0     # m/s
}

# Substitute values
k_val = sp.N(k.subs(values))

# Display symbolic expression
display(Math(r"k = \frac{1}{2} m v^2"))

# Display numerical result
display(Math(r"k = " + f"{float(k_val):.2e}" + r"\ \text{J}"))
```

## Problem 2: Work as Energy Transfer in One Dimension

**LOC addressed:** B1, B2, B3, B4, C1

A crate of mass

$$
m = 6.00\ \text{kg}
$$

slides along a horizontal floor in one dimension.  
A **constant net force**

$$
f_x = -18.0\ \text{N}
$$

acts on the crate over a **directed displacement**

$$
\Delta x = +4.0\ \text{m}.
$$

1. Compute the **net work** done on the crate.  
2. Determine the **change in kinetic energy** of the crate.  
3. State whether energy is transferred **to** or **from** the crate, and explain why based on the sign of the work.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 2 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

In one dimension, work done by a constant force is

$$
w = f_x\,\Delta x.
$$

Substitute the given values:

$$
w = (-18.0)(+4.0) = -72.0\ \text{J}.
$$

By the work–energy theorem,

$$
w_{\text{net}} = \Delta k.
$$

Thus,

$$
\Delta k = -72.0\ \text{J}.
$$

---

**Interpretation**

- The work is **negative**, meaning the force acts opposite the displacement.  
- Energy is therefore transferred **from** the crate.  
- The crate’s kinetic energy decreases by 72.0 J during this interaction.

$$
\boxed{w = \Delta k = -72.0\ \text{J}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 2 — Work as Energy Transfer (Symbolic + Numeric)

import sympy as sp
from IPython.display import display, Math

# Define symbols
f_x, dx = sp.symbols('f_x dx', real=True)

# Work definition
w = f_x * dx

# Numerical values
values = {
    f_x: -18.0,   # N
    dx: 4.0       # m
}

# Compute work
w_val = sp.N(w.subs(values))

# Display symbolic expressions
display(Math(r"w = f_x\,\Delta x"))

# Display numerical result
display(Math(r"w = " + f"{float(w_val):.1e}" + r"\ \text{J}"))

# Work–energy theorem statement
display(Math(r"\Delta k = w_{\text{net}} = " + f"{float(w_val):.1e}" + r"\ \text{J}"))

```

## Problem 3: Gravitational Potential Energy and Energy Transfer

**LOC addressed:** D1, D2, D3, C1, C3

A ball of mass

$$
m = 2.00\ \text{kg}
$$

is lifted vertically upward near Earth’s surface.  
The ball is raised from an initial height

$$
y_i = 1.5\ \text{m}
$$

to a final height

$$
y_f = 6.0\ \text{m}.
$$

Take the gravitational field strength to be

$$
g = 9.8\ \text{m/s}^2.
$$

1. Compute the **change in gravitational potential energy** of the ball.  
2. Determine the **work done by gravity** during this process.  
3. Explain the **sign relationship** between the work done by gravity and the change in gravitational potential energy.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 3 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

Near Earth’s surface, gravitational potential energy is defined as

$$
u_g = m g y.
$$

The change in gravitational potential energy is

$$
\Delta u_g = m g (y_f - y_i).
$$

Substitute values:

$$
\Delta u_g = (2.00)(9.8)(6.0 - 1.5)
= 88.2\ \text{J}.
$$

The work done by gravity is related by

$$
w_g = -\Delta u_g.
$$

Therefore,

$$
w_g = -88.2\ \text{J}.
$$

---

**Interpretation**

- Gravitational potential energy **increases** as the ball is lifted.  
- Gravity does **negative work**, since the gravitational force acts opposite the displacement.  
- Energy is transferred **into stored gravitational potential energy**.

$$
\boxed{\Delta u_g = +88.2\ \text{J}, \quad w_g = -88.2\ \text{J}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 3 — Gravitational Potential Energy and Work by Gravity

import sympy as sp
from IPython.display import display, Math

# Define symbols
m, g, y_i, y_f = sp.symbols('m g y_i y_f', real=True)

# Gravitational potential energy change
delta_u = m * g * (y_f - y_i)

# Numerical values
values = {
    m: 2.00,     # kg
    g: 9.8,      # m/s^2
    y_i: 1.5,    # m
    y_f: 6.0     # m
}

# Compute values
delta_u_val = sp.N(delta_u.subs(values))
w_g_val = -delta_u_val

# Display symbolic expressions
display(Math(r"\Delta u_g = m g (y_f - y_i)"))

# Display numerical results
display(Math(r"\Delta u_g = " + f"{float(delta_u_val):.1e}" + r"\ \text{J}"))
display(Math(r"w_g = -\Delta u_g = " + f"{float(w_g_val):.1e}" + r"\ \text{J}"))
```

## Problem 4: Speed at a Given Height (Straight Up, No Air Resistance)

**LOC addressed:** C1, C3, D1, D2, E1, E4

A ball of mass

$$
m = 0.400\ \text{kg}
$$

is tossed straight upward from height

$$
y_i = 1.20\ \text{m}
$$

with initial speed

$$
v_i = 16.0\ \text{m/s}.
$$

Ignore air resistance and use

$$
g = 9.80\ \text{m/s}^2.
$$

Before the ball reaches its peak, it passes the height

$$
y = 6.00\ \text{m}.
$$

1. Assuming mechanical energy is conserved, find the **speed** of the ball at $y=6.00$ m.  
2. Briefly explain why the speed found corresponds to the upward part of the motion (before the peak).

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 4 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

With no non-conservative forces, mechanical energy is conserved:

$$
k_i + u_i = k + u.
$$

Using

$$
k = \frac12 m v^2,
\qquad
u_g = mgy,
$$

we write

$$
\frac12 m v_i^2 + mgy_i = \frac12 m v^2 + mgy.
$$

Mass cancels, and solving for $v$ gives

$$
v = \sqrt{v_i^2 - 2g(y-y_i)}.
$$

Substitute:

$$
v = \sqrt{(16.0)^2 - 2(9.80)(6.00-1.20)}
= \sqrt{256 - 94.08}
= \sqrt{161.92}
\approx 12.7\ \text{m/s}.
$$

This is the **speed** at $y=6.00$ m.  
That height occurs once on the way up and once on the way down, but the **speed magnitude** is the same at that height (in the absence of air resistance). Since the problem states “before the peak,” we interpret this as the upward pass.

$$
\boxed{v \approx 12.7\ \text{m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 4 — Speed at a Given Height (Symbolic + Numeric)

import sympy as sp
from IPython.display import display, Math

# Symbols
g, y_i, y, v_i, v = sp.symbols('g y_i y v_i v', positive=True)

# Mechanical energy relation (mass cancels)
# (1/2)v_i^2 + g y_i = (1/2)v^2 + g y
eq = sp.Eq(sp.Rational(1, 2)*v_i**2 + g*y_i, sp.Rational(1, 2)*v**2 + g*y)

# Solve for v (speed = positive root)
sol_v = sp.solve(eq, v)[0]

# Numerical values
vals = {
    g: 9.80,
    y_i: 1.20,
    y: 6.00,
    v_i: 16.0
}

v_val = sp.N(sol_v.subs(vals))

# Display math
display(Math(r"\frac12 v_i^2 + g y_i = \frac12 v^2 + g y"))
display(Math(r"v = " + sp.latex(sol_v)))
display(Math(r"v = " + f"{float(v_val):.2e}" + r"\ \text{m/s}"))
```

## Problem 5: Mechanical Energy Changes Due to a Non-Conservative Force

**LOC addressed:** E1, E2, E3, D1, D2, C1, C3

A 1D cart of mass

$$
m = 2.50\ \text{kg}
$$

moves along a straight track. The cart starts at height

$$
y_i = 4.00\ \text{m}
$$

with speed

$$
v_i = 0.00\ \text{m/s}.
$$

It travels down to a lower height

$$
y_f = 1.20\ \text{m}.
$$

During this motion, air drag does a known amount of non-conservative work on the cart:

$$
w_{nc} = -12.0\ \text{J}.
$$

Use

$$
g = 9.80\ \text{m/s}^2.
$$

1. Write the mechanical-energy form of the work–energy theorem for this situation.  
2. Find the cart’s final speed at $y_f$.  
3. Compare the result to the case with **no** air drag: should the final speed be larger or smaller? Explain briefly.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 5 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

Mechanical energy is

$$
e_{\text{mech}} = k + u_g.
$$

The mechanical-energy form of the work–energy theorem is

$$
\Delta e_{\text{mech}} = w_{nc}.
$$

So,

$$
(k_f + u_f) - (k_i + u_i) = w_{nc}.
$$

Substitute $k_i=0$ and $u_g=mgy$:

$$
\left(\frac12 m v_f^2 + mg y_f\right) - \left(0 + mg y_i\right) = w_{nc}.
$$

Solve for $v_f$:

$$
\frac12 m v_f^2 = mg(y_i-y_f) + w_{nc}
$$

$$
v_f = \sqrt{\frac{2}{m}\left[m g(y_i-y_f) + w_{nc}\right]}.
$$

Because $w_{nc}$ is negative, the final speed is **smaller** than it would be with no air drag.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 5 — Mechanical Energy with Non-Conservative Work (Symbolic + Numeric)

import sympy as sp
from IPython.display import display, Math

# Symbols
m, g, y_i, y_f, w_nc, v_f = sp.symbols('m g y_i y_f w_nc v_f', positive=True)

# Energies
k_i = 0
u_i = m * g * y_i
k_f = sp.Rational(1, 2) * m * v_f**2
u_f = m * g * y_f

# Mechanical-energy work–energy theorem
eq = sp.Eq((k_f + u_f) - (k_i + u_i), w_nc)

# Solve for v_f (positive root)
sol_vf = sp.solve(eq, v_f)[0]

# Numerical values
vals = {
    m: 2.50,
    g: 9.80,
    y_i: 4.00,
    y_f: 1.20,
    w_nc: -12.0
}

vf_val = sp.N(sol_vf.subs(vals))

# Display math
display(Math(r"\Delta e_{\text{mech}} = w_{nc}"))
display(Math(r"\left(\frac12 m v_f^2 + mgy_f\right) - \left(mgy_i\right) = w_{nc}"))
display(Math(r"v_f = " + sp.latex(sol_vf)))

# Numerical result
display(Math(r"v_f = " + f"{float(vf_val):.3e}" + r"\ \text{m/s}"))
```

## Problem 6: Power from Energy Transfer

**LOC addressed:** F1, F2, F3, B1, C1

A motor transfers energy to a 1D cart by doing net work on it.  
The cart’s kinetic energy increases from

$$
k_i = 120\ \text{J}
$$

to

$$
k_f = 680\ \text{J}
$$

over a time interval

$$
\Delta t = 4.0\ \text{s}.
$$

1. Compute the net work done on the cart.  
2. Compute the average power delivered to the cart.  
3. Explain what the power value tells you physically (in words).

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 6 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

From the work–energy theorem,

$$
w_{\text{net}} = \Delta k = k_f - k_i.
$$

So,

$$
w_{\text{net}} = 680 - 120 = 560\ \text{J}.
$$

Average power is the rate of energy transfer:

$$
p_{\text{avg}} = \frac{w_{\text{net}}}{\Delta t}
= \frac{560}{4.0}
= 140\ \text{W}.
$$

**Interpretation:**  
A power of 140 W means the motor transfers energy to the cart at an average rate of 140 joules every second.

$$
\boxed{w_{\text{net}} = 560\ \text{J}, \quad p_{\text{avg}} = 140\ \text{W}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem 6 — Power from Energy Transfer (Symbolic + Numeric)

import sympy as sp
from IPython.display import display, Math

# Symbols
k_i, k_f, dt = sp.symbols('k_i k_f dt', positive=True)

# Work-energy theorem
w_net = k_f - k_i

# Average power
p_avg = sp.simplify(w_net / dt)

# Numerical values
vals = {
    k_i: 120,
    k_f: 680,
    dt: 4.0
}

w_val = sp.N(w_net.subs(vals))
p_val = sp.N(p_avg.subs(vals))

# Display math
display(Math(r"w_{\text{net}} = \Delta k = k_f - k_i"))
display(Math(r"w_{\text{net}} = " + sp.latex(w_net)))
display(Math(r"p_{\text{avg}} = \frac{w_{\text{net}}}{\Delta t} = " + sp.latex(p_avg)))

# Numerical results
display(Math(r"w_{\text{net}} = " + f"{float(w_val):.3e}" + r"\ \text{J}"))
display(Math(r"p_{\text{avg}} = " + f"{float(p_val):.3e}" + r"\ \text{W}"))
```

## Problem 7: Work from a Force–Position Graph and Final Speed

**LOC addressed:** B1, B2, B4, C1, C3

A 1D cart of mass

$$
m = 3.00\ \text{kg}
$$

moves along the +x direction on a track. A net force acts on the cart as a function of position.  
The measured force data are:

| Position $x$ (m) | 0.00 | 0.20 | 0.40 | 0.60 | 0.80 | 1.00 |
|------------------|------|------|------|------|------|------|
| Net Force $F$ (N) | 0.0  | 10.0 | 18.0 | 16.0 | 8.0  | 0.0  |

Assume the cart starts from rest at $x=0$.

1. Plot $F$ vs. $x$.  
2. Use a trapezoidal-sum (area under the curve) to estimate the net work done from $x=0$ to $x=1.00$ m.  
3. Use the work–energy theorem to find the final speed $v_f$ at $x=1.00$ m.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 7 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

The net work is the area under the force–position curve:

$$
w_{\text{net}} \approx \sum \frac{F_i + F_{i+1}}{2}\,\Delta x.
$$

Then apply the work–energy theorem (starting from rest):

$$
w_{\text{net}} = \Delta k = \frac12 m v_f^2,
\qquad
v_f = \sqrt{\frac{2w_{\text{net}}}{m}}.
$$

The Python script below computes the trapezoidal work and then the final speed.

</div>
</details>


```python
# DIY Cell
```

```python
# Problem 7 — Work from F(x) graph (trapezoid) and final speed

import sympy as sp
from IPython.display import display, Math

# -----------------------------
# Data
# -----------------------------
x_data = [0.00, 0.20, 0.40, 0.60, 0.80, 1.00]   # m
F_data = [0.0, 10.0, 18.0, 16.0, 8.0, 0.0]      # N

# Symbols
m, W, v_f = sp.symbols('m w v_f', positive=True)

# Trapezoidal work (symbolic sum evaluated numerically)
W_num = 0.0
for i in range(len(x_data) - 1):
    dx = x_data[i+1] - x_data[i]
    W_num += 0.5 * (F_data[i] + F_data[i+1]) * dx

# Work-energy theorem from rest: W = (1/2) m v_f^2
eq = sp.Eq(W, sp.Rational(1, 2) * m * v_f**2)

# Solve for v_f
sol_vf = sp.solve(eq, v_f)[0]  

# Substitute numerical values
m_val = 3.00
vf_val = sp.N(sol_vf.subs({W: W_num, m: m_val}))

# -----------------------------
# Plot (matplotlib) — simple and clear
# -----------------------------
import matplotlib.pyplot as plt

plt.figure()
plt.plot(x_data, F_data, marker='o')
plt.xlabel("x (m)")
plt.ylabel("F (N)")
plt.title("Net Force vs Position")
plt.grid(True)
plt.show()

# -----------------------------
# Display math + results
# -----------------------------
display(Math(r"w_{\text{net}} \approx \sum \frac{F_i + F_{i+1}}{2}\,\Delta x"))
display(Math(r"w_{\text{net}} = " + f"{W_num:.1e}" + r"\ \text{J}"))

display(Math(r"w_{\text{net}} = \frac12 m v_f^2"))
display(Math(r"v_f = " + sp.latex(sol_vf)))

display(Math(r"v_f = " + f"{float(vf_val):.1e}" + r"\ \text{m/s}"))
```
