## D4.8.1 Motivation — How Fast Is Energy Transferred?

So far, we have focused on **how much energy** is transferred to or from a system through work.  
However, in many real situations, an equally important question is:

> **How fast does that energy transfer occur?**

Two machines may both transfer the same amount of energy, but if one does so in half the time, we experience it as being *more powerful*.  
This distinction matters in contexts ranging from engines and motors to human performance and electrical devices.

To capture this idea, we introduce a new quantity: **power**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.8.2 Conceptual Definition of Power

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Power</h3>

<strong>Power is the rate at which energy is transferred.</strong>

</div>

Power does not tell us how much energy is transferred —  it tells us **how quickly** that transfer occurs.

---

## D4.8.3 Average Power: Mathematical Definition

Mathematically, we define the **average power** over a time interval $\Delta t$ as

$$
P_{\text{avg}} = \frac{W_{\text{net}}}{\Delta t}.
$$

At first glance, this expression may look slightly uncomfortable from a **mathematical point of view**.

We are used to defining rates as a **ratio of changes**, such as

$$
\frac{\Delta x}{\Delta t}
\quad \text{or} \quad
\frac{\Delta v}{\Delta t}.
$$

By contrast, the expression
$$
P_{\text{avg}} = \frac{W_{\text{net}}}{\Delta t}
$$

appears to place a single quantity in the numerator rather than a difference.

This apparent asymmetry is what makes the definition feel uncomfortable to the mathematical eye.

The resolution comes from remembering what **work actually represents**.

Work is not an independent state variable—it is *already a change*.  
Specifically, work is the **change in kinetic energy**:

$$
W_{\text{net}} = \Delta k.
$$

Substituting this into the power definition gives

$$
P_{\text{avg}} = \frac{\Delta k}{\Delta t},
$$

which has the familiar and expected structure of a **ratio of differences**.

The definition is therefore mathematically consistent; the change is simply embedded in the definition of work itself. Writing power in terms of work emphasizes the physical process of **energy transfer**, while writing it in terms of $\Delta k$ makes the mathematical structure explicit.

Later, in Phase B, this structure will be made fully transparent using calculus, where instantaneous power is defined as

$$
P = \frac{dE}{dt}.
$$


which removes any ambiguity and makes the structure of the definition completely explicit.

For now, it is enough to remember:

- Work tells us **how much** energy is transferred.  
- Power tells us **how fast** that energy transfer occurs.

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Average Power from Energy Transfer</h3>

<strong>Problem</strong>  

A motor transfers $3.6 \times 10^{4}\ \text{J}$ of energy to a cart over a time interval of $12\ \text{s}$.

1. What is the <em>average power</em> delivered by the motor?  
2. Interpret the result physically.

---

<strong>Methodology</strong>  

Average power is defined as the rate of energy transfer:

$$
P_{\text{avg}} = \frac{W_{\text{net}}}{\Delta t}.
$$

Because work represents a change in energy, this expression can also be viewed as

$$
P_{\text{avg}} = \frac{\Delta E}{\Delta t}.
$$

---

<strong>Solution</strong>  

Substitute the given values:

$$
P_{\text{avg}} = \frac{3.6 \times 10^{4}\ \text{J}}{12\ \text{s}} = 3.0 \times 10^{3}\ \text{J/s}.
$$

Since $1\ \text{J/s} = 1\ \text{W}$, this becomes

$$
P_{\text{avg}} = 3.0 \times 10^{3}\ \text{W}.
$$

---

<strong>Answer</strong>  

$$
\boxed{P_{\text{avg}} = 3.0 \times 10^{3}\ \text{W}}
$$

---

<strong>Interpretation</strong>  

The motor transfers energy to the cart at an average rate of $3000\ \text{W}$.  
This tells us **how fast** energy is delivered, not how much energy is transferred in total.

Two motors could transfer the same total energy, but the one that does so in less time would have the greater power.
