## D4.2.1 Comparative Concept: Impulse and Momentum Transfer

In an earlier module, we introduced **impulse** as a way to describe how interactions change motion.

Rather than starting with force, we defined **net linear impulse** directly as a **change in linear momentum**:

$$
i_{\text{net},x} = \Delta p_x
$$

This definition emphasized a key idea:

> **Impulse records how motion changes due to an interaction.**

Only later did we connect impulse to a particular mechanism—**force acting through time**.  
The important concept was not force itself, but the **transfer of momentum**.

This perspective allowed us to reason about interactions, collisions, and conservation without relying on acceleration-based arguments.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D4.2.2 Work as Change in Kinetic Energy

In this section, we introduce an **analogous idea for energy**. In some situations, an interaction is not most naturally described by how long it acts, but by **how far** it acts. For example, we may know that a box is pushed with a constant force of 50 N across a floor through a distance of 15 m, without knowing how long the push lasted.

This motivates a second, complementary way of describing interactions.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Work as Energy Transfer</h3>

Work is the transfer of energy caused by a force acting through a displacement.

</div>

While **impulse** records interactions through **time**, **work** records interactions through **traveled distance**.

Just as impulse was defined as **change in momentum**, we define **work** as **change in kinetic energy**.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Work</h3>

Work is the net change in kinetic energy of a particle.

$$
w_{\text{net}} = \Delta k
$$

</div>

This definition emphasizes that work is not a new physical substance, but a way of **accounting for energy transfer** during an interaction.

Several important observations follow immediately:

- Work depends only on the **initial and final states** of motion through $k$.
- If the kinetic energy of a particle increases, **positive work** has been done on it (or energy was transferred **to** the particle).
- If the kinetic energy decreases, **negative work** has been done on it (or energy was transferred **from** the particle).
- If the kinetic energy does not change, the **net work is zero**, even if interactions occurred.

At this stage, we do not need to know *how* the interaction took place—only that it resulted in a change in kinetic energy.

In the next sub-module, we will connect this definition of work to a specific physical mechanism:  
**forces acting through distance**, in direct analogy with how impulse was later connected to forces acting through time.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Work as Change in Kinetic Energy (No Force Needed)</h3>

**Problem**

A 0.50 kg cart speeds up from 2.0 m/s to 6.0 m/s on a horizontal track.  
Find the <em>net work</em> done on the cart during this change in motion.

<br><br>

<strong>Solution</strong>  

We use the definition of work introduced in this section:

$$
w_{\text{net}} = \Delta k.
$$

First compute the initial and final kinetic energies:

$$
k_i = \frac{1}{2}mv_i^2 = \frac{1}{2}(0.50)(2.0)^2 = 1.0\ \text{J},
$$

$$
k_f = \frac{1}{2}mv_f^2 = \frac{1}{2}(0.50)(6.0)^2 = 9.0\ \text{J}.
$$

Then the change in kinetic energy is

$$
\Delta k = k_f - k_i = 9.0 - 1.0 = 8.0\ \text{J}.
$$

Therefore,

$$
w_{\text{net}} = 8.0\ \text{J}.
$$

<br>

<strong>Answer</strong>  

$$
\boxed{w_{\text{net}} = +8.0\ \text{J}}
$$

The positive sign means the particle gained kinetic energy: energy was transferred <em>to</em> the cart.

</div>

<h3 style="margin-top:0; color:#000000;">Box Activity 2 – Exploring How Work Depends on Speed</h3>

In this activity, you will use Python to explore how the net work depends on speed changes, using

$$
w_{\text{net}} = \Delta k = \frac{1}{2}m\left(v_f^2 - v_i^2\right).
$$

**Task:** Complete the following steps.

1. Use a particle mass of  

   $$
   m = 2.0\ \text{kg}.
   $$

2. Choose an initial speed $v_i = 0$ m/s (starting from rest).

3. Compute the kinetic energy $k$ for speeds from 0 to 10 m/s in steps of 1 m/s.

4. Make a plot of kinetic energy $k$ versus speed $v$.

5. Compute the work required to go from rest to each speed:

   $$
   w_{\text{net}}(v) = \Delta k = k(v) - k(0).
   $$

6. **Reflect**
   - Does $k$ grow linearly with $v$ or nonlinearly?
   - If you double the speed, by what factor does the required work change?
   - How does this connect to real-world stopping distance at higher speeds?

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

<strong>What you should observe</strong>

See code and output below.

- The graph of $k$ vs. $v$ curves upward because  

  $$
  k \propto v^2.
  $$

- Doubling the speed requires <strong>four times</strong> the work (starting from rest), since

  $$
  w_{\text{net}} = \Delta k \propto v^2.
  $$

<strong>Example check</strong>

From 0 to 5 m/s:

$$
w_{\text{net}} = \frac{1}{2}(2.0)(5^2) = 25\ \text{J}.
$$

From 0 to 10 m/s:

$$
w_{\text{net}} = \frac{1}{2}(2.0)(10^2) = 100\ \text{J}.
$$

Doubling the speed from 5 m/s to 10 m/s increases the required work by a factor of <strong>four</strong>, not two.

</div>
</details>


```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

m = 2.0  # kg

v = np.arange(0, 11, 1)          # speeds from 0 to 10 m/s
k = 0.5 * m * v**2               # kinetic energy
w_net = k - k[0]                 # work from rest to speed v

# Print a small table
print(" v (m/s)    k (J)     w_net (J)")
for vi, ki, wi in zip(v, k, w_net):
    print(f"{vi:6.1f}   {ki:7.2f}   {wi:9.2f}")

# Plot k vs v
plt.figure()
plt.plot(v, k, marker='o')
plt.xlabel("Speed v (m/s)")
plt.ylabel("Kinetic energy k (J)")
plt.title("Kinetic Energy vs Speed for a Point Particle")
plt.grid(True)
plt.show()
```
