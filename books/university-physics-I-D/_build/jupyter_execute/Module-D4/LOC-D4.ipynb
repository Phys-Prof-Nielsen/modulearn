# D4.0 Learning Outcomes — Work–Energy Framework
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## Why this module matters

In everyday language, *work* is often described in terms of effort, fatigue, or how hard something feels to do.  
In physics, these intuitive ideas are formalized through the concepts of **energy**, **work**, and **power**.

This module develops the **work–energy framework** step by step. We begin by introducing **kinetic energy** as a scalar property associated with motion. We then define **work** as the mechanism by which energy is transferred through interactions that act over a distance. These ideas come together naturally in the **work–energy theorem**, which connects interactions to changes in motion without relying on acceleration-based reasoning.

After establishing this foundation, we introduce **gravitational potential energy** as a stored form of energy associated with position. This allows us to distinguish between **conservative** interactions, which store energy, and **non-conservative** interactions, which transfer energy into or out of mechanical energy. These ideas are unified through the concept of **mechanical energy**, which provides a powerful bookkeeping tool for understanding motion.

We conclude the module with **power**, which describes how quickly energy is transferred.

Throughout this module, all motion and interactions are treated in **one dimension**, with direction handled using **signed quantities**. Energy itself remains a **scalar**, allowing us to focus on physical interpretation, energy transfer, and conservation rather than mathematical complexity.

By the end of this module, you will be able to:

---

### A. Kinetic Energy as a Property of Motion  
*Establishes kinetic energy as the foundational energy quantity.*

- **A1.** Describe **kinetic energy** as a scalar physical property associated with the motion of an object.  
- **A2.** Define kinetic energy in one dimension as  

  $$
  k = \frac{1}{2} m v^2
  $$

- **A3.** Explain why kinetic energy depends on the **magnitude of velocity**, not its direction.  
- **A4.** Interpret changes in kinetic energy as indicators of energy transfer.

---

### B. Work as Energy Transfer  
*Introduces work as the mechanism that changes kinetic energy.*

- **B1.** Define **work** as energy transferred by a force acting through a displacement.  
- **B2.** Calculate work done by a constant force in one dimension using  

  $$
  w = f_x\,\Delta x
  $$

- **B3.** Use the sign of work to distinguish between energy **added to** or **removed from** a particle.  
- **B4.** Interpret work physically as a process that changes an object’s kinetic energy.

---

### C. Work–Energy Theorem  
*Connects interactions to changes in motion.*

- **C1.** State and apply the **work–energy theorem**,  

  $$
  w_{\text{net}} = \Delta k
  $$

- **C2.** Interpret net work as the combined effect of multiple interactions acting over a distance.  
- **C3.** Use the work–energy theorem to predict changes in speed without using acceleration or kinematic equations.

---

### D. Gravitational Potential Energy  
*Introduces stored energy associated with position.*

- **D1.** Describe **gravitational potential energy** as energy associated with vertical position near Earth’s surface.  
- **D2.** Use the near-Earth expression  

  $$
  u_g = m g y
  $$

  without formal derivation.  
- **D3.** Relate the work done by gravity to changes in gravitational potential energy using  

  $$
  w_g = -\Delta u_g.
  $$

---

### E. Mechanical Energy and Energy Accounting  
*Unifies kinetic and potential energy.*

- **E1.** Define **mechanical energy** as  

  $$
  e_{\text{mech}} = k + u.
  $$

- **E2.** Distinguish between **conservative** and **non-conservative** forces based on how they affect mechanical energy.  
- **E3.** Apply the mechanical-energy form of the work–energy theorem,

  $$
  \Delta e_{\text{mech}} = w_{nc},
  $$

  to analyze motion.  
- **E4.** Identify conditions under which **mechanical energy is conserved**.

---

### F. Power as the Rate of Energy Transfer  
*Connects energy ideas to performance and real systems.*

- **F1.** Define **power** as the rate at which work is done or energy is transferred.  
- **F2.** Calculate average power using  

  $$
  p_{\text{avg}} = \frac{w}{\Delta t}.
  $$

- **F3.** Interpret power physically as how **quickly** energy is transferred, not how much energy is involved.

---

### Performance Criteria

- Energy quantities are treated consistently as **scalars** with clear physical interpretation.  
- Work, energy, and power are used in a **coherent conceptual hierarchy**.  
- Directional reasoning is handled through **signed forces and displacements**, not vector formalism.  
- Conservative and non-conservative interactions are clearly distinguished.  
- Explanations emphasize **energy transfer, storage, and transformation**, not memorized formulas.
