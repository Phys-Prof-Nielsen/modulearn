## D4.6.1 Motivation — Where Does the Energy Go?

In earlier sections, we saw that kinetic energy is **not always conserved**.  

During some interactions, kinetic energy decreases even though momentum remains conserved.

This raises a natural question:

> **If kinetic energy disappears, where does the energy go?**

In general, energy can be transferred into other forms, such as **internal energy** or **thermal energy**. These ideas are extremely important, but they require additional tools and concepts that will be developed later when we study **thermodynamics**.

For now, we focus on a different and more immediately useful idea: **potential energy**—a way to account for energy that is stored in the configuration of a system and can be fully converted back into kinetic energy.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D4.6.2 Potential Energy — Energy of Configuration

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Potential Energy</h3>

<strong>Potential energy is energy associated with the configuration of a system, rather than its motion.</strong>

</div>

Unlike kinetic energy, which depends on how fast objects move, potential energy depends on:
- the relative positions of objects, and
- the nature of the interaction between them.

Potential energy is not associated with individual particles in isolation.  
It belongs to a **system** of interacting objects.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.6.3 Conservative Forces 

Some forces allow energy to be stored and recovered through changes in configuration. Such forces are called **conservative forces**.

For conservative forces:
- energy transferred by the force can be stored as potential energy,
- that energy can later be fully converted back into kinetic energy.

Not all forces have this property. For example, friction converts energy into thermal energy in a way that cannot be fully recovered.

A detailed discussion of conservative and non-conservative forces will come later in **Phases C and B**. For now, we focus on the most familiar conservative interaction: **gravity**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.6.4 Gravitational Potential Energy (Near Earth)

Near Earth’s surface, the gravitational force on an object is approximately constant and directed downward.

When an object moves vertically under the influence of gravity:
- gravity does work on the object,
- that work can be associated with a change in **gravitational potential energy**.

We define the change in gravitational potential energy as

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Gravitational Potential Energy (Near Earth)</h3>

$$
\Delta U_g = mg\,\Delta y
$$

</div>

Here,
- $m$ is the mass of the object,
- $g$ is the magnitude of the gravitational acceleration,
- $\Delta y$ is the vertical displacement.

---

### Important Notes

- Only **changes** in gravitational potential energy matter.  
- The zero level of potential energy is a **choice**, not a physical requirement.  
- Raising an object increases $U_g$; lowering it decreases $U_g$.

Gravitational potential energy provides a way to keep track of energy when motion changes due to gravity, even when no contact forces are present.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Change in Gravitational Potential Energy</h3>

<strong>Problem</strong>  

A $2.0\ \text{kg}$ book is lifted vertically upward by $1.5\ \text{m}$ near Earth’s surface.

Find the change in the book’s gravitational potential energy.

---

<strong>Solution</strong>  

Using

$$
\Delta U_g = mg\,\Delta y,
$$

we substitute the given values:

$$
\Delta U_g = (2.0)(9.8)(1.5) = 29.4\ \text{J}.
$$

---

<strong>Answer</strong>  

$$
\boxed{\Delta U_g = +29\ \text{J}}
$$

---

<strong>Interpretation</strong>  

The gravitational potential energy increases because the object is raised against the force of gravity.  
That energy can later be converted back into kinetic energy if the object is released.

</div>
