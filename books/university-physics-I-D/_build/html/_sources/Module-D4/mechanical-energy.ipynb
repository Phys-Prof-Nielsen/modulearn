# D4.7 Mechanical Energy and the Work–Energy Theorem
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.7.1 Motivation — Combining Motion and Configuration

So far, we have introduced two distinct ways that energy can appear in a system:

- **Kinetic energy** $k$, associated with motion  
- **Potential energy** $u$, associated with configuration (position)

In many physical situations, energy moves back and forth between these two forms without being lost.  
To describe this exchange efficiently, we introduce a new quantity: **mechanical energy**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.7.2 Mechanical Energy

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Mechanical Energy</h3>

<strong>Mechanical energy is the sum of kinetic and potential energy:</strong>

$$
e_{\text{mech}} = k + u
$$

</div>

Mechanical energy accounts for both:
- how fast objects are moving, and
- how they are positioned within a system.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.7.3 Rewriting the Work–Energy Theorem

We begin with the work–energy theorem in its original form:

$$
w_{\text{net}} = \Delta k.
$$

The net work can be separated into two contributions:

- work done by **conservative forces**, $w_c$  
- work done by **non-conservative forces**, $w_{nc}$  

Thus,

$$
w_{\text{net}} = w_c + w_{nc}.
$$

Substituting into the work–energy theorem gives

$$
w_c + w_{nc} = \Delta k.
$$

For conservative forces, we define potential energy such that

$$
w_c = -\Delta u.
$$

This relationship reflects the fact that work done by a conservative force is stored as potential energy.

Substituting this into the previous equation yields

$$
-\Delta u + w_{nc} = \Delta k.
$$

Rearranging,

$$
\Delta k + \Delta u = w_{nc}.
$$

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

### Mechanical Energy Form of the Theorem

Recognizing that $\Delta(k + u) = \Delta e_{\text{mech}}$, we arrive at the **mechanical energy form of the work–energy theorem**:

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Work–Energy Theorem (Mechanical Energy Form)</h3>

$$
\Delta e_{\text{mech}} = w_{nc}
$$

</div>
<br>
<div style="margin-top:10px; padding:10px; background-color:#d6e4ff; border-left:4px solid #003366;">
<strong>Key Insight:</strong>  
<strong>Only non-conservative forces can change the mechanical energy of a particle.</strong>  
Conservative forces merely transfer energy between kinetic and potential forms.
</div>

---

### Special Case — Mechanical Energy Conservation

If no non-conservative work is done on the system ($w_{nc} = 0$), then

$$
\Delta e_{\text{mech}} = 0,
$$

which means

$$
e_{\text{mech}} = k + u = \text{constant}.
$$

This is the **principle of conservation of mechanical energy**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## Working Equation (Near Earth’s Surface)

For motion near Earth’s surface involving gravity and possibly friction:

- $k = \dfrac{1}{2}mv^2$  
- $u = mgy$  

The mechanical energy equation becomes

$$
\frac{1}{2}mv_i^2 + mg y_i
+ w_{nc} = \frac{1}{2}mv_f^2 + mg y_f.
$$

If friction is the only non-conservative force acting along the direction of motion,

$$
w_{nc} = f_x\,\Delta x,
$$

and the full working equation is

$$
\frac{1}{2}mv_i^2 + mg y_i
+ f_x\,\Delta x = \frac{1}{2}mv_f^2 + mg y_f.
$$

---

### Interpretation

- Conservative forces move energy between $k$ and $u$.  
- Non-conservative forces transfer energy into or out of mechanical energy.  
- Mechanical energy conservation is not a new law—it is a **rearranged form of the work–energy theorem**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Mechanical Energy with Air Drag (Non-Conservative Force)</h3>

<strong>Problem</strong>  

A $0.50\ \text{kg}$ ball is thrown straight upward with an initial speed of  

$$
v_i = 10.0\ \text{m/s}.
$$

As the ball rises, it experiences:
- the force of gravity, and  
- a constant air-drag force of magnitude  

  $$
  f_{\text{drag}} = 1.5\ \text{N},
  $$

  directed downward.

The ball rises a vertical distance of  

$$
\Delta y = +3.0\ \text{m}.
$$

Using mechanical energy, determine the speed of the ball after it has risen $3.0\ \text{m}$.

---

<strong>Methodology</strong>  

We use the mechanical-energy form of the work–energy theorem:

$$
\Delta e_{\text{mech}} = w_{nc}.
$$

Gravity is a **conservative force** and is accounted for through gravitational potential energy.  
Air drag is **non-conservative** and contributes through its work.

---

<strong>Solution</strong>  

Initial mechanical energy:

$$
e_{\text{mech},i} = \frac{1}{2}mv_i^2 + mg y_i.
$$

Choose $y_i = 0$, so

$$
e_{\text{mech},i} = \frac{1}{2}(0.50)(10.0)^2 = 25\ \text{J}.
$$

Final mechanical energy:

$$
e_{\text{mech},f} = \frac{1}{2}mv_f^2 + mg y_f
= \frac{1}{2}(0.50)v_f^2 + (0.50)(9.8)(3.0).
$$

The work done by air drag is

$$
w_{\text{drag}} = f_{\text{drag}}\Delta y = (-1.5)(3.0) = -4.5\ \text{J}.
$$

Applying the work–energy theorem:

$$
e_{\text{mech},f} - e_{\text{mech},i} = w_{\text{drag}}.
$$

Substitute values:

$$
\left[\frac{1}{2}(0.50)v_f^2 + 14.7\right] - 25 = -4.5.
$$

Simplifying:

$$
\frac{1}{4}v_f^2 = 5.8
\quad \Rightarrow \quad
v_f^2 = 23.2.
$$

---

<strong>Answer</strong>  

$$
\boxed{v_f \approx 4.8\ \text{m/s}}
$$

---

<strong>Interpretation</strong>  

- Gravity transfers energy between kinetic and potential forms but does not change the total mechanical energy.  
- Air drag removes mechanical energy from the system.  
- Because of air drag, the ball has less kinetic energy than it would in the absence of non-conservative forces.

This example illustrates that **only non-conservative forces, such as air drag, can change mechanical energy**.

</div>

**Initial mechanical energy**

The ball is released from rest, so

$$
e_{\text{mech},i} = k_i + u_i = 0 + mg y_i.
$$

Substituting values:

$$
e_{\text{mech},i} = (0.60)(9.8)(4.0) = 23.5\ \text{J}.
$$

---

**Final mechanical energy**

Just before impact, $y_f = 0$, so

$$
e_{\text{mech},f} = \frac{1}{2}mv_f^2.
$$

---

**Apply conservation of mechanical energy**

Since no non-conservative forces act,

$$
e_{\text{mech},f} = e_{\text{mech},i}.
$$

Thus,

$$
\frac{1}{2}mv_f^2 = 23.5.
$$

Solving:

$$
v_f = \sqrt{\frac{2(23.5)}{0.60}} \approx 8.9\ \text{m/s}.
$$

---

**Interpretation**

- Mechanical energy is conserved because no non-conservative forces act.  
- Gravitational potential energy is fully converted into kinetic energy.  
- The result is independent of the chosen reference height.

<h3 style="margin-top:0; color:#000000;">Box Activity 2 – Mechanical Energy with a Non-Conservative Push</h3>

**Physical Setup**

A box of mass

m = 4.0\ \text{kg}

A constant horizontal **pushing force** of magnitude

f_{\text{push}} = +12\ \text{N}

**Task**

3. Use the mechanical-energy form of the work–energy theorem,

\Delta e_{\text{mech}} = w_{nc},
$$

to determine the final speed of the box.  
4. State explicitly why mechanical energy is **not conserved** in this situation.

**Reflection**

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Work done by the push**

$$
w_{\text{push}} = f_{\text{push}}\Delta x = (12)(5.0) = 60\ \text{J}.
$$

---

**Initial mechanical energy**

The box starts from rest and does not change height:

$$
e_{\text{mech},i} = k_i + u_i = 0.
$$

---

**Final mechanical energy**

$$
e_{\text{mech},f} = \frac{1}{2}mv_f^2.
$$

---

**Apply the work–energy theorem**

$$
\Delta e_{\text{mech}} = e_{\text{mech},f} - e_{\text{mech},i} = w_{nc}.
$$

Thus,

$$
\frac{1}{2}mv_f^2 = 60.
$$

Solving,

$$
v_f = \sqrt{\frac{2(60)}{4.0}} \approx 5.5\ \text{m/s}.
$$

---

**Interpretation**

- The pushing force adds energy to the system.  
- Mechanical energy increases because **non-conservative work is done**.  
- Mechanical energy is therefore **not conserved**, even though momentum may change smoothly.
