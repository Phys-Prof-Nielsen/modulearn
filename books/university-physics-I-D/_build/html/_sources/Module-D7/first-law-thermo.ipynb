# D7.5 The First Law of Thermodynamics
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.5.1 Why the First Law matters

In earlier modules, we introduced **energy** in a mechanical context. We learned to track changes in kinetic energy and relate them to **work** done by forces.

In thermodynamics, we broaden this view. Systems can exchange energy not only through mechanical work, but also through **heat transfer**, and they can store energy internally in microscopic forms that are not directly visible as motion.

The **First Law of Thermodynamics** provides a single, unified statement of **energy conservation** that applies to *all* physical systems—mechanical, thermal, and beyond.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.5.2 Statement of the First Law

The First Law of Thermodynamics states that the change in a system’s internal energy is equal to the energy transferred into the system as heat minus the energy transferred out of the system as work:

$$
\Delta U = Q - W.
$$

This equation is not a special thermodynamic rule—it is simply **energy conservation**, written in a form that explicitly accounts for **how energy enters and leaves a system**.

Internal energy **$U$ is a state variable**, meaning that $\Delta U$ depends only on the initial and final states of the system, not on the path taken between them.


<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.5.3 Sign conventions for heat and work

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Concept — Sign Conventions for Heat and Work</h3>

In thermodynamics, signs indicate the <strong>direction of energy transfer</strong> relative to the system.

<br>

<strong>Heat (same convention in physics in most disciplines)</strong><br>

- <strong>$Q > 0$</strong>: heat flows <strong>into</strong> the system  
- <strong>$Q < 0$</strong>: heat flows <strong>out of</strong> the system  

Heat represents energy transferred due to a <strong>temperature difference</strong> and is always treated consistently across disciplines.

<br>

<strong>Work (physics convention used in this course)</strong><br>

In physics, <strong>work is defined as work done by the system</strong> on its surroundings.  
The sign of the work follows directly from the direction of the force exerted by the system and the displacement of the boundary.

With this definition:
- <strong>$W > 0$</strong> corresponds to the system doing work on its surroundings (e.g., expansion).  
- <strong>$W < 0$</strong> corresponds to the surroundings doing work on the system (e.g., compression).

With this convention, the First Law is written as

$$
\Delta U = Q - W.
$$

<br>

<strong>Note on chemistry conventions</strong><br>

In chemistry, work is often defined with the <strong>opposite sign</strong>, by taking work done <strong>on</strong> the system as positive.  
With that bookkeeping choice, the First Law is written as

$$
\Delta U = Q + W.
$$

The underlying physics is the same; only the <strong>sign convention for work</strong> differs.

In this course, we will consistently use the <strong>physics definition: work done by the system</strong>.
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.5.4 The First Law as the true energy equation

In mechanics, we often write an “energy equation” such as

$$
\Delta K = W,
$$

or, when potential energy is included,

$$
\Delta K + \Delta U_{\text{mech}} = 0,
$$

for isolated systems.

These equations are **special cases** of energy conservation that apply when:
- energy transfer occurs only through mechanical work,
- and internal thermal effects are negligible.

The First Law makes no such restrictions.

It states that **all** changes in a system’s energy—mechanical or thermal—must be accounted for through **heat** and **work**:

$$
\boxed{\Delta U = Q - W}
$$

Here, the internal energy $U$ includes:
- kinetic energy of microscopic motion,
- potential energy of microscopic interactions,
- and, when relevant, macroscopic mechanical energy.

From this perspective:
- The mechanical energy equations are **limited forms** of the First Law.
- The First Law is the **general energy accounting principle**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.5.5 Interpreting the First Law qualitatively

The First Law allows us to reason about processes without detailed calculations:

- If a system absorbs heat and does no work, its internal energy increases.  
- If a system does work without absorbing heat, its internal energy decreases.  
- If heat enters and work is done simultaneously, the net change depends on the balance between the two.

The power of the First Law lies not in computation, but in **consistent bookkeeping** of energy flow.

---

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Internal Energy Is a State Variable</h3>

A closed system starts in an initial state A and ends in a final state B.  
The system undergoes the same net change in state in two different ways.

<br>

<strong>Process 1</strong><br>

- Heat absorbed by the system:  

  $$ Q_1 = +800\,\text{J} $$

- Work done by the system:  

  $$ W_1 = +300\,\text{J} $$

Using the First Law,

$$
\Delta U_1 = Q_1 - W_1 = 800 - 300 = 500\,\text{J}.
$$

<br>

<strong>Process 2</strong><br>

- Heat absorbed by the system:  

  $$ Q_2 = +1200\,\text{J} $$

- Work done by the system:  

  $$ W_2 = +700\,\text{J} $$

Again applying the First Law,

$$
\Delta U_2 = Q_2 - W_2 = 1200 - 700 = 500\,\text{J}.
$$

<br>

<strong>Interpretation</strong><br>

Even though the **amount of heat absorbed** and the **amount of work done** are different for the two processes, the change in internal energy is the same:

$$
\Delta U_1 = \Delta U_2 = 500\,\text{J}.
$$

This shows that internal energy depends only on the **initial and final states**, not on the path taken between them.

<br>

<strong>Key takeaway</strong><br>

Internal energy $U$ is a <strong>state variable</strong>.  
Heat $Q$ and work $W$ are <strong>process variables</strong>: they depend on *how* the change occurs, but their combination $Q - W$ does not.
</div>

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Friction, Newton’s Second Law, and Internal Energy</h3>

A block of mass

$$
m = 2.0\,\text{kg}
$$

slides across a rough horizontal floor with an initial speed

$$
v_0 = 6.0\,\text{m/s}.
$$

The kinetic friction force acting on the block has magnitude

$$
f_k = 4.0\,\text{N}.
$$

The block eventually comes to rest after sliding some distance.

<br>

<strong>Mechanical description (Newton’s laws)</strong><br>

The friction force acts opposite the direction of motion, producing a constant acceleration

$$
a = -\frac{f_k}{m} = -\frac{4.0}{2.0} = -2.0\,\text{m/s}^2.
$$

Using kinematics, the stopping distance is

$$
0 = v_0^2 + 2a\Delta x
\quad\Rightarrow\quad
\Delta x = \frac{v_0^2}{2|a|}
= \frac{36}{4}
= 9.0\,\text{m}.
$$

<br>

<strong>Work done by friction</strong><br>

The work done by the friction force is

$$
W_{\text{fric}} = -f_k \Delta x = -(4.0)(9.0) = -36\,\text{J}.
$$

This equals the change in the block’s kinetic energy:

$$
\Delta K = K_f - K_i = 0 - \frac{1}{2}mv_0^2 = -36\,\text{J}.
$$

From a purely mechanical viewpoint, the kinetic energy is “lost.”

<br>

<strong>Thermodynamic interpretation</strong><br>

The energy is not destroyed.  
The friction force converts mechanical energy into **internal energy** of the block and the floor (microscopic motion of atoms).

Applying the First Law to the block–floor system:

- No heat enters or leaves the system:  

  $$ Q = 0 $$

- The system does no macroscopic work on the surroundings:  

  $$ W = 0 $$

So the First Law gives

$$
\Delta U = Q - W = 0.
$$

However, the internal energy of the system increases by

$$
\Delta U_{\text{thermal}} = +36\,\text{J},
$$

exactly equal to the kinetic energy lost.

<br>

<strong>Key takeaway</strong><br>

Friction does not violate energy conservation.  
It transfers energy from **macroscopic motion** into **internal energy**.

Newton’s laws describe the motion, but the **First Law of Thermodynamics completes the energy accounting**.
</div>

<h3 style="margin-top:0; color:#000000;">Box Activity 1 — Friction, Work, and Temperature Increase</h3>

In this activity, you will connect **Newton’s Second Law**, **mechanical work**, and the **First Law of Thermodynamics** to determine how mechanical energy becomes internal energy.

<strong>Task:</strong>

A solid metal block slides across a rough horizontal surface.

The block has:
- mass  

  $$ m = 3.0\,\text{kg} $$

- initial speed  

  $$ v_0 = 5.0\,\text{m/s} $$

A constant frictional force of magnitude

$$
f = 6.0\,\text{N}
$$

acts on the block opposite the direction of motion.

Assume:
- specific heat capacity of the block $ c = 450 $ J/(kg$\cdot$ K)

Assume **all mechanical energy lost goes into the internal energy of the block**.

Answer the following questions.

1. Use Newton’s Second Law to determine the **acceleration** of the block.  
2. Determine the **distance** the block slides before coming to rest.  
3. Calculate the **work done by friction**.  
4. Use the **First Law of Thermodynamics** to determine the change in internal energy of the block.  
5. Determine the **increase in temperature** of the block.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide
</summary>

<strong>Step 1: Acceleration (Newton’s Second Law)</strong>

The friction force is the only horizontal force, so

$$
a = -\frac{f}{m} = -\frac{6.0}{3.0}
= -2.0\,\text{m/s}^2.
$$

<br>

<strong>Step 2: Stopping distance</strong>

Using kinematics,

$$
0 = v_0^2 + 2a\Delta x
\quad\Rightarrow\quad
\Delta x = -\frac{v_0^2}{2a}
= \frac{25}{4}
= 6.25\,\text{m}.
$$

<br>

<strong>Step 3: Work done by friction</strong>

$$
W_{\text{fric}} = f\,\Delta x
= (-6.0)(6.25)
= -37.5\,\text{J}.
$$

<br>

<strong>Step 4: First Law of Thermodynamics</strong>

For the block:
- No heat enters or leaves the "system" during the slide:  

  $$ Q = 0 $$

- The work done by the block:  

  $$ W = -37.5\,\text{J} $$

Applying the First Law,

$$
\Delta U = Q - W = 0 - (-37.5) = +37.5\,\text{J}.
$$

The internal energy of the block increases by 37.5 J.

<br>

<strong>Step 5: Temperature increase</strong>

Using

$$
\Delta U = mc\Delta T,
$$

$$
\Delta T = \frac{\Delta U}{mc}
= \frac{37.5}{(3.0)(450)}
\approx 0.028\,\text{K}.
$$

<br>

<strong>Interpretation</strong>

The temperature increase is small, but the energy accounting is exact:  
every joule of mechanical energy lost becomes internal energy.

This demonstrates how **Newton’s laws**, **work**, and the **First Law of Thermodynamics** describe the same physical process from different perspectives.
</details>

## D7.5.6 Key takeaway

The First Law of Thermodynamics is not a new kind of physics.  
It is **energy conservation**, written in its most complete and general form.

All other energy equations—mechanical or thermal—are special cases of this law.
