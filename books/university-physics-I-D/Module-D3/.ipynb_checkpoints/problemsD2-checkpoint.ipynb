<h2>Problem D2.1</h2>
    </header>

Calculate the linear momentum of the following:

1. A proton moving with in the +x-direction with a speed that is 99.9% the speed of light.
2. A 2000.0 kg sedan traveling with a velocity of -70.0 mph.

### Solution

```python
#DIY Cell
```

```python
%reset -f

# PART 1
#-------

c = 2.9979e8
m = 1.6726e-27

v = 0.999*c

p = m*v
answer = "{:+.2e}".format(p)
print('Linear momentum of proton: '+answer+' kgm/s')

# PART 2
#-------

m = 2000.0
v = -70.0  #mi/hr

cf1 = (1609.34)/(1)  # 1609.34 m per 1 mi
cf2 = (1)/(3600)  # 1 hr per 3600 s

v = v*cf1*cf2

p = m*v
answer = "{:.2e}".format(p)
print('Linear momentum of sedan: '+answer+' kgm/s')
```

<h2>Problem D2.2</h2>
    </header>

A semi truck with a mass of 1.20E4 kg and speed of 65 mph is northbound on I-15. At the same time, a sedan with a mass of 2.40E3 kg is southbound on I-15 with a speed of 85 mph. What are the individual momentums and the magnitudes of momentums of the two vehicles? Consider the northbound direction to be positive.
   
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

### Solution

```python
#DIY Cell
```

```python
%reset -f

import numpy as np  #we will use the fucntion "abs" to find the magnitude of momentum

cf1 = (1609.34)/(1)  # 1609.34 m per 1 mi
cf2 = (1)/(3600)  # 1 hr per 3600 s

#semi-truck
#-------
m = 1.20e4
v = 65*cf1*cf2

p = m*v
answer = "{:+.1e}".format(p)
print('Linear momentum of semitruck: '+answer+' kgm/s')

p_mag = np.abs(p)
answer = "{:.1e}".format(p_mag)
print('Magnitude of linear momentum of semitruck: '+answer+' kgm/s')

print(' ') #provide a space between the answers of semi and sedan
#sedan
#-------
m = 2.40e3
v = -85*cf1*cf2  #negative since it is velocity and moving in negative direction

p = m*v
answer = "{:.1e}".format(p)
print('Linear momentum of sedan: '+answer+' kgm/s')

p_mag = np.abs(p)
answer = "{:.1e}".format(p_mag)
print('Magnitude of linear momentum of sedan: '+answer+' kgm/s')
```

## Problem D2.3 üêçüíª

In a particle accelerator, an electron of mass $m_e = 9.11 \times 10^{-31}\ \text{kg}$ is accelerated along a straight line. Its velocity increases linearly with time according to the relation

$$ v_x(t) = \frac{v_{f_x}}{t_f}t $$

where $t_f = 2.0 \times 10^{-9}\ \text{s}$ is the total acceleration time, and $v_f$ is the final velocity at $t = t_f$.

At the end of the acceleration period, the electron‚Äôs momentum is measured to be $p_x = 1.82E-22~\text{kgm/s}$.

1. Use Sympy‚Äôs solve function to find the final velocity $v_f$.

2. Plot the velocity vs. time curve from $t=0$ to $t_f$.

```python
# DIY Cell
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Known values
m_e = 9.11e-31    # kg, electron mass
p = 1.82e-22      # kg*m/s, measured momentum
t_f = 2.0e-9      # s, acceleration time

# Unknown final velocity
v_f = sp.Symbol('v_f')

# Momentum equation at final time (alternative way of writing it compared to example in notes)
equation = sp.Eq(m_e * v_f, p)

# Solve for v_f
solution = sp.solve(equation, v_f)
print("Final electron velocity v_f =", solution[0], "m/s")

# ---- Plotting velocity vs time ----
t = np.linspace(0, t_f, 100)
v = (solution[0]/t_f) * t   # linear increase from 0 to v_f

plt.plot(t*1e9, v)   # convert time to ns for readability
plt.xlabel("Time (ns)")
plt.ylabel("Velocity (m/s)")
plt.title("Electron acceleration in a particle accelerator")
plt.grid(True)
plt.show()
```
