## Why this module matters

Many physical processes involve changes that cannot be fully understood using mechanics alone. Objects warm up, cool down, expand, contract, and exchange energy with their surroundings in ways that depend on **temperature** and **thermal interactions**, not just visible forces and motion.

However, these thermal processes do **not** lie outside the laws of physics already introduced. At the microscopic level, thermodynamic behavior arises from the same fundamental principles—**impulse, work, energy, and momentum**—acting between vast numbers of atoms and molecules. Each microscopic interaction obeys familiar physical laws; what changes is the **scale** and **collective complexity** of the system.

In this module, we therefore adopt a **macroscopic, phenomenological description** of thermodynamics. Rather than tracking individual particles, we describe systems using observable quantities such as **temperature**, **heat**, **work**, and **internal energy**. These quantities allow us to account for energy transfer and storage without resolving the detailed microscopic motion that gives rise to them.

We begin by introducing **temperature** as a measurable property associated with thermal equilibrium, followed by **thermal expansion**, which reveals how microscopic motion manifests as macroscopic size changes. We then define **heat** as a mode of energy transfer and examine the **mechanisms of heat transfer**. Finally, we introduce the **First Law of Thermodynamics** as a unifying principle that connects mechanical work and heat through changes in internal energy.

In later phases of the course, we will return to these ideas in greater depth. There, we will connect macroscopic thermodynamic laws to their microscopic origins and introduce elements of **statistical mechanics**, where probability, large numbers of particles, and—at times—quantum effects become essential. For now, the focus is on building a **coherent, energy-based framework** that remains fully consistent with fundamental physics while remaining accessible at the macroscopic level.

This module establishes the conceptual foundation needed for deeper thermodynamic analysis in subsequent phases.

By the end of this module, you will be able to:

---

### **A. Temperature and Thermal Equilibrium**
*Introduces temperature as a measurable physical property.*

- **A1.** Define **temperature** as a property associated with thermal equilibrium between systems.  
- **A2.** Explain how temperature is measured using thermometers and why temperature comparisons require a reference system.  
- **A3.** Distinguish between **temperature** and **thermal energy**, recognizing that objects at the same temperature may contain different amounts of internal energy.

---

### **B. Thermal Expansion**
*Describes how temperature changes affect physical dimensions.*

- **B1.** Describe **thermal expansion** as the tendency of materials to change size when temperature changes.  
- **B2.** Identify **linear**, **area**, and **volume** expansion as different manifestations of the same underlying effect.  
- **B3.** Interpret simple expansion relations qualitatively, recognizing the role of material properties and temperature change.  
- **B4.** Apply thermal expansion concepts to real-world situations involving constraints, such as rails, bridges, or composite materials.

---

### **C. Heat as Energy Transfer**
*Clarifies the role of heat in thermodynamic processes.*

- **C1.** Define **heat** as energy transferred between systems due to a temperature difference.  
- **C2.** Distinguish between **heat**, **work**, and **internal energy**, recognizing that heat is not a property stored in a system.  
- **C3.** Interpret the sign of heat transfer using clear physical reasoning (energy entering or leaving a system).

---

### **D. The First Law of Thermodynamics**
*Unifies work and heat through energy conservation.*

- **D1.** Define **internal energy** as the energy stored within a system due to microscopic motion and interactions.  
- **D2.** State the **First Law of Thermodynamics** in words and in symbolic form,

  $$
  \Delta U = Q - W,
  $$

  and explain the physical meaning of each term.

- **D3.** Apply the First Law qualitatively to simple processes, such as heating at fixed volume, compression, or frictional dissipation.  
- **D4.** Explain how mechanical work and heat are alternative mechanisms for changing a system’s internal energy.

---

### **E. Heat Transfer Mechanisms**
*Identifies how thermal energy moves between systems.*

- **E1.** Describe **conduction** as heat transfer through direct contact at the microscopic level.  
- **E2.** Describe **convection** as heat transfer involving bulk motion of a fluid.  
- **E3.** Describe **radiation** as heat transfer via electromagnetic waves, requiring no material medium.  
- **E4.** Identify everyday examples where multiple heat transfer mechanisms act simultaneously.

---

### **F. Scope and Preparation for Later Phases**
*Clarifies the level and limits of this introduction.*

- **F1.** Recognize that thermodynamic relations in this module are **descriptive and phenomenological**, not derived from microscopic models.  
- **F2.** Identify concepts intentionally deferred to later phases, such as equations of state, entropy, and thermodynamic cycles.  
- **F3.** Describe this module as a **conceptual bridge** between mechanical energy and full thermodynamic analysis.

---

### **Performance Criteria**

- Temperature, heat, work, and internal energy are defined **clearly and distinctly**.  
- Thermal processes are explained using **energy flow and storage**, not forces or motion alone.  
- The First Law is used as a **conceptual accounting principle**, not a formula-matching exercise.  
- Heat transfer mechanisms are identified correctly in real-world contexts.  
- Explanations emphasize **physical meaning, directionality, and system boundaries**, rather than computation.
