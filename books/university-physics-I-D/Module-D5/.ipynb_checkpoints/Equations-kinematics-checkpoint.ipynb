## D7.2.1 Average definitions

In the previous note we defined the average velocity and average acceleration. Let us summarize the equations here:

<div class="alert alert-success">
    
$$ v_{av,x} = \frac{\Delta x}{\Delta t} \tag{1}$$
</div>

<div class="alert alert-success">
    
$$ a_{av,x} = \frac{\Delta v_x}{\Delta t} \tag{2}$$
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.2.2 Position with constant velocity

We can use the definition of average velocity (since an average velocity is a **constant** velocity) to investigate the position of a particle as a function of time. Rewriting Equation (1) from above:

$$ \Delta x= v_{av,x}\Delta t$$

If we write out the differences and re-arrange the terms, we get the postion at **any** time $t_2$:

$$x_2 = x_1 + v_{av,x}(t_2 - t_1) $$
    
We observe that to determine the position at **any** time $t_1$, we **must** know the initial position and time as well as the velocity. We call this **an initial condition**. Often the initial conditions are at $t_1 = 0$ and the above is then written as 

<div class="alert alert-success">
    
$$ x(t) = x_0 + v_{av,x}t \tag{3}$$
</div>

where we have made clear that the position is a function of time.


<div class="alert alert-danger">

The above relation is **only** valid for cases with a **constant** velocity. If you apply the above equation to a problem with non-constant velocity, you will get a wrong answer. Plain out wrong!
</div>

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 1</h3>
    </header>
    
Consider a cosmic ray traveling through the Earth's atmospheric with an average speed of $3.0\times 10^6$ m/s. It is initially located 10.0 km above the Earth's surface. How far above the surface is it at t = 1.0 ms?

**Solution**
    
There are several ways of interpretating the numbers for this problem as we are free to choose our reference system. I will choose the following:

My reference frame is a 1D reference frame in the z-direction with positive being upward (I know, I am so boring) and the origin at the surface. With this reference frame, we have

$z_0 = 10.0~\textrm{km} = 1.00\times 10^4~\textrm{m}$,  
$v_{av,z} = -3.0\times 10^6~\textrm{m/s}$: since the particle is moving in the negative direction,<br> 
$t = 1.0\times 10^{-3}$ s<br>.

Using Equation (3) from above but for the z-direction, we have

$$ z(t) = z_0 + v_{av,z}t $$

Inserting our values, we get the new position above the surface:

$$ z(t = 1.0~\textrm{ms}) = (1.00\times 10^4~\textrm{m}) + (-3.0\times 10^6~\textrm{m/s})~(1.0\times 10^{-3}~\textrm{s}) $$

$$ z(t = 1.0~\textrm{ms}) = (1.00\times 10^4~\textrm{m}) + (-3.0\times 10^3~\textrm{m}) $$

$$ z(t = 1.0~\textrm{ms}) = 7.0\times 10^3~\textrm{m} $$

The particle is located 7.0 km above the surface after 1 ms has elapsed from its initial position at t = 0 s.
    
</div>

## D7.2.4  Velocity with constant acceleration

We can follow the same approach as above to find a function for the velocity, provided we have information about the initial velocity, and a constant acceleration. Without further ado:

<div class="alert alert-success">
    
$$ v_{x}(t) = v_{0,x} + a_{av,x}t \tag{4}$$
</div>

<div class="alert alert-danger">

The above relation is **only** valid for cases with a **constant** acceleration. If you apply the above equation to a problem with non-constant acceleration, you will get a wrong answer. Plain out wrong!

## D7.2.4 Derived Equations: constant acceleration

Two derived equations are the most essential to kinematics and are **only valid for constant acceleration**. We will derive these in Phase B using calculus. 

<div class="alert alert-success">
    
$$ x(t)  = x_0 +  v_{0,x}t +  \frac{1}{2}a_{x}t^2 \tag{5}$$

and 

$$ v_{x}^2 = v_{0,x}^2 + 2a_{x}\Delta x \tag{6}$$
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 3</h3>
    </header>

A rocket is launched vertically with an initial speed of 0.0 m/s. The rocket accelerates at a net rate of 50.0-g's (1 g = 9.81 m/s$^2$). How high is the rocket above its launch point at t = 2.0 s?

<b>Solution</b><br>

Let the launch point be $z_0 = 0$ m, and insert our given numbers into equation (5) and the z-direction instead:

$$ z(t = 2)  = (0) +  (0)(2.0) +  \frac{1}{2}(50.0)(9.81)(2.0)^2 = 980~\textrm{m} $$
