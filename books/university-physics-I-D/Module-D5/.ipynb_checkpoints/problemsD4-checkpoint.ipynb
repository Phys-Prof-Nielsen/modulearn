<h2>Problem D4.1</h2>
    </header>

Before a crash-test collision, a car with a crash test dummy of mass 75-kg is traveling at 25.0 m/s towards a wall. The impact time is 10.00 ms. Assume the impact force is much greater than any other forces involved in the collision. 

1. If the crast test dummy is fasten to the car, what impact force is experienced by the dummy?
2. Instead of being fasten to the car, the dummy is wearing seat belts and the impact time is increased to 20.0 ms. What is the impact force then?
3. In addition to seat belt, the car is equipped with airbags increasing the impact time to a whooping 0.100 s. What is the impact force in this scenario?
4. The acceleration of gravity is 9.81 m/s$^2$. How many g's did the dummy experience in each crash scenario? HINT: take the acceleration and divide by 9.81 to obtain it in units of g.

```python
# DIY Cell
```

```python
%reset -f

# declare given parameters
m = 75.0
#let +x be in initial direction of travel
v1x = 25.0
v2x = 0.0  # crash test dummy comes to a stop
dt = 10.0E-3   # time in seconds

# part 1
# ------
p1x = m*v1x
p2x = m*v2x
fx = (p2x - p1x)/(dt)

answer = "{:.1e}".format(fx)
print('Impact force is '+answer+' N.')

# Part 2
# ------
dt = 20.0E-3
fx = (p2x - p1x)/(dt)

answer = "{:.1e}".format(fx)
print('Impact force is '+answer+' N.')

# Part 3
# ------
dt = 0.100
fx = (p2x - p1x)/(dt)

answer = "{:.1e}".format(fx)
print('Impact force is '+answer+' N.')

print(' ')

# Part 4
# ------
g = 9.81
dt = 10.0E-3
ax = (v2x - v1x)/dt
g_acc = ax/g

answer = "{:.2e}".format(g_acc)
print('Scenario 1: '+answer+' g.')

dt = 20.0E-3
ax = (v2x - v1x)/dt
g_acc = ax/g

answer = "{:.2e}".format(g_acc)
print('Scenario 2: '+answer+' g.')

dt = 0.100
ax = (v2x - v1x)/dt
g_acc = ax/g

answer = "{:.2e}".format(g_acc)
print('Scenario 2: '+answer+' g.')
```

<h2>Problem D4.2</h2>
    </header>

A 5.0 kg particle is traveling in 1D at -35 m/s when it interacts with another object. After the interaction the velocity of the particle is -5.0 m/s.

1. Make a prediction about the direction of the interaction in terms (positive or negative direction) and explain your answer.
2. If the net average force during the collision was 3.0E3 N, what was the impact time?
3. What was the acceleration of the of the particle during the interaction? 

```python
# DIY Cell
```

```python
%reset -f

# declare given parameters
m = 5.0
v1x = -35.0
v2x = -5.0

# Part 1
# ------
print('Since the particle is slowing down, the interaction must be in opposite direction of motion.')

# Part 2
# ------
fx = 3.0E3
p1x = m*v1x
p2x = m*v2x
dpx = p2x - p1x
dt = dpx/fx

answer = "{:.1e}".format(dt)
print('The impact time was: '+answer+' s.')

# Part 3
# ------
ax = (v2x - v1x)/dt   #could also use f = ma to find acceleration.

answer = "{:.1e}".format(ax)
print('Acceleration was: '+answer+' m/s\u00B2.')
```

<h2>Problem D4.3</h2>
    </header>

A net force of $f_x = +5.0$ N is in contact with a 0.50 kg particle over the duration of 8.0 s. 

1. Is momentum transferred to or from the particle? Explain your answer.
2. What is the momentum transfer experienced by the particle?
3. If the initial velocity of the particle was +25 m/s, what is the velocity of the particle after the interaction?
4. What acceleration did the particle experience during the interaction?

```python
# DIY Cell
```

```python
%reset -f
# declare given parameters
fx = 5.0
m = 0.50
dt = 8.0

# Part 1
# ------
print('Since force is positive, momentum is transferred to the particle.')

# Part 2
# ------
dpx = fx*dt

print('The momentum transfer is: '+str(dpx)+' kgm/s.')

# Part 3
# ------
v1x = 25.0
p1x = m*v1x

p2x = p1x + dpx
v2x = p2x/m

print('Final velocity is: '+str(v2x)+' m/s.')

# Part 4
# ------
ax = (v2x - v1x)/dt

print('Acceleration is: '+str(ax)+' m/s\u00B2.')
```

<h2>Problem D4.4</h2>
    </header>

Andrea, a $63.0$ kg sprinter, starts a race with an acceleration of $+4.200$ m/s$^2$. What is the net external force on her?

[This problem is a slightly modified version from OpenStax. Access for free here](https://openstax.org/books/university-physics-volume-1/pages/5-problems)

```python
# DIY Cell
```

```python
%reset -f

import sympy as sym

m = 63.0
ax = 4.200

#declare unknown variable
fx = sym.Symbol('fx')

#equation to solve: 0 = f - ma (comes from f = ma)
eq = fx - m*ax

sol = sym.solve(eq,fx)
fx = sol[0]

answer = "{:.2e}".format(fx)
print('The force on Andrea is: '+str(answer)+' N.')
```

<h2>Problem D4.5</h2>
    </header>

A cleaner pushes a $4.50$ kg laundry cart in such a way that the net external force on it is $60.0$ N. Calculate the magnitude of his cart‚Äôs acceleration.

[This problem is a slightly modified version from OpenStax. Access for free here](https://openstax.org/books/university-physics-volume-1/pages/5-problems)

```python
# DIY Cell
```

```python
%reset -f

import sympy as sym

m = 4.50
fx = 60.0

#declare unknown variable
ax = sym.Symbol('ax')

#equation to solve: 0 = f - ma (comes from f = ma)
eq = fx - m*ax

sol = sym.solve(eq,ax)
ax = sol[0]
a = sym.Abs(ax)

answer = "{:.2e}".format(a)
print('The magnitude of the acceleration is: '+str(answer)+' m/s\u00B2.')
```

<h2>Problem D4.6</h2>
    </header>

A car with a mass of $1000.0$ kg accelerates from $0$ to $+90.0$ km/h in $10.0$ s. 

1. What is its acceleration?
2. What is the net force on the car?

[This problem is a slightly modified version from OpenStax. Access for free here](https://openstax.org/books/university-physics-volume-1/pages/5-problems)

```python
# DIY Cell
```

```python
%reset -f

import sympy as sym

m = 1000.0
v1x = 0.0
v2x = 90.0
dt = 10.0

#convert v2x from km/hr to m/s
v2x = (90.0)*(1000./1.0)*(1.0/3600.0)

# Part 1
# -----
#declare unknown variable
ax = sym.Symbol('ax')

#equation to solve: 0 = a - (v2 - v1)/(t2 - t1) 
eq1 = ax - (v2x - v1x)/dt

sol1 = sym.solve(eq1,ax)
ax = sol1[0]

answer = "{:.2e}".format(ax)
print('The acceleration is: '+str(answer)+' m/s\u00B2.')

# Part 2
# -----
#declare unknown variable
fx = sym.Symbol('fx')

#equation to solve
eq2 = fx - m*ax

sol2 = sym.solve(eq2,fx)
fx = sol2[0]

answer = "{:.2e}".format(fx)
print('The force is: '+str(answer)+' N.')
```

## Problem D4.7 üêçüíª Impulse‚ÄìMomentum for a Muon in a Beamline

In a beamline, a **muon** (mass $m_\mu = 1.884\times10^{-28}\ \text{kg}$) moves along the $x$-axis with initial speed $v_0 = 5.00\times10^{5}\ \text{m/s}$. A fast kicker applies a **time-varying force** $f(t)$ along $+x$ for a total of $200\ \text{ns}$ with the shape:

- **Ramp up** linearly from $0$ to $f_0$ during $0\text{‚Äì}40\ \text{ns}$  
- **Hold** constant at $f_0$ during $40\text{‚Äì}140\ \text{ns}$  
- **Ramp down** linearly from $f_0$ to $0$ during $140\text{‚Äì}200\ \text{ns}$  

After the kicker, the muon‚Äôs momentum is measured to be  
$$ p_f = 2.50\times10^{-22}\ \text{kg¬∑m/s}. $$

Use the **impulse‚Äìmomentum theorem for a single particle**: 

$$ i_{net_x} = \Delta p = f_{net_x}\Delta t$$

**Tasks**  
1. Using geometry only, express the **total area** under $f(t)$ (the impulse) in terms of the unknown peak force $f_0$ and the given time segments.
2. Set $i_{net_x} = \Delta p = p_f - p_0$ and use **Sympy‚Äôs `solve`** to find the **unknown peak force $f_0$**.  
3. Create a **plot of $f(t)$ vs. time** (in ns) that shows the ramp-up, flat, and ramp-down segments clearly.  
4. Using Python (no calculus), compute the **numerical area** under your plotted $f(t)$ curve (e.g., trapezoid-sum over the sampled data) and verify it matches the impulse you used to solve for $f_0$.  
5. Compute the **final speed** $v_f$ from $p_f = m_\mu\,v_f$ and report it (assume non-relativistic for this exercise).  

*Hints:*  
- Triangle area: $\tfrac{1}{2}\times\text{base}\times\text{height}$  
- Rectangle area: $\text{base}\times\text{height}$  
- Keep all times in **seconds** in your code; label the plot in **ns** for readability.


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# --------------------------
# Given constants (SI units)
# --------------------------
m_mu = 1.884e-28          # kg, muon mass
v0   = 5.00e5             # m/s, initial speed
p_f  = 2.50e-22           # kg*m/s, measured final momentum after the kick

# Time breakpoints (in seconds)
# These define the three regions of the force-time graph:
#   0‚Äì40 ns   : ramp up (triangle area)
#   40‚Äì140 ns : flat top (rectangle area)
#   140‚Äì200 ns: ramp down (triangle area)
t0, t1, t2, t3 = 0.0, 40e-9, 140e-9, 200e-9

# --------------------------
# Step 1: Momentum change
# --------------------------
p0 = m_mu * v0            # initial momentum of the muon before the kicker
Delta_p = p_f - p0        # momentum change required (this must equal the impulse J)

# --------------------------
# Step 2: Symbolic solve for F0
# --------------------------
# Impulse = area under f(t).
# Since the shape is "triangle + rectangle + triangle",
# we can compute it by geometry only (no calculus needed).
#
#   Up-ramp area   = (1/2) * (t1 - t0) * F0
#   Rectangle area = (t2 - t1) * F0
#   Down-ramp area = (1/2) * (t3 - t2) * F0
#
# Combine into one "effective time factor" multiplied by f0:
f0 = sp.Symbol('f0', positive=True) #This symbol must always represent a positive number (strictly > 0).
#Sympy can then simplify expressions more cleanly because it doesn‚Äôt have to consider negative cases.

# try and see if you can figure this one out :)  HINT: has something to do with average and straight lines
effective_dt = 0.5*(t1 - t0) + (t2 - t1) + 0.5*(t3 - t2)
i_symbolic = f0 * effective_dt

# Solve J_symbolic = Œîp for the unknown peak force F0: two equations
f0_sol = sp.solve(sp.Eq(i_symbolic, Delta_p), f0)[0]
print(f"Peak force f0 = {f0_sol:.3e} N")

# --------------------------
# Step 3: Build and plot f(t)
# --------------------------
# Define f(t) piecewise for plotting:
#   - Linear increase 0 ‚Üí f0 during [t0, t1]
#   - Flat f0 during [t1, t2]
#   - Linear decrease f0 ‚Üí 0 during [t2, t3]
def f_of_t(tt, f0_value):
    f = np.zeros_like(tt) #creates a new NumPy array of zeros with the same shape and data type as the array tt
    # ramp up: 0 -> f0 on [t0, t1]
    up = (tt >= t0) & (tt < t1)
    f[up] = f0_value * (tt[up] - t0) / (t1 - t0)
    # flat: constant f0 on [t1, t2]
    flat = (tt >= t1) & (tt < t2)
    f[flat] = f0_value
    # ramp down: f0 -> 0 on [t2, t3]
    down = (tt >= t2) & (tt <= t3)
    f[down] = f0_value * (t3 - tt[down]) / (t3 - t2)
    return f

# Create a fine time array for smooth plotting
t = np.linspace(t0, t3, 1000)
f_vals = f_of_t(t, float(f0_sol))

# Plot the force profile
plt.plot(t*1e9, f_vals, label="Kicker force $f(t)$")
plt.xlabel("Time (ns)")      # label in ns for readability
plt.ylabel("Force (N)")
plt.title("Beam kicker force profile (geometry-only area)")
plt.grid(True)
plt.legend()
plt.show()

# --------------------------
# Step 4: Numerical area check (no calculus)
# --------------------------
# Use trapezoidal integration (sum of trapezoids under curve)
# to approximate the impulse from sampled data.
# This verifies that our geometry formula matches the actual F(t) profile.
i_numeric = np.trapezoid(f_vals, t)
print(f"Required impulse Œîp = {Delta_p:.3e} N¬∑s")
print(f"Area from sampled f(t) = {i_numeric:.3e} N¬∑s")

# --------------------------
# Step 5: Final speed from p_f = m_mu * v_f
# --------------------------
# Finally, solve for v_f using definition of momentum.
v_f = p_f / m_mu
print(f"Final speed v_f = {v_f:.3e} m/s")
```
