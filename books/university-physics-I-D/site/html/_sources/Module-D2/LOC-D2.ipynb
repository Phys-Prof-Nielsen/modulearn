## D2.0.1 Why this module matters
Before we can analyze motion, forces, energy, or fields, we need a shared language for describing **measurements** and **evidence**. Physics is not just equations—it is the art of turning observations into trustworthy numerical claims.  
This module builds the foundation for that art.

Here you will learn how to:
- ask testable questions using the scientific method,  
- express quantities in consistent units,  
- represent numbers clearly using scientific notation and significant figures,  
- analyze repeated measurements statistically, and  
- visualize relationships using clean, meaningful plots.

Mastery of these tools will make all later problem solving smoother, more accurate, and more transparent—to others and to your future self.

By the end of this module, you will be able to:

---

### A. Scientific Method & Experimental Thinking
- **A1.** Formulate a testable physics question and identify independent, dependent, and controlled variables.  
- **A2.** Predict the qualitative trend expected from physical reasoning and outline a minimal data-collection plan.  
- **A3.** Describe how measurement choices influence uncertainty and what evidence would support or refute a hypothesis.

---

### B. Units, Dimensional Consistency & Conversions  
*Moved earlier because units underpin all later numerical representation.*

- **B1.** Convert between units using the **factor–label method**, including multi-step conversions (e.g., mile → m → cm).  
- **B2.** Use SI base and derived units (e.g., N = kg·m·s$^{-2}$) and check equations for dimensional consistency.  
- **B3.** Diagnose and correct inconsistent or missing units in worked solutions.

---

### C. Scientific Notation & Representing Numbers  
*Comes after units so students understand what quantities they are representing.*

- **C1.** Convert numbers to and from **scientific notation** and interpret powers of ten.  
- **C2.** Compare magnitudes and isolate scale differences using exponent rules.  
- **C3.** Evaluate whether a numerical expression conveys appropriate clarity and order of magnitude.

---

### D. Orders of Magnitude & Estimation  
*Builds intuition before precision.*

- **D1.** Perform **Fermi estimates** and justify simplifying assumptions.  
- **D2.** Compare quantities using **dimensionless ratios** and interpret physical meaning (≪1, ~1, ≫1).  
- **D3.** Report an estimate to the correct order of magnitude and state its physical relevance.

---

### E. Measurement Precision & Significant Figures  
*Placed after number representation and estimation.*

- **E1.** Read analog and digital instruments and report a measurement with a justified final digit.  
- **E2.** Apply **significant-figure rules** for addition/subtraction and multiplication/division, including proper rounding.  
- **E3.** Distinguish **accuracy** vs. **precision**, and compute absolute and relative percent error.

---

### F. Basic Statistics for Experimental Data  
*Placed after precision to quantify variation.*

- **F1.** Compute **mean**, **standard deviation**, and **standard error of the mean (SEM)** for repeated measurements.  
- **F2.** Interpret what each quantity measures (spread, uncertainty in the mean, repeatability).  
- **F3.** Identify outliers and discuss their possible causes (instrument error, procedural error, etc.).

---

### G. Graphing & Data Presentation  
*Placed last, as it depends on all earlier skills.*

- **G1.** Create a clear, labeled Cartesian plot with correct units and a meaningful title.  
- **G2.** Determine **slope and intercept** from a roughly linear dataset and attach units and physical interpretation.  
- **G3.** Present results with appropriate precision, consistent units, and a concise conclusion supported by the graph.

---

### Performance Criteria
- Reported results use **consistent units**, **correct significant figures**, and clearly state assumptions.  
- Conversions, statistics, and plots reflect **logical precision**, **dimensional correctness**, and **scientific clarity**.  
- Explanations connect numerical results to physical meaning (e.g., what a slope represents, why a unit cancels, what a ratio implies).
