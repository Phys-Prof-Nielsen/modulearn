# D3.0 Learning Outcomes - Impulse-Momentum Framework
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## Why this module matters

In everyday language, motion is often described in terms of *speed* or *how fast something is going*.  
In physics, motion is understood more deeply through how **moving matter changes** when interactions occur.

This module develops that understanding step by step. We begin by introducing **momentum** as a property of moving matter. We then build toward **impulse** as the process by which momentum is transferred during an interaction. Only after that foundation is established do we introduce **force acting through time** as one specific mechanism for momentum transfer.

Throughout this module, we work strictly in **one dimension**. Direction is handled using **positive and negative signs** relating to a coordinate axis, allowing us to reason carefully about motion, interactions, and conservation without additional mathematical machinery.

By the end of this sequence, momentum conservation will emerge naturally—not as a rule to memorize, but as a logical consequence of how interactions redistribute momentum within a system.

By the end of this module, you will be able to:

---

### A. Momentum as a Property of Moving Matter  
*Establishes momentum as the foundational quantity.*

- **A1.** Describe momentum as a physical property associated with **moving matter**.  
- **A2.** Define one-dimensional momentum as  

  $$
  p_x = m v_x
  $$
  
- **A3.** Interpret the **sign of momentum** as indicating direction of motion and distinguish it from quantities without direction (e.g., mass).

---

### B. Impulse as Momentum Transfer  
*Introduces impulse as the cause of momentum change.*

- **B1.** Define **impulse** as the change in momentum,  

  $$
  j_x = \Delta p_x
  $$
   
- **B2.** Interpret impulse physically as **momentum transferred** during an interaction.  
- **B3.** Explain how multiple interactions can add or remove momentum, leading to a **net impulse**.

---

### C. Force Acting Through Time  
*Connects impulse to forces without introducing calculus.*

- **C1.** Describe force as a quantity that contributes to momentum transfer over a finite time interval.  
- **C2.** Relate impulse to force acting over time using  

  $$
  j_x = f_x\,\Delta t
  $$
   
  for constant or average force.  
- **C3.** Explain how the same impulse can result from different combinations of force magnitude and interaction time.
- **C4.** Interpret a **force–time graph** and use its **area** to reason about impulse as momentum transfer.

---

### D. Impulse–Momentum Theorem  
*Unifies momentum change and force-based impulse.*

- **D1.** Apply the **impulse–momentum theorem**,  

  $$
  \Delta p_{\text{net},x} = f_{\text{net},x}\Delta t
  $$
    
  to one-dimensional situations.  
- **D2.** Interpret the net force as the combined effect of multiple interactions acting over time.  
- **D3.** Use the impulse–momentum theorem to predict changes in motion without relying on acceleration-based reasoning.

---

### E. Conservation of Momentum (Two-Particle Systems)  
*Introduces conservation as a consequence of interactions.*

- **E1.** State the conditions under which **total momentum is conserved** for a simple two-particle system.  
- **E2.** Apply momentum conservation to interactions such as collisions and separations using signed quantities.  
- **E3.** Explain conservation of momentum as a redistribution of momentum within the system, not a disappearance or creation of motion.

---

### Performance Criteria
- Direction is represented **consistently and explicitly** using sign conventions.  
- Momentum, impulse, and force are used in a **conceptually coherent hierarchy**.  
- Calculations and explanations correctly distinguish between **individual** and **net** quantities.  
- Reasoning emphasizes physical meaning: how and why momentum changes, and when it is conserved.
