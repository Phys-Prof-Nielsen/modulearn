## Problem 1: Momentum as a Property of Moving Matter

**LOC addressed:** A1, A2, A3

A point particle has mass

$$
m = 1.50\ \text{kg}
$$

and velocity  

$$
v_x = -12.0\ \text{m/s}.
$$

1. Compute the particle’s momentum $p_x$.  
2. Explain what the **sign** of the momentum tells you physically.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 1 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

Momentum in 1D is defined as

$$
p_x = m v_x.
$$

Substitute:

$$
p_x = (1.50)(-12.0) = -18.0\ \text{kg·m/s}.
$$

The **negative sign** indicates that the particle’s motion (and momentum) is directed in the **negative $x$-direction**.

$$
\boxed{p_x = -18.0\ \text{kg·m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Given values
m = 1.50      # kg
v_x = -12.0   # m/s

# Momentum
p_x = m * v_x

# Output with appropriate formatting
print(f"Momentum p_x = {p_x:.2e} kg·m/s")
```

## Problem 2: Impulse as Momentum Transfer

**LOC addressed:** B1, B2, B3

A $0.60\ \text{kg}$ particle changes velocity from 

$$
v_{x,i} = +4.0\ \text{m/s}
$$

to  

$$
v_{x,f} = -1.0\ \text{m/s}.
$$

1. Predict the **direction** of the net impulse before calculating.  
2. Calculate the net impulse $j_x$.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 2 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

The velocity changes from positive to negative, so momentum becomes **more negative**.  
Therefore, the impulse must be in the **negative $x$-direction**.

Impulse is defined as momentum transfer:

$$
j_x = \Delta p_x = m(v_{x,f} - v_{x,i}).
$$

Compute:

$$
j_x = (0.60)(-1.0 - 4.0) = -3.0\ \text{kg·m/s}.
$$

$$
\boxed{j_x = -3.0\ \text{kg·m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Given values
m = 0.60        # kg
v_i = 4.0       # m/s
v_f = -1.0      # m/s

# Momentum transfer (impulse)
j_x = m * (v_f - v_i)

# Output with proper formatting
print(f"Net impulse j_x = {j_x:.1e} kg·m/s")
```

## Problem 3: Force Acting Through Time

**LOC addressed:** C1, C2, C3

A constant force 

$$
f_x = +6.0\ \text{N}
$$

acts on a particle for 

$$
\Delta t = 0.25\ \text{s}.
$$

1. Calculate the impulse delivered.  
2. Explain how the same impulse could be produced by a smaller force acting longer.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 3 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

For constant (or average) force,

$$
j_x = f_x\,\Delta t.
$$

Compute:

$$
j_x = (6.0)(0.25) = 1.50\ \text{N·s}
= 1.50\ \text{kg·m/s}.
$$

The **same impulse** could result from a smaller force acting over a longer time, illustrating that impulse depends on the **product** of force and time.

$$
\boxed{j_x = 1.50\ \text{kg·m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Given values
f_x = 6.0        # N
delta_t = 0.25   # s

# Impulse
j_x = f_x * delta_t

# Output with proper formatting
print(f"Impulse j_x = {j_x:.1e} kg·m/s")
```

## Problem 4: Impulse–Momentum Theorem

**LOC addressed:** D1, D2, D3

A particle of mass $m=2.0\ \text{kg}$ starts from rest.  
A net force acts for $0.40\ \text{s}$, producing a net impulse of  

$$
j_{\text{net},x} = +8.0\ \text{kg·m/s}.
$$

1. Determine the final momentum of the particle.  
2. Find the final velocity.  
3. Explain why the elapsed time (and indirectly acceleration) is not required to solve this problem.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 4 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

Impulse–momentum theorem:

$$
j_{\text{net},x} = \Delta p_x.
$$

Since the particle starts from rest,

$$
p_{x,f} = j_{\text{net},x} = 8.0\ \text{kg·m/s}.
$$

Final velocity:

$$
v_{x,f} = \frac{p_{x,f}}{m} = \frac{8.0}{2.0} = 4.0\ \text{m/s}.
$$

The elapsed time is not necessary since we are already given the net impulse. If not provided, we would need either the elapsed time to find acceleration --> net force, or have the net force given.

$$
\boxed{v_{x,f} = 4.0\ \text{m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp

# Given values
m = 2.0        # kg
j_net = 8.0    # kg·m/s
v_i = 0.0      # m/s

# Symbols
v_f = sp.symbols('v_f')

# Initial and final momentum
p_i = m * v_i
p_f = m * v_f

# Impulse–momentum theorem: j_net = p_f - p_i
eq = sp.Eq(j_net, p_f - p_i)

# Solve for final velocity
v_f_sol = sp.solve(eq, v_f)[0]

# Compute final momentum
p_f_sol = p_f.subs(v_f, v_f_sol)

# Output with proper formatting
print(f"Final momentum p_x,f = {float(p_f_sol):.1e} kg·m/s")
print(f"Final velocity v_x,f = {float(v_f_sol):.1e} m/s")
```

## Problem 5: Conservation of Momentum in a Two-Particle System

**LOC addressed:** E1, E2, E3

Two particles interact in 1D with **no external impulse**:

- $m_1 = 1.0\ \text{kg}$, $v_{1,x,i}=+2.0\ \text{m/s}$  
- $m_2 = 2.0\ \text{kg}$, $v_{2,x,i}=0$

After the interaction,

$$
v_{1,x,f} = -1.0\ \text{m/s}.
$$

1. Find the final velocity of particle 2.  
2. Explain why total momentum is conserved.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem 5 solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

With zero external impulse, total momentum is conserved:

$$
P_{x,i} = P_{x,f}.
$$

Compute:

$$
(1.0)(2.0) + 0 = (1.0)(-1.0) + 2.0\,v_{2,x,f}.
$$

Solve:

$$
3.0 = 2.0\,v_{2,x,f}
\quad\Rightarrow\quad
v_{2,x,f} = 1.5\ \text{m/s}.
$$

Momentum is conserved because **internal impulses cancel**, redistributing momentum within the system rather than changing the total.

$$
\boxed{v_{2,x,f} = +1.5\ \text{m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp

# Given values
m1 = 1.0        # kg
m2 = 2.0        # kg
v1_i = 2.0      # m/s
v2_i = 0.0      # m/s
v1_f = -1.0     # m/s

# Symbol for unknown final velocity of particle 2
v2_f = sp.symbols('v2_f')

# Total momentum before and after
P_i = m1*v1_i + m2*v2_i
P_f = m1*v1_f + m2*v2_f

# Conservation of momentum: P_i = P_f
eq = sp.Eq(P_i, P_f)

# Solve for v2_f
v2_f_sol = sp.solve(eq, v2_f)[0]

# Final momentum of particle 2
p2_f = m2 * v2_f_sol

# Output with proper formatting
print(f"Final velocity v_2,x,f = {float(v2_f_sol):.1e} m/s")
print(f"Final momentum p_2,x,f = {float(p2_f):.1e} kg·m/s")
```

## Problem 6: Interpreting a Force–Time Graph to Determine Impulse

**LOC addressed:** C4, B1, B2, D1

A point particle moves along the $x$-axis and experiences a force that varies with time as follows:

- From $t = 0$ to $t = 0.20\ \text{s}$, the force increases linearly from $0$ to $+10\ \text{N}$.  
- From $t = 0.20\ \text{s}$ to $t = 0.50\ \text{s}$, the force remains constant at $+10\ \text{N}$.  

The particle has mass

$$
m = 2.0\ \text{kg}.
$$

1. Sketch (or plot) the force–time graph $f_x(t)$.  
2. Use the **area under the force–time graph** to determine the impulse delivered to the particle.  
3. If the particle starts from rest, determine its final velocity.  
4. Explain why the area under the force–time graph represents impulse.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Interpreting the force–time graph**

The force–time graph consists of:
- a **triangle** from $t=0$ to $0.20\ \text{s}$ with height $10\ \text{N}$,
- followed by a **rectangle** from $t=0.20$ to $0.50\ \text{s}$ with height $10\ \text{N}$.

---

**Step 2 — Impulse from the graph (area)**

Impulse equals the **area under the force–time curve**.

**Triangle area:**

$$
j_{\text{tri}} = \frac{1}{2}(0.20)(10) = 1.0\ \text{N·s}.
$$

**Rectangle area:**

$$
j_{\text{rect}} = (0.30)(10) = 3.0\ \text{N·s}.
$$

**Total impulse:**

$$
j_x = 1.0 + 3.0 = 4.0\ \text{N·s}.
$$

Using $1\ \text{N·s} = 1\ \text{kg·m/s}$:

$$
j_x = 4.0\ \text{kg·m/s}.
$$

---

**Step 3 — Final velocity (starts from rest)**

Impulse–momentum theorem:

$$
j_x = \Delta p_x = m v_{x,f}.
$$

Solve for final velocity:

$$
v_{x,f} = \frac{j_x}{m} = \frac{4.0}{2.0} = 2.0\ \text{m/s}.
$$

---

**Step 4 — Physical interpretation**

A force–time graph shows **how strongly** and **for how long** an interaction acts.  
The **area under the curve** accumulates the effect of force over time, which is precisely the definition of **impulse**—the total momentum transferred during the interaction.

---

**Final Answers**

$$
\boxed{j_x = 4.0\ \text{kg·m/s}}, \qquad
\boxed{v_{x,f} = +2.0\ \text{m/s}}
$$

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt

# Given values
m = 2.0          # kg
F_max = 10.0     # N
t1 = 0.20        # s
t2 = 0.50        # s

# Time segments
t_ramp = np.linspace(0.0, t1, 300)
t_const = np.linspace(t1, t2, 300)

# Force segments
f_ramp = (F_max / t1) * t_ramp          # linear increase
f_const = F_max * np.ones_like(t_const) # constant force

# Plot force vs time
plt.figure()
plt.plot(t_ramp, f_ramp)
plt.plot(t_const, f_const)

# Shade area under the curve (impulse)
plt.fill_between(t_ramp, f_ramp, alpha=0.3)
plt.fill_between(t_const, f_const, alpha=0.3)

plt.xlabel("time (s)")
plt.ylabel("force $f_x$ (N)")
plt.title("Impulse as Area Under Force–Time Graph")
plt.show()

# Numerical impulse (area under curve)
t_all = np.concatenate((t_ramp, t_const))
f_all = np.concatenate((f_ramp, f_const))
j = np.trapezoid(f_all, t_all)

# Final velocity (starts from rest)
v_f = j / m

# Output
print(f"Impulse j_x = {j:.1e} kg·m/s")
print(f"Final velocity v_x,f = {v_f:.1e} m/s")
```
