## Problem 1: Linear Momentum with Direction and Units

Calculate the **linear momentum** for each of the following cases.  
Be careful with **units**, **signs**, and **scientific notation**.

1. A **proton** moving in the **+x-direction** with a speed equal to **99.9% of the speed of light**.  
2. A **2000.0 kg sedan** traveling with a velocity of **–70.0 mph**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equation**

For 1D motion, linear momentum is

$$
p_x = m v_x.
$$

---

**Part 1 — Proton at relativistic speed (Newtonian treatment)**

**Given**
- Proton mass:  

  $$m_p = 1.67\times10^{-27}\ \text{kg}$$
  
- Speed:  

  $$v_x = +0.999\,c = +0.999(3.00\times10^8)\ \text{m/s}$$

**Compute velocity**

$$
v_x = +2.997\times10^8\ \text{m/s}.
$$

**Momentum**

$$
p_x = (1.67\times10^{-27})(2.997\times10^8)
= 5.00\times10^{-19}\ \text{kg·m/s}.
$$

**Result**

$$
\boxed{p_x \approx +5.00\times10^{-19}\ \text{kg·m/s}}
$$

> *Note:* At this speed, a fully correct treatment requires **relativistic momentum**.  
> Here we are intentionally using the Newtonian definition for comparison and scale.

---

**Part 2 — Sedan traveling at –70.0 mph**

**Given**
- Mass:  

  $$m = 2000.0\ \text{kg}$$
  
- Velocity:  

  $$v_x = -70.0\ \text{mph}$$

**Convert mph to m/s**

$$
1\ \text{mph} = 0.44704\ \text{m/s}
$$

$$
v_x = -70.0(0.44704) = -31.3\ \text{m/s}.
$$

**Momentum**

$$
p_x = (2000.0)(-31.3) = -6.26\times10^4\ \text{kg·m/s}.
$$

**Result**

$$
\boxed{p_x = -6.26\times10^4\ \text{kg·m/s}}
$$

---

**Final Summary**

- Proton: very small momentum due to extremely small mass  
- Sedan: much larger momentum due to large mass, despite much lower speed  

The **sign** of each momentum reflects the **direction of motion along the x-axis**.

</div>

Take the **northbound direction as positive**.

## Problem 4: Impulse as Momentum Transfer

A point particle has mass  

$$
m = 0.40\ \text{kg}.
$$

It moves initially with velocity  

$$
v_{x,i} = +6.0\ \text{m/s},
$$

and after an interaction its velocity becomes  

$$
v_{x,f} = +2.0\ \text{m/s}.
$$

1. Compute the impulse $j_x$ delivered to the particle.  
2. State whether the impulse is positive or negative.  
3. Describe what the sign tells you about the direction of the interaction.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1: Relevant equation**

Impulse is momentum transfer:

$$
j_x = \Delta p_x = p_{x,f} - p_{x,i}.
$$

For a point particle,

$$
p_x = m v_x,
$$

so

$$
j_x = m(v_{x,f} - v_{x,i}).
$$

---

**Step 2: Calculate the impulse**

$$
j_x = (0.40)\left(2.0 - 6.0\right)
= (0.40)(-4.0)
= -1.6\ \text{kg·m/s}.
$$

---

**Step 3: Sign and interpretation**

The impulse is **negative**, which means the net momentum transfer was in the **negative $x$-direction**.

Even though the particle continued moving in the positive direction, the interaction transferred momentum in the opposite direction, reducing the particle’s momentum.

---

**Final Answer**

$$
\boxed{j_x = -1.6\ \text{kg·m/s}}
$$

</div>
</details>

## Problem 6: Momentum Change and Impulse (1D Signed Motion)

A particle of mass  

$$
m = 5.0\times10^{-3}\ \text{kg}
$$

moves in 1D along the $z$-axis.

- Initially, it has a **speed** of $335\ \text{m/s}$ in the **$-z$ direction**.
- After some elapsed time, its **velocity** is  

  $$
  v_{z,f} = -875\ \text{m/s}.
  $$

1. What is the initial velocity of the particle, $v_{z,i}$?  
2. What are the initial and final momenta, $p_{z,i}$ and $p_{z,f}$?  
3. Before calculating the net impulse, predict the **direction** of the impulse. Explain briefly.  
4. Calculate the momentum transfer (net impulse) $j_{\text{net},z} = \Delta p_z$.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Initial velocity from speed and direction**

A **speed** of $335\ \text{m/s}$ in the $-z$ direction means the signed velocity is

$$
v_{z,i} = -335\ \text{m/s}.
$$

---

**Step 2 — Initial and final momentum**

For 1D motion,

$$
p_z = m v_z.
$$

**Initial momentum**

$$
p_{z,i} = (5.0\times10^{-3})(-335)
= -1.675\ \text{kg·m/s}
\approx -1.7\ \text{kg·m/s}.
$$

**Final momentum**

$$
p_{z,f} = (5.0\times10^{-3})(-875)
= -4.375\ \text{kg·m/s}
\approx -4.4\ \text{kg·m/s}.
$$

---

**Step 3 — Predict the direction of impulse (before calculating)**

The velocity changes from $-335$ to $-875\ \text{m/s}$, meaning the particle becomes **more negative** (it speeds up in the $-z$ direction).

So the momentum becomes **more negative**, which implies

$$
\Delta p_z = p_{z,f} - p_{z,i} < 0.
$$

Therefore, the net impulse should be in the **negative $z$ direction**.

---

**Step 4 — Net impulse (momentum transfer)**

Net impulse is momentum transfer:

$$
j_{\text{net},z} = \Delta p_z = p_{z,f} - p_{z,i}.
$$

Compute:

$$
j_{\text{net},z} = (-4.375) - (-1.675) = -2.700\ \text{kg·m/s}.
$$

With appropriate significant figures,

$$
j_{\text{net},z} \approx -2.7\ \text{kg·m/s}.
$$

---

**Final Answers**

1. $$\boxed{v_{z,i} = -335\ \text{m/s}}$$  
2. $$\boxed{p_{z,i} \approx -1.7\ \text{kg·m/s}},\qquad \boxed{p_{z,f} \approx -4.4\ \text{kg·m/s}}$$  
3. Impulse is predicted to be in the **$-z$ direction** because the momentum becomes more negative.  
4. $$\boxed{j_{\text{net},z} \approx -2.7\ \text{kg·m/s}}$$

## Problem 7: Impulse from a Constant Force (Units + Interpretation)

A constant force acts on a point particle along the $x$-axis:

$$
f_x = -5.0\ \text{N}
$$

for a time interval

$$
\Delta t = 0.30\ \text{s}.
$$

1. Compute the impulse $j_x$ delivered by this force.  
2. Show that the unit of impulse can be written as **kg·m/s**.  
3. Explain what the negative sign of your answer means physically.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1: Relevant equation**

For a constant (or average) force,

$$
j_x = f_x\,\Delta t.
$$

---

**Step 2: Compute the impulse**

$$
j_x = (-5.0)(0.30) = -1.5\ \text{N·s}.
$$

---

**Step 3: Show impulse units as kg·m/s**

Using

$$
1\ \text{N} = 1\ \text{kg·m/s}^2,
$$

we get

$$
\text{N·s} = \left(\text{kg·m/s}^2\right)\text{s} = \text{kg·m/s}.
$$

So the impulse can be written as

$$
j_x = -1.5\ \text{kg·m/s}.
$$

---

**Step 4: Interpret the sign**

The negative sign means the momentum transfer (impulse) is in the **negative $x$-direction**.  
In other words, the interaction pushes/pulls in the negative direction during the time interval.

---

**Final Answer**

$$
\boxed{j_x = -1.5\ \text{kg·m/s}}
$$

</div>
</details>

## Problem 8: Frictional Force from Momentum Change

You observe a **point particle–like object** of mass  

$$
m = 2.0\ \text{kg}
$$

moving in 1D with an initial velocity  

$$
v_{x,i} = -45\ \text{m/s}.
$$

The object is subjected **only to a frictional force** and comes to rest after  

$$
\Delta t = 8.0\ \text{s}.
$$

1. Determine the **direction** of the frictional force before doing any calculations.  
2. Calculate the **magnitude and direction** of the frictional force.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Predict the direction of the force**

The object is initially moving in the **negative $x$-direction** and comes to rest.  
Friction always acts **opposite the direction of motion**, so the frictional force must act in the **positive $x$-direction**.

---

**Step 2 — Use impulse–momentum theorem**

The net impulse equals the change in momentum:

$$
j_{\text{net},x} = \Delta p_x = p_{x,f} - p_{x,i}.
$$

---

**Step 3 — Compute initial and final momentum**

Initial momentum:

$$
p_{x,i} = m v_{x,i} = (2.0)(-45) = -90\ \text{kg·m/s}.
$$

Final velocity is zero, so final momentum:

$$
p_{x,f} = 0.
$$

Change in momentum:

$$
\Delta p_x = 0 - (-90) = +90\ \text{kg·m/s}.
$$

---

**Step 4 — Relate impulse to force**

Since friction is the only force acting and it acts over the entire time interval,

$$
j_{\text{net},x} = f_x\,\Delta t.
$$

Solve for the frictional force:

$$
f_x = \frac{\Delta p_x}{\Delta t} = \frac{90}{8.0} = 11.25\ \text{N}.
$$

---

**Step 5 — Interpret the result**

The force is **positive**, meaning it acts in the **positive $x$-direction**, opposite the object’s initial motion—as expected for friction.

---

**Final Answer**

$$
\boxed{f_x \approx +11\ \text{N}}
$$

The frictional force acts in the **positive $x$-direction**.

</div>
</details>

## Problem 9: Impact Force in Two Head-On Collisions

A $2.0\ \text{kg}$ particle travels at  

$$
v_{x,i} = +25\ \text{m/s}
$$

toward a fixed wall. Two experiments are performed, each with the **same impact time**

$$
\Delta t = 0.020\ \text{s}.
$$

- **Experiment A:** The particle sticks to the wall after the collision.  
- **Experiment B:** The particle bounces off the wall with the **same speed**.

Assume the **impact force is much larger than any other forces** during the collision.

1. Predict which experiment produces the **larger impact force**, and explain why.  
2. Quantify your prediction by calculating the **average impact force** in both cases.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Conceptual prediction**

Impact force is tied to **impulse**, which equals the **change in momentum**:

$$
j_{\text{net},x} = \Delta p_x.
$$

- **Sticking**: momentum goes from $+mv$ to $0$.  
- **Bouncing back**: momentum goes from $+mv$ to $-mv$ (a larger change).

Therefore, the **bounce-back collision** produces a **larger momentum change** and thus a **larger impact force** (for the same $\Delta t$).

---

**Step 2 — Compute impulse in each experiment**

Given $m=2.0\ \text{kg}$ and $v=25\ \text{m/s}$:

**Experiment A (sticks):**

$$
p_{x,i} = (2.0)(25)=50\ \text{kg·m/s}, \qquad p_{x,f}=0
$$

$$
\Delta p_x = 0-50 = -50\ \text{kg·m/s}.
$$

**Experiment B (bounces with same speed):**

$$
p_{x,f} = (2.0)(-25)= -50\ \text{kg·m/s}
$$

$$
\Delta p_x = -50-50 = -100\ \text{kg·m/s}.
$$

---

**Step 3 — Average impact force**

Using

$$
\bar f_x = \frac{\Delta p_x}{\Delta t},
$$

**Experiment A (sticks):**

$$
\bar f_x = \frac{-50}{0.020} = -2.5\times10^3\ \text{N}.
$$

**Experiment B (bounces):**

$$
\bar f_x = \frac{-100}{0.020} = -5.0\times10^3\ \text{N}.
$$

(The negative sign indicates the force acts opposite the incoming direction.)

---

**Final Answers**

- **Larger impact force:** **Bounce-back collision**  
- **Average impact forces:**

$$
\boxed{\bar f_x = -2.5\times10^3\ \text{N}\ \text{(sticks)}},\qquad
\boxed{\bar f_x = -5.0\times10^3\ \text{N}\ \text{(bounces)}}
$$

For the same impact time, **doubling the momentum change doubles the impact force**.

</div>
</details>

## Problem 10: Impulse from a Force–Time Graph (Area Interpretation)

A point particle experiences a force along the $x$-axis that varies with time as shown conceptually by a **triangular force–time pulse**:

- The force increases linearly from $0$ to $8.0\ \text{N}$ over a time interval of $1.2\ \text{s}$.
- After reaching $8.0\ \text{N}$, the force immediately drops back to zero.

1. Compute the impulse delivered to the particle.  
2. If the particle has mass $m = 0.60\ \text{kg}$ and initial velocity $v_{x,i} = +1.5\ \text{m/s}$, find the final velocity.  
3. Explain why the detailed shape of the force–time curve is less important than its area.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1: Impulse from the force–time graph**

Impulse is given by the **area under the force–time curve**.

For a triangular pulse,

$$
j_x = \frac{1}{2}(\text{base})(\text{height}).
$$

Here,

$$
j_x = \frac{1}{2}(1.2)(8.0) = 4.8\ \text{N·s}.
$$

Using $1\ \text{N·s} = 1\ \text{kg·m/s}$,

$$
j_x = 4.8\ \text{kg·m/s}.
$$

---

**Step 2: Apply impulse–momentum theorem**

Initial momentum:

$$
p_{x,i} = m v_{x,i} = (0.60)(1.5) = 0.90\ \text{kg·m/s}.
$$

Final momentum:

$$
p_{x,f} = p_{x,i} + j_x = 0.90 + 4.8 = 5.7\ \text{kg·m/s}.
$$

Final velocity:

$$
v_{x,f} = \frac{p_{x,f}}{m} = \frac{5.7}{0.60} = 9.5\ \text{m/s}.
$$

---

**Step 3: Conceptual interpretation**

Although the force changes with time, the **total momentum transfer** depends only on the **area under the force–time curve**, not on the exact shape of the curve.

Different force–time profiles with the same area produce the same impulse and therefore the same change in momentum.

---

**Final Answer**

$$
\boxed{j_x = 4.8\ \text{kg·m/s}}, \qquad
\boxed{v_{x,f} = 9.5\ \text{m/s}}
$$

</div>
</details>

## Problem 11: Two-Particle Momentum Conservation (1D)

Two carts move along a straight, frictionless track and interact briefly.

- Cart 1 has mass  

  $$
  m_1 = 1.0\ \text{kg}
  $$

  and initial velocity  

  $$
  v_{1,x,i} = +3.0\ \text{m/s}.
  $$

- Cart 2 has mass  

 $$
  m_2 = 2.0\ \text{kg}
  $$

  and is initially at rest.

After the interaction, cart 1 is observed to move with velocity  

$$
v_{1,x,f} = -1.0\ \text{m/s}.
$$

1. Compute the initial total momentum of the system.  
2. Determine the final velocity of cart 2.  
3. Explain why cart 1 reverses direction even though no external impulse acts on the system.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1: Conservation of momentum**

Because the net external impulse on the two-cart system is zero, total momentum is conserved:

$$
P_{x,i} = P_{x,f}.
$$

For a two-particle system in 1D,

$$
m_1 v_{1,x,i} + m_2 v_{2,x,i}
=
m_1 v_{1,x,f} + m_2 v_{2,x,f}.
$$

---

**Step 2: Compute the initial momentum**

Initial total momentum:

$$
P_{x,i} = (1.0)(+3.0) + (2.0)(0) = 3.0\ \text{kg·m/s}.
$$

---

**Step 3: Solve for the final velocity of cart 2**

Substitute known values:

$$
3.0 = (1.0)(-1.0) + (2.0)v_{2,x,f}.
$$

Rearranging,

$$
3.0 + 1.0 = 2.0\,v_{2,x,f},
$$

$$
v_{2,x,f} = \frac{4.0}{2.0} = 2.0\ \text{m/s}.
$$

---

**Step 4: Physical interpretation**

## Problem 12: Net Impulse from Multiple Forces (Same Time Interval)

A point particle experiences three forces along the $x$-axis for the same time interval

$$
\Delta t = 0.40\ \text{s}.
$$

The forces are:
- $f_{1,x} = +6.0\ \text{N}$
- $f_{2,x} = -2.0\ \text{N}$
- $f_{3,x} = -1.0\ \text{N}$

The particle has mass

$$
m = 2.0\ \text{kg}
$$

and starts from rest.

1. Compute the net force $f_{\text{net},x}$.  
2. Compute the net impulse $j_{\text{net},x}$.  
3. Determine the final velocity $v_{x,f}$.  
4. State what the sign of $v_{x,f}$ tells you physically.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1: Net force**

Since all forces act over the same time interval, we first add them:

$$
f_{\text{net},x} = f_{1,x} + f_{2,x} + f_{3,x}
= 6.0 - 2.0 - 1.0
= +3.0\ \text{N}.
$$

---

**Step 2: Net impulse**

For forces acting over the same $\Delta t$,

$$
j_{\text{net},x} = f_{\text{net},x}\,\Delta t
= (3.0)(0.40)
= 1.2\ \text{N·s}.
$$

Using $1\ \text{N·s} = 1\ \text{kg·m/s}$,

$$
j_{\text{net},x} = 1.2\ \text{kg·m/s}.
$$

---

**Step 3: Use impulse–momentum theorem to find final velocity**

Impulse–momentum theorem:

$$
j_{\text{net},x} = \Delta p_x = p_{x,f} - p_{x,i}.
$$

Since the particle starts from rest, $p_{x,i} = 0$.

So,

$$
p_{x,f} = j_{\text{net},x} = 1.2\ \text{kg·m/s}.
$$

Then

$$
v_{x,f} = \frac{p_{x,f}}{m} = \frac{1.2}{2.0} = 0.60\ \text{m/s}.
$$

---

**Step 4: Interpret the sign**

The positive final velocity means the particle ends up moving in the **positive $x$-direction**.

---

**Final Answer**

$$
\boxed{f_{\text{net},x} = +3.0\ \text{N}},\quad
\boxed{j_{\text{net},x} = 1.2\ \text{kg·m/s}},\quad
\boxed{v_{x,f} = +0.60\ \text{m/s}}
$$

**Step 1: Compute each impulse separately**

Because the forces act over **different time intervals**, we must add impulses:

$$
j_{\text{net},x} = j_{1,x} + j_{2,x} + j_{3,x}.
$$

For each force,

$$
j_{k,x} = f_{k,x}\,\Delta t_k.

j_{1,x} = (8.0)(0.10) = 0.80\ \text{N·s},

## Problem 14: Use Python to Solve (Two-Particle Momentum Conservation)

Two carts interact on a frictionless track (1D). Assume the net external impulse on the two-cart system is zero.

- $m_1 = 0.75\ \text{kg}, \qquad v_{1,x,i} = +3.20\ \text{m/s}$
- $m_2 = 1.50\ \text{kg}, \qquad v_{2,x,i} = -0.80\ \text{m/s}$
- After the interaction: $v_{1,x,f} = +0.50\ \text{m/s}$

**Task:** Use Python to compute the final velocity $v_{2,x,f}$.

- Show **two solution methods** in Python:
  1. a direct calculator approach (algebra solved by you),
  2. a SymPy equation-solver approach.

Format your final numerical answer using scientific notation with **three significant figures**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Method 1 — Direct calculator approach**

`m1 = 0.75`  
`m2 = 1.50`  
`v1_i = 3.20`  
`v2_i = -0.80`  
`v1_f = 0.50`  

`P_i = m1*v1_i + m2*v2_i`  
`v2_f = (P_i - m1*v1_f)/m2`

`import sympy as sp`  

`m1 = 0.75`

`sol = sp.solve(eq, v2_f)[0]`  

`print(f"{float(sol):.2e}")`

## Problem 15: Electron Acceleration and Momentum (Python + SymPy)

In a particle accelerator, an electron of mass  

$$
m_e = 9.11\times10^{-31}\ \text{kg}
$$

is accelerated along a straight line.

Its velocity increases linearly with time according to

$$
v_x(t) = \frac{v_f}{t_f}\,t,
$$

where

$$
t_f = 2.0\times10^{-9}\ \text{s}
$$

is the total acceleration time, and $v_f$ is the final velocity at $t=t_f$.

At the end of the acceleration period, the electron’s momentum is measured to be

$$
p_x = 1.82\times10^{-22}\ \text{kg·m/s}.
$$

1. Use **SymPy’s `solve` function** to determine the final velocity $v_f$.  
2. Plot the velocity–time curve $v_x(t)$ from $t=0$ to $t=t_f$.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Momentum relation at final time**

At $t=t_f$, the velocity equals $v_f$, so the electron’s momentum is

$$
p_x = m_e v_f.
$$

We solve this equation symbolically for $v_f$ using SymPy.

---

**Step 2 — Solve for the final velocity**

`import sympy as sp`

`m_e = 9.11e-31`  
`p_x = 1.82e-22`  

`v_f = sp.symbols('v_f')`

`eq = sp.Eq(m_e*v_f, p_x)`  
`sol = sp.solve(eq, v_f)[0]`

`print(f"{float(sol):.3e}")`

$$
v_f \approx 2.00\times10^8\ \text{m/s}.
$$

---

**Step 3 — Define and plot the velocity–time function**

The velocity increases linearly:

$$
v_x(t) = \frac{v_f}{t_f}\,t.

The following Python script plots $v_x(t)$ from $t=0$ to $t=t_f$.

`import numpy as np`  
`import matplotlib.pyplot as plt`

`t_f = 2.0e-9`  
`v_f_val = float(sol)`

`t = np.linspace(0, t_f, 200)`  
`v = (v_f_val/t_f)*t`

`plt.figure()`  
`plt.plot(t, v)`  
`plt.xlabel("time (s)")`  
`plt.ylabel("velocity v_x (m/s)")`  
`plt.title("Electron Velocity vs. Time")`  
`plt.show()`

---

**Interpretation**

- The straight-line graph confirms **constant acceleration**.
- The final velocity is a significant fraction of the speed of light, signaling that **relativistic effects** would be important in a more accurate treatment.
- Here, we intentionally use the Newtonian momentum definition to focus on impulse, momentum, and computational tools.

**Final Answer**

$$
\boxed{v_f \approx 2.00\times10^8\ \text{m/s}}
$$

## Problem 17: Force–Time Plot, Impulse, and Initial Velocity

A point particle of mass  

$$
m = 1.50\ \text{kg}
$$
moves along the $x$-axis and experiences a time-varying force during a short interaction.

The force acts over the time interval $0 \le t \le 0.40\ \text{s}$ and is described by:

- The force increases linearly from $0$ to $+6.0\ \text{N}$ during the first $0.20\ \text{s}$.
- The force then decreases linearly back to $0$ by $t = 0.40\ \text{s}$.

After the interaction, the particle is observed to have a velocity

$$
v_{x,f} = +3.0\ \text{m/s}.
$$

1. Use Python to **plot the force–time curve**.  
2. Use the graph to **calculate the impulse** delivered to the particle.  
3. Determine the **initial velocity** of the particle.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 1 — Force–time model and plot**

The force–time curve forms a **triangle** with:
- base = $0.40\ \text{s}$
- height = $6.0\ \text{N}$

Python script to plot the force:

`import numpy as np`  
`import matplotlib.pyplot as plt`

`t = np.linspace(0, 0.40, 400)`  
`f = np.where(t <= 0.20, 30*t, 30*(0.40 - t))`

`plt.figure()`  
`plt.plot(t, f)`  
`plt.xlabel("time (s)")`  
`plt.ylabel("force $f_x$ (N)")`  
`plt.title("Force vs. Time")`  
`plt.show()`

---

**Step 2 — Impulse from the graph**

Impulse equals the **area under the force–time curve**:

$$
j_x = \frac{1}{2}(\text{base})(\text{height})
= \frac{1}{2}(0.40)(6.0)
= 1.20\ \text{N·s}.
$$

Using $1\ \text{N·s} = 1\ \text{kg·m/s}$:

$$
j_x = 1.20\ \text{kg·m/s}.
$$

---

**Step 3 — Use impulse–momentum theorem**

Impulse–momentum theorem:

$$
j_x = p_{x,f} - p_{x,i}.
$$

Final momentum:

$$
p_{x,f} = m v_{x,f} = (1.50)(3.0) = 4.50\ \text{kg·m/s}.

Initial momentum:

$$
p_{x,i} = p_{x,f} - j_x = 4.50 - 1.20 = 3.30\ \text{kg·m/s}.
$$

Initial velocity:

$$
v_{x,i} = \frac{p_{x,i}}{m} = \frac{3.30}{1.50} = 2.2\ \text{m/s}.
$$

---

**Interpretation**

- The **area** under the force–time curve determines the momentum transfer.
- Knowing the final velocity allows us to work **backward in time** to find the initial velocity.
- Different force shapes with the same area would produce the same impulse.

---

**Final Answers**

$$
\boxed{j_x = 1.20\ \text{kg·m/s}},\qquad
\boxed{v_{x,i} = +2.2\ \text{m/s}}
