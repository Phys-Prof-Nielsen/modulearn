## Motivation

Modern physics is inseparable from computation. Whether modeling planetary motion, analyzing image data, processing data from sensors, or exploring the behavior of oscillations and waves, **Python has become the universal scientific language**.  

In this module, you will begin using the interactive **code cells embedded directly in this e-book** (via Thebe). These cells allow you to type Python code, run simulations, make quick calculations, and visualize results—all within the same environment where you learn the physics.

This early exposure matters. Python helps you check your intuition, explore “what-if?” scenarios, build computational thinking skills, and prepare for later phases where you will model motion, forces, energy, oscillations, and more. Even students who do not plan to become programmers benefit from understanding the basics, as computational tools have become standard in science, engineering, and data-driven disciplines.

## Learning Outcomes

By the end of this module, you should be able to:

### Conceptual Understanding
1. Explain why computation—and Python in particular—is important in physics and modern scientific work.  
2. Describe how embedded code cells in this e-book function and why live, interactive computation supports learning.  
3. Understand the purpose of **pseudocode** as a bridge between conceptual reasoning and actual programming.  

### Basic Python Skills
4. Use Python as a **scientific calculator** for arithmetic, exponents, and variable assignments.  
5. Execute and modify code in the e-book’s interactive cells (including running, resetting, and downloading code).  
6. Import and use essential scientific functions such as square root function, and constants from **NumPy** or **SymPy**.  

### Computational Problem-Solving Skills
7. Write simple pseudocode describing the steps for solving a physics problem computationally.  
8. Use Python to solve 1–2 algebraic equations numerically or symbolically.  
9. Use Python to evaluate expressions related to basic physics situations (e.g., motion, forces, energy).  

### Digital Literacy & Tool Awareness
10. Identify additional computational tools (e.g., vPython, Trinket, Replit, Wolfram Alpha, online graphing utilities) and know when each is appropriate.  
11. Recognize the difference between “black-box” tools and transparent computation where you control each step.
