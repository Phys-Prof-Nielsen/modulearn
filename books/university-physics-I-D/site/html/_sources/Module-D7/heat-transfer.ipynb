## D7.4.1 Motivation
When there is a temperature difference, heat flows from hotter regions to colder regions. But *how* does this transfer happen?  
In nature and technology, there are three main mechanisms: **conduction**, **convection**, and **radiation**.  
Each mechanism has distinct physical processes, but all are governed by the same principle: energy transfer due to temperature difference.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.4.2 Conduction
Conduction is **heat transfer through a solid (or stationary fluid)** without any bulk movement of matter.  
- Microscopic picture: energetic particles (atoms, electrons, phonons) collide with neighbors, transferring kinetic energy step by step.  
- Macroscopic description: heat flows from the warmer side of a solid to the cooler side.  

The rate of heat conduction is approximately proportional to the temperature gradient:

$$
\frac{Q}{t} = k A \frac{\Delta T}{\Delta x}
$$

where  
- $Q/t$ = heat transfer rate (Watts : W)  
- $k$ = thermal conductivity of the material (W/m·K)  
- $A$ = cross-sectional area (m²)  
- $\Delta x$ = thickness of the material (m)  
- $\Delta T$ = temperature difference across the material (K)  

**Examples:**  
- A metal spoon heating up in a pot of soup.  
- Heat loss through walls and windows of a house.

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Heat Loss by Conduction Through a House Wall</h3>

Consider a section of an exterior house wall during winter.  
The indoor temperature is

$$
T_{\text{inside}} = 20^\circ\text{C},
$$

while the outdoor temperature is

$$
T_{\text{outside}} = -5^\circ\text{C}.
$$

The wall has:
- thickness $\Delta x = 0.20\,\text{m},$

- surface area $= 10\,\text{m}^2$

- effective thermal conductivity $k = 0.04$ W/(m$\cdot$ K)

  (typical of insulated construction).

<br>

<strong>Calculation</strong><br>

The rate of heat transfer by conduction is approximated by

$$
\frac{Q}{t} = k A \frac{\Delta T}{\Delta x},
$$

where

$$
\Delta T = T_{\text{inside}} - T_{\text{outside}} = 25\,\text{K}.
$$

Substituting the values,

$$
\frac{Q}{t} = (0.04)(10)\frac{25}{0.20}
= 50\,\text{W}.
$$

<br>

<strong>Interpretation</strong><br>

This wall section loses heat at a rate of about

$$
50\,\text{J/s}.
$$

Over one hour, the energy lost is

$$
Q = (50)(3600) \approx 1.8\times10^5\,\text{J}.
$$

This continual energy loss must be replaced by a heating system to maintain a comfortable indoor temperature.

<br>

<strong>Key takeaway</strong><br>

Conduction through walls depends strongly on:
- the **temperature difference**,
- the **wall thickness**,
- and the **thermal conductivity** of the materials.

Good insulation reduces heat loss by having a **low thermal conductivity**, which is why modern buildings are designed with layered walls and insulating materials.

### Conduction and R-Value (Thermal Resistance)

In building construction and energy efficiency, heat conduction is often described using **thermal resistance**, commonly called the **R-value**, rather than thermal conductivity.

Starting from the conduction equation for steady, one-dimensional heat flow through a flat layer,

$$
\frac{Q}{t} = k A \frac{\Delta T}{\Delta x},
$$

we can rewrite it in the form

$$
\frac{Q}{t} = \frac{A\,\Delta T}{R},
$$

where the **thermal resistance** $R$ is defined as

$$
R = \frac{\Delta x}{k}.
$$

---

#### Physical meaning of R-value

The R-value measures how strongly a material **resists heat flow**:

- A **larger R-value** means **less heat transfer** for the same temperature difference.  
- Thicker materials have larger R-values.  
- Materials with lower thermal conductivity have larger R-values.

Because $R$ increases when heat flow becomes more difficult, R-value provides an intuitive way to describe insulation effectiveness.

---

#### R-values in construction

In real buildings, walls and roofs consist of multiple layers (drywall, insulation, sheathing, air gaps).  
For layers in series, the total thermal resistance is approximately the **sum of the individual R-values**:

$$
R_{\text{total}} = R_1 + R_2 + R_3 + \cdots
$$

This makes R-value especially useful in construction, where materials are selected and combined to achieve a target overall resistance to heat flow.

---

#### Why R-value is practical

Expressing conduction in terms of R-value shifts the focus from *how easily heat flows* to *how strongly a structure resists heat flow*.  
For a fixed temperature difference, increasing the R-value directly reduces the rate of heat loss or gain, lowering heating and cooling energy demands.

This is why building codes and insulation standards are typically specified in terms of **R-values** rather than thermal conductivity.

---

#### Units of thermal resistance (R-value)

In physics, thermal resistance is defined as

$$
R = \frac{\Delta x}{k},
$$

which has SI units

$$
[R]_{\text{SI}} = \text{m}^2\cdot\text{K/W}.
$$

This form is convenient for theoretical calculations and direct use in the conduction equation

$$
\frac{Q}{t} = \frac{A\,\Delta T}{R}.
$$

---

#### Construction R-values (Btu-based units)

In **building construction**, insulation performance is almost always quoted using a different unit system based on **British thermal units (Btu)**.  
In this context, R-value is defined with units

$$
[R]_{\text{construction}} = \text{ft}^2\cdot^\circ\text{F}\cdot\text{h/Btu}.
$$

These are the R-values seen on insulation labels such as **R-13, R-19, or R-30**.

Although the units differ from SI, the **physical meaning is the same**:
- A **larger R-value** means **greater resistance to heat flow**.
- Higher R-values correspond to **better insulation**.

---

#### Important clarification

The numerical value of an R-value **depends on the unit system** being used.

- SI R-values (m²·K/W)  
- Construction R-values (ft²·°F·h/Btu)

are **not numerically equal**, even though they describe the same physical concept.

---

#### Key takeaway

Construction R-values are expressed in **Btu-based units**, but they represent the same idea as thermal resistance in physics:  
**how strongly a material resists heat flow**.

## D7.4.3 Convection
Convection is **heat transfer by the bulk motion of a fluid (liquid or gas)**.  
- Hotter, less dense fluid regions rise, while cooler, denser regions sink.  
- This creates a cycle that moves heat from one place to another.  

There are two types:  
- **Natural convection:** driven by buoyancy due to density differences (e.g., warm air rising, ocean currents).  
- **Forced convection:** driven by external means such as a fan, pump, or wind.  

Convection is often modeled with Newton’s Law of Cooling:

$$
\frac{Q}{t} = h A \Delta T
$$

where  
- $h$ = convection heat transfer coefficient (W/m²·K)  
- $A$ = surface area in contact with fluid (m²)  
- $\Delta T$ = temperature difference between surface and fluid (K)  

**Examples:**  
- Boiling water circulating in a pot.  
- Warm air currents heating a room from a radiator.  
- Cooling of a car engine by circulating coolant.

## D7.4.4 Radiation
Radiation is **heat transfer by electromagnetic waves**, primarily in the infrared range.  
- Requires no medium: works through vacuum as well as matter.  
- All objects emit radiant energy according to their temperature.  

The Stefan–Boltzmann Law gives the power radiated from a surface:

$$
\frac{Q}{t} = \epsilon \sigma A T^4
$$

where  
- $\epsilon$ = emissivity (0–1), a measure of how effectively a surface emits radiation  
- $\sigma = 5.67 \times 10^{-8}\ \mathrm{W/m^2K^4}$ = Stefan–Boltzmann constant  
- $A$ = surface area (m²)  
- $T$ = absolute temperature (K)  

**Examples:**  
- Heat from the Sun reaching Earth through space.  
- A glowing stove element radiating infrared energy.  
- Energy loss from the human body even in cold, still air.

<div style="background-color:#e0f2f1; border-left:6px solid #00695c; padding:14px; border-radius:4px; margin-top:16px;">
<h3 style="margin-top:0; color:#000000;">Example — Feeling Heat from a Campfire</h3>

On a cool evening, you stand near a campfire.  
Even though the surrounding air may be cold, you feel **warmth on your face and hands** when facing the fire.

This heating occurs primarily through **radiation**, not conduction or convection.

<br>

<strong>Physical description</strong><br>

The burning wood in the fire reaches a high temperature and emits **infrared electromagnetic radiation**.  
This radiation travels through the air and is absorbed by your skin, increasing its internal energy and temperature.

No physical contact is required, and the warming occurs even if the air between you and the fire is still.

<br>

<strong>Energy transfer perspective</strong><br>

Because the fire is much hotter than your skin, radiant energy flows from the fire to you.  
If you turn away from the fire, the radiant heating decreases sharply, even though the air temperature around you remains nearly the same.

This directional nature of radiation explains why:
- You feel warmer on the side of your body facing the fire.
- A reflective surface (like a shiny metal sheet) can redirect heat toward you.
- Fireplaces often use reflective back panels to improve heating efficiency.

<br>

<strong>Key takeaway</strong><br>

Radiation transfers energy by **electromagnetic waves** and does not require a medium.  
In everyday life, radiation is responsible for the warmth felt from fires, sunlight, and hot surfaces—even across empty space.

<h3 style="margin-top:0; color:#000000;">Example — Solar Radiation and Intensity at Earth</h3>

On a clear day near noon, sunlight delivers a large amount of energy to Earth by **radiation**.  
This energy transfer is described using the **radiative intensity**, defined as the power received per unit area.

Near the top of Earth’s atmosphere, the Sun provides an average intensity of approximately

$$
I \approx 1.4\times10^3\ \text{W/m}^2,
$$

a value known as the *solar constant*.

<br>

<strong>Calculation</strong><br>

Suppose sunlight shines directly onto a flat surface of area

$$
A = 0.50\ \text{m}^2.
$$

The rate at which radiant energy is received by the surface is

$$
\frac{Q}{t} = IA.
$$

Substituting values,

$$
\frac{Q}{t} = (1.4\times10^3)(0.50)
= 7.0\times10^2\ \text{W}.
$$

<br>

<strong>Energy over time</strong><br>

If the surface is exposed to sunlight for

$$
t = 10\ \text{minutes} = 600\ \text{s},
$$

the total energy absorbed is

$$
Q = (700)(600)
= 4.2\times10^5\ \text{J}.
$$

<br>

<strong>Interpretation</strong><br>

In just ten minutes, sunlight delivers over

$$
400\ \text{kJ}
$$

of energy to a surface smaller than a typical tabletop.  
This explains why:
- dark surfaces heat up quickly in sunlight,
- solar panels can generate useful electrical power,
- and shade makes a dramatic difference in comfort on hot days.

<br>

<strong>Key takeaway</strong><br>

Radiation transfers energy in the form of electromagnetic waves.  
The **intensity** (power per area) determines how much energy is delivered, and even modest areas can receive large amounts of energy from the Sun.

## D7.4.5 Comparing the Mechanisms
- **Conduction:** particle-to-particle transfer through solids.  
- **Convection:** bulk fluid motion carries heat.  
- **Radiation:** electromagnetic waves carry energy through space.  

Often, more than one mechanism occurs together. For example, a hot cup of coffee cools by conduction through the cup walls, convection to the surrounding air, and radiation from the surface.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D7.4.6 Summary
- Heat always flows from high temperature to low temperature.  
- The three modes are **conduction, convection, and radiation**.  
- Each has its own governing equation:  
  - Conduction: $\tfrac{Q}{t} = k A \tfrac{\Delta T}{L}$  
  - Convection: $\tfrac{Q}{t} = h A \Delta T$  
  - Radiation: $\tfrac{Q}{t} = \epsilon \sigma A T^4$  
- Real-world heat transfer often involves a combination of all three.
