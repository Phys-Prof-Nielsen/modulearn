# D5.0 Learning Outcomes — Kinematics and the Emergence of Newton’s Second Law
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## Why this module matters

In Module D4, motion was analyzed without reference to acceleration or Newton’s laws. Instead, changes in motion were explained using **impulse–momentum** and **energy**, which apply broadly—even in situations where forces or accelerations vary in complicated ways.

In this module, we introduce **kinematics**, the mathematical description of motion itself. Rather than explaining *why* motion changes, kinematics focuses on **how position, velocity, and acceleration are related in time**.

We begin by defining the core kinematic quantities—**position**, **displacement**, **speed**, **velocity**, and **acceleration**—all in **one dimension**, using signed quantities to represent direction. We then derive the familiar kinematic equations for **constant acceleration**, emphasizing both their usefulness and their **limitations**.

Only after this foundation is established do we introduce **Newton’s second law**,

$$
f_x = ma_x,
$$

not as a starting axiom, but as a **derived relationship** that emerges naturally from the **impulse–momentum theorem** under specific assumptions. In this way, acceleration becomes a **measurable consequence of interaction**, not a core physics concept.

Together, kinematics and Newton’s second law allow us to make **quantitative predictions** about motion in the past, present, and future—highlighting the **deterministic structure of Newtonian mechanics** within its domain of validity.

Throughout the module, all motion is treated in **one dimension**, without vectors or calculus, preparing the ground for more general formulations introduced later in Phase B.

By the end of this module, you will be able to:

### A. Kinematic Quantities in One Dimension  
*Defines the language used to describe motion.*

- **A1.** Define **position** as a signed coordinate along a one-dimensional axis.  
- **A2.** Distinguish between **displacement** and **distance traveled**, interpreting displacement as a directed change in position.  
- **A3.** Define **average speed** and **average velocity**, and explain how they differ conceptually.  
- **A4.** Define **average acceleration** as the rate of change of velocity over a time interval.  
- **A5.** Use sign conventions consistently to describe direction in one-dimensional motion.
- **A6.** Interpret **position–time**, **velocity–time**, and **acceleration–time** graphs in one dimension by relating  
  - slopes to **rates of change** (velocity and acceleration), and  
  - areas to **accumulated quantities** (displacement and change in velocity),  
  and explain which graphical areas and slopes have physical meaning in kinematics.


---

### B. Constant-Acceleration Kinematics  
*Develops predictive equations under a specific physical assumption.*

- **B1.** State the assumption of **constant acceleration** and explain when it is physically reasonable.  
- **B2.** Use the definition of acceleration to relate velocity and time,

  $$
  v_x = v_{0_x} + a_{x}t.
  $$

- **B3.** Use the definition of average velocity under constant acceleration to relate position and time,

  $$
  x = x_0 + v_{0_x} t + \tfrac{1}{2} a_x t^2.
  $$

- **B4.** Use the time-independent kinematic equation (Torricelli’s equation),

  $$
  v_x^2 = v_{0_x}^2 + 2a_x(x - x_0),
  $$

  to relate speed and position without explicit reference to time.

---

### C. Scope and Limitations of Kinematic Equations  
*Clarifies what these equations do—and do not—describe.*

- **C1.** Identify the **physical assumptions** underlying constant-acceleration kinematics.  
- **C2.** Recognize situations where the kinematic equations **do not apply**, such as variable forces or accelerations.  
- **C3.** Explain why more general descriptions of motion require new mathematical tools introduced in later phases.

---

### D. Newton’s Second Law from Impulse–Momentum  
*Derives $ f_x = ma_x $ rather than postulating it.*

- **D1.** State the **impulse–momentum theorem**,

  $$
  j_x = \Delta p_x.
  $$

- **D2.** Recall that, for a constant force acting over a time interval, impulse can be written as

  $$
  j_x = f_x\,\Delta t.
  $$

- **D3.** Combine impulse–momentum with constant mass to obtain

  $$
  f_x = m a_x,
  $$

  without the use of calculus.  
- **D4.** Interpret Newton’s second law as a **special case** of a more general momentum-based description of motion.

---

### E. Determinism and Prediction in Newtonian Motion  
*Connects equations to physical meaning.*

- **E1.** Explain how knowing initial conditions and interactions allows prediction of motion forward and backward in time.  
- **E2.** Describe Newtonian dynamics as **deterministic** within its domain of validity.  
- **E3.** Use $ f_x = ma_x $ together with kinematics to make quantitative predictions about one-dimensional motion.

---

### Performance Criteria

- Motion is described using **clear physical definitions**, not memorized formulas.  
- Direction is handled consistently using **signed quantities**, not vector notation.  
- Kinematic equations are applied only when their **assumptions are valid**.  
- Newton’s second law is understood as a **derived relationship**, not a starting axiom.  
- Explanations emphasize **causal structure, assumptions, and limitations** of Newtonian mechanics.
