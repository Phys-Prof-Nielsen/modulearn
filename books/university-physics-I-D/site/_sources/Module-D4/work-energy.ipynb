# D4.3 Work–Energy: Force Acting Through Displacement
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## Motivation — Newton’s First Law Revisited (Energy View)

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Newton’s First Law</h3>

An object continues in its state of motion unless acted upon by an unbalanced interaction.
</div>

From an energy perspective, this statement can be interpreted as follows:

- If the motion of an object changes, then its **kinetic energy must have changed**.  
- If the motion does not change, then the **net change in kinetic energy must be zero**.

Seen this way, Newton’s First Law is not fundamentally a statement about force. It is a statement about the *absence of energy transfer*.

This perspective will guide how we now connect **force** to **work**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D4.3.1 Work Due to a Single Force

In earlier section, we defined **work** purely in terms of its effect on motion:

$$
w_{\text{net}} = \Delta k.
$$

This definition emphasized that work is fundamentally about **energy transfer**, not force.

We now ask the natural next question:

**What physical mechanism causes a change in kinetic energy?**

---

### From Energy Change to Physical Cause

In some situations, an interaction is not most naturally described by how long it acts, but by **how far** it acts.  
For example, we may know that a box is pushed with a constant force across a floor through a known distance, without knowing how long the push lasted.

This motivates a complementary way of describing energy transfer.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Concept: Work as Energy Transfer</h3>

<strong>Work is the transfer of energy caused by a force acting through a displacement.</strong>

</div>

This statement plays the same role for energy that force–time played for momentum:

- **Impulse** records interactions through **time**
- **Work** records interactions through **distance**

---

### Definition of Linear Work (Single Force)

When a single force acts on a point particle and causes it to move through a displacement, the energy transfer can be quantified directly.

For a **constant force**, or when a force can be reasonably replaced by its **average value**, the work done by that force is

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Linear Work on a Point Particle</h3>

$$
w = f_x\,\Delta x \tag{1}
$$

</div>

Here,

- $f_x$ is the force acting along the $x$-direction, and  
- $\Delta x$ is the **directed displacement** of the particle during the interaction.

The displacement $\Delta x$ represents a **distance traveled with direction**.  
Using our current notation, it could equivalently be written as $d_x$, emphasizing that displacement carries a sign just like velocity and force.  
The use of $\Delta x$ highlights the parallel with the impulse expression $f_x\,\Delta t$.

This expression makes an important physical point:

> **A large force acting over a short distance can transfer the same amount of energy as a smaller force acting over a longer distance.**

---

### Sign of Work

Because both force and displacement are directed quantities in one dimension, the **sign of the work** has clear physical meaning:

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">

- If the force and displacement are in the **same direction**, the work is **positive**.  
- If the force and displacement are in **opposite directions**, the work is **negative**.

</div>

The sign tells us how energy is transferred:

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">

- **Positive work:** energy is transferred *to* the particle.  
- **Negative work:** energy is transferred *from* the particle.

</div>

---

### Units of Work

From the definition,

$$
[w] = \text{N·m},
$$

since work is the product of force and displacement.

Because work represents **energy transfer**, this unit is defined as the **joule (J)**:

$$
1\ \text{N·m} = 1\ \text{J}.
$$

This confirms the consistency of our definitions:  
**work and energy share the same unit because they describe the same physical quantity from different perspectives.** We must emphasize though, that **work is not a form of energy, it is a transfer of energy**.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D4.3.2 Net Work from Multiple Forces

In many situations, a point particle experiences **more than one force** while moving through a displacement.  
Each force may transfer energy to or from the particle.

As with impulse, the most general and physically correct approach is to start with **work itself**, not force.

### Net Work — General Statement

The **net work** done on a point particle is the **sum of the work** done by each individual force:

$$
w_{\text{net}} = \sum w.
$$

Each individual contribution depends on:
- the force involved, and  
- the **displacement over which that force acts**.

For a single force acting over a displacement $\Delta x$,

$$
w = f_x\,\Delta x \quad \text{(constant or average force)}.
$$

Putting this together,

$$
w_{\text{net}} = \sum (f_x\,\Delta x)_i.
$$

<div style="background-color:#ffe6e6; border-left:6px solid #990000; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Important Caution</h3>

The displacement over which each force acts **does not have to be the same**.

Some forces may act only over part of the motion (such as friction or contact forces),  
while others may act over the entire displacement (such as gravity).

You must therefore **never assume** that forces can be added first and multiplied by a single displacement unless you have verified that they act over the same distance.
</div>

---

### Special Case — Forces Acting Over the Same Displacement

In simplified situations, all forces act over the **same displacement** $\Delta x$.  
When this is true, the net work simplifies to

$$
w_{\text{net}} = \left( \sum f_x \right)\Delta x.
$$

Defining the **net force** as

$$
f_{\text{net},x} = \sum f_x,
$$

we obtain

$$
w_{\text{net}} = f_{\text{net},x}\,\Delta x.
$$

This form is convenient, but it is valid **only** when all contributing forces act over the same displacement.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Work–Energy Theorem (1D)</h3>

The **net work** done on a point particle equals the **change in its kinetic energy**:

$$
w_{\text{net}} = \Delta k.
$$

This relation connects **forces acting through displacement** to **changes in motion**.
</div>


<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Verifying the Work–Energy Theorem</h3>

<strong>Problem</strong>  

A $0.50\ \text{kg}$ cart initially moves at $2.0\ \text{m/s}$ along a straight track.  
A constant force of $f_x = +2.0\ \text{N}$ acts on the cart over a displacement of $4.0\ \text{m}$ in the direction of motion.

The cart’s speed at the end of this displacement is $6.0\ \text{m/s}$.

Verify that the work done by the force equals the change in kinetic energy.

---

<strong>Solution</strong>

<strong>1. Work done by the force</strong>

Since the force is constant and acts in the direction of motion, the work done is

$$
w = f_x\,\Delta x = (2.0)(4.0) = 8.0\ \text{J}.
$$

---

<strong>2. Change in kinetic energy</strong>

Initial kinetic energy:

$$
k_i = \frac{1}{2}(0.50)(2.0)^2 = 1.0\ \text{J}.
$$

Final kinetic energy:

$$
k_f = \frac{1}{2}(0.50)(6.0)^2 = 9.0\ \text{J}.
$$

Change in kinetic energy:

$$
\Delta k = k_f - k_i = 9.0 - 1.0 = 8.0\ \text{J}.
$$

---

<strong>Conclusion</strong>

$$
w = \Delta k = 8.0\ \text{J}.
$$

The work done by the force exactly equals the change in kinetic energy, verifying the **work–energy theorem** for this situation.

</div>

---

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Net Work from Multiple Forces and Resulting Speed</h3>

<strong>Problem</strong>  

A box is pushed across a horizontal floor by a constant pushing force of $25\ \text{N}$. A frictional force of $5.0\ \text{N}$ acts on the box in the opposite direction.

The box moves a distance of $12\ \text{m}$ and starts from rest.  

1. How much <em>net work</em> is done on the box by the net force?  
2. How fast is the box moving after it has traveled this distance?

Assume the mass of the box is $m = 10\ \text{kg}$.

---

<strong>Methodology</strong>  

First determine the **net force** acting on the box and compute the **net work** using

$$
w_{\text{net}} = f_{\text{net},x}\,\Delta x.
$$

Then use the **work–energy theorem**

$$
w_{\text{net}} = \Delta k
$$

to determine the final speed of the box.

---

<strong>Solution</strong>  

Let the direction of motion be the positive $x$-direction.

- Pushing force: $+25\ \text{N}$  
- Frictional force: $-5.0\ \text{N}$  

The net force is

$$
f_{\text{net},x} = (+25) + (-5.0) = +20\ \text{N}.
$$

---

<strong>1. Net work</strong>  

The net work done over the displacement $\Delta x = 12\ \text{m}$ is

$$
w_{\text{net}} = f_{\text{net},x}\,\Delta x = (20)(12) = 240\ \text{J}.
$$

---

<strong>2. Final speed</strong>  

Since the box starts from rest, its initial kinetic energy is

$$
k_i = 0.
$$

From the work–energy theorem,

$$
w_{\text{net}} = \Delta k = k_f - k_i = k_f.
$$

Thus,

$$
k_f = 240\ \text{J}.
$$

Solving for the final speed:

$$
k_f = \frac{1}{2} m v^2
\quad \Rightarrow \quad
v = \sqrt{\frac{2k_f}{m}}.
$$

Substituting values,

$$
v = \sqrt{\frac{2(240)}{10}} = \sqrt{48} \approx 6.9\ \text{m/s}.
$$

---

<strong>Answers</strong>  

$$
\boxed{w_{\text{net}} = 2.4 \times 10^{2}\ \text{J}}
$$

$$
\boxed{v \approx 6.9\ \text{m/s}}
$$

---

<strong>Evaluation</strong>  

The net force acts in the same direction as the displacement, so the work is **positive**.  
As a result, energy is transferred to the box, increasing its kinetic energy and causing it to speed up from rest.

</div>

$$
w_{\text{net}} = \Delta k
\qquad \text{and} \qquad
w = f_x\,\Delta x.
$$

---

**Important Setup Condition**

In this activity, **all forces act on the particle over the same displacement** $\Delta x$.

$$
w_{\text{net}} = f_{\text{net},x}\,\Delta x.
$$

   $$
   m = 2.0\ \text{kg}.

2. **Forces acting on the particle**  

   All of the following forces act **throughout the entire displacement**:

   - Applied force: $f_{\text{app}} = +6.0\ \text{N}$  
   - Resistive force: $f_{\text{res}} = -2.0\ \text{N}$  

3. **Displacement**  

   The particle moves a distance  

   $$
   \Delta x = 5.0\ \text{m}.
   $$

4. **Compute the net force**

   Determine the net force $f_{\text{net},x}$ acting on the particle.

5. **Compute the net work**

   Use

   $$
   w_{\text{net}} = f_{\text{net},x}\,\Delta x
   $$

   to find the net work done on the particle.

7. **Connect to kinetic energy**

   Using the work–energy theorem,

   $$
   w_{\text{net}} = \Delta k,
   $$
 
   determine the final speed of the particle.

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 4: Net force**

$$
f_{\text{net},x} = (+6.0) + (-2.0) = +4.0\ \text{N}.
$$

**Step 5: Net work**

Because both forces act over the same displacement,

$$
w_{\text{net}} = (4.0)(5.0) = 20\ \text{J}.

**Step 6: Final speed**

Since the particle starts from rest, $k_i = 0$, so

$$
k_f = w_{\text{net}} = 20\ \text{J}.
$$

Using $k_f = \frac{1}{2}mv^2$,

$$
20 = \frac{1}{2}(2.0)v^2
\quad \Rightarrow \quad
v^2 = 20
\quad \Rightarrow \quad
v \approx 4.5\ \text{m/s}.
$$

**Interpretation**

- The net work is positive, so energy is transferred *to* the particle.  
- Combining forces is valid here only because they act over the same displacement.  
- The calculation depends on displacement, not time.

<div style="background-color:#e8f5e9; border-left:5px solid #006633; padding:12px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Box Activity 2 – Work When Forces Act Over Different Displacements</h3>

In many real situations, **different forces do not act over the same displacement**.  
In such cases, it is **not valid** to combine forces first and multiply by a single distance.

Instead, work must be computed **force by force**, and then summed.

Recall the general definition:

$$
w_{\text{net}} = \sum w_i = \sum (f_{x,i}\,\Delta x_i).
$$

---

**Important Setup Condition**

In this activity, **each force acts over a different displacement**.  
You must therefore compute the work done by each force **separately**.

---

**Task:** Complete the following steps.

1. **Given quantities**

   A box of mass  

   $$
   m = 4.0\ \text{kg}
   $$

   starts from rest and moves along the $+x$ direction.

2. **Forces acting on the box**

   - A pushing force  

     $$
     f_{\text{push}} = +10\ \text{N}
     $$

     acts over the first  

     $$
     \Delta x_1 = 6.0\ \text{m}.
     $$

   - A frictional force  

     $$
     f_{\text{fric}} = -4.0\ \text{N}
     $$

     acts over the **entire motion**, for a total displacement  

     $$
     \Delta x_2 = 10.0\ \text{m}.
     $$

3. **Compute individual work contributions**

   - Compute the work done by the pushing force.  
   - Compute the work done by friction.

4. **Compute the net work**

   Add the individual work contributions to find $w_{\text{net}}$.

5. **Connect to kinetic energy**

   Using

   $$
   w_{\text{net}} = \Delta k,
   $$

   determine the final speed of the box.

6. **Reflect**

   - Why would combining the forces and using a single displacement give an incorrect result?  
   - Which force removes energy from the system, and how can you tell?

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Step 3: Individual work contributions**

Work by the pushing force:

$$
w_{\text{push}} = (+10)(6.0) = +60\ \text{J}.
$$

Work by friction:

$$
w_{\text{fric}} = (-4.0)(10.0) = -40\ \text{J}.
$$

---

**Step 4: Net work**

$$
w_{\text{net}} = 60 - 40 = 20\ \text{J}.
$$

---

**Step 5: Final speed**

Since the box starts from rest,

$$
k_i = 0,
\quad k_f = w_{\text{net}} = 20\ \text{J}.
$$

Using $k_f = \frac{1}{2}mv^2$:

$$
20 = \frac{1}{2}(4.0)v^2
\quad \Rightarrow \quad
v^2 = 10
\quad \Rightarrow \quad
v \approx 3.2\ \text{m/s}.
$$

---

**Interpretation**

- The pushing force transfers energy *to* the box.  
- Friction transfers energy *from* the box.  
- Because the forces act over different displacements, work must be computed **term by term**.
