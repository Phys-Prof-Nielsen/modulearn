## Problem 1 â€” Work, Kinetic Energy, and Motion

A **5.0 kg box** is pushed across a horizontal surface by a **constant net force** of  

$$
f_x = +235\ \text{N}.
$$

The box starts from rest at  

$$
x = 0\ \text{m}.
$$

After being pushed a distance of  

$$
\Delta x = 45\ \text{m},
$$

answer the following:

1. How much **work** was done on the box?  
2. What is the **kinetic energy** at $x = 45\ \text{m}$?  
3. How **fast** is the box traveling at $x = 45\ \text{m}$?  
4. What is the **velocity** of the box at $x = 45\ \text{m}$?

Be careful with **units**, **signs**, and **significant figures**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Work by a constant force in one dimension:

$$
w = f_x\,\Delta x
$$

Workâ€“energy theorem:

$$
w_{\text{net}} = \Delta k
$$

Kinetic energy:

$$
k = \frac{1}{2}mv^2
$$

---

**Part 1 â€” Work done on the box**

**Given**
- $f_x = +235\ \text{N}$
- $\Delta x = +45\ \text{m}$

**Compute work**

$$
w = (235)(45) = 1.0575\times10^4\ \text{J}
$$

With proper significant figures:

$$
\boxed{w \approx 1.1\times10^4\ \text{J}}
$$

---

**Part 2 â€” Kinetic energy at 45 m**

The box starts from rest, so

$$
k_i = 0.
$$

Using the workâ€“energy theorem,

$$
k_f = k_i + w = w.
$$

Thus,

$$
\boxed{k_f \approx 1.1\times10^4\ \text{J}}
$$

---

**Part 3 â€” Speed at 45 m**

Use the definition of kinetic energy:

$$
k_f = \frac{1}{2}mv^2.
$$

Solve for speed:

$$
v = \sqrt{\frac{2k_f}{m}}
= \sqrt{\frac{2(1.0575\times10^4)}{5.0}}
$$

$$
v \approx 65\ \text{m/s}.
$$

**Result**

$$
\boxed{v \approx 65\ \text{m/s}}
$$

---

**Part 4 â€” Velocity at 45 m**

The force acts in the **positive $x$-direction**, and the box never reverses direction.  
Therefore, the velocity has the same sign as the motion.

$$
\boxed{v_x \approx +65\ \text{m/s}}
$$

---

**Final Summary*

- Positive work increases kinetic energy.  
- Because the net force and displacement are in the same direction, the work is positive.  
- The workâ€“energy theorem allows us to determine the speed **without using acceleration or time**.  

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import math

# Given values
m = 5.0            # kg
f_x = 235.0        # N
delta_x = 45.0     # m
v_i = 0.0          # m/s (starts from rest)

# Work done
w = f_x * delta_x

# Final kinetic energy (from workâ€“energy theorem)
k_f = w

# Final speed
v_f = math.sqrt(2 * k_f / m)

# Output results
print(f"Work done on the box: {w:.1e} J")
print(f"Kinetic energy at 45 m: {k_f:.1e} J")
print(f"Speed at 45 m: {v_f:.1e} m/s")
print(f"Velocity at 45 m: +{v_f:.1e} m/s")
```

## Problem 2 â€” Gravitational Work and Energy

During a pumpkin drop, a **9.5 kg pumpkin** is dropped from the roof of a **35 m** building.  
The force exerted by gravity on the pumpkin is

$$
f_g = mg,
$$

where

$$
g = 9.8\ \text{m/s}^2.
$$

Ignore air resistance.

The pumpkin is **thrown downward** with an initial speed of

$$
v_i = 25\ \text{m/s}.
$$

How fast is the pumpkin traveling **just before impact** with the ground?

Be careful with **signs**, **units**, and **physical interpretation**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Kinetic energy:

$$
k = \frac{1}{2}mv^2
$$

Gravitational potential energy (near Earth):

$$
u_g = mgy
$$

Mechanical energy conservation (no non-conservative forces):

$$
k_i + u_i = k_f + u_f
$$

---

**Step 1 â€” Choose a reference and write energies**

Let the ground be $y = 0$.

Initial state (roof):
- Height: $y_i = 35\ \text{m}$  
- Speed: $v_i = 25\ \text{m/s}$  

Initial kinetic energy:

$$
k_i = \frac{1}{2}(9.5)(25)^2 = 2.97\times10^3\ \text{J}.
$$

Initial gravitational potential energy:

$$
u_i = (9.5)(9.8)(35) = 3.26\times10^3\ \text{J}.
$$

---

**Step 2 â€” Final energy at the ground**

At impact:
- $y_f = 0 \Rightarrow u_f = 0$

Final mechanical energy is purely kinetic:

$$
k_f = \frac{1}{2}mv_f^2.
$$

---

**Step 3 â€” Apply mechanical energy conservation**

$$
k_i + u_i = k_f.
$$

Substitute values:

$$
2.97\times10^3 + 3.26\times10^3 = \frac{1}{2}(9.5)v_f^2.
$$

$$
6.23\times10^3 = 4.75\,v_f^2.
$$

Solve for $v_f$:

$$
v_f^2 = 1.31\times10^3
\quad \Rightarrow \quad
v_f \approx 36\ \text{m/s}.
$$

---

**Answer**

$$
\boxed{v_f \approx 36\ \text{m/s}}
$$

---

**Interpretation**

- Both the initial kinetic energy and the gravitational potential energy contribute to the final speed.  
- Gravity transfers energy from potential to kinetic, increasing the pumpkinâ€™s speed.  
- Mechanical energy is conserved because no non-conservative forces act.

</div>
</details>

```python
# DIY Cell
```

```python
%reset -f

import sympy as sym

m = 9.5
h = 35
g = 9.8
v1 = 25

# work
fx = m*g
dx = 35

w = fx*dx

# solve using work-energy theorem
v2 = sym.Symbol('v2')
eq = w - (0.5*m*v2**2 - 0.5*m*v1**2)
sol = sym.solve(eq,v2)
v2 = sol[1]
answer = "{:.1e}".format(float(v2))
print('The speed of the pumpkin right before impact is '+str(answer)+' m/s.')
```

## Problem 3 â€” Mechanical Energy in One Dimension (Vertical Motion)

A **3.0 kg object** moves only in the vertical ($y$) direction.  
The object is initially at rest at height

$$
y_i = 1.5\ \text{m}.
$$

A constant **upward force** of magnitude

$$
f_y = +40.o\ \text{N}
$$

acts on the object over a vertical displacement of

$$
\Delta y = +2.0\ \text{m}.
$$

Assume:
- gravity acts downward with $g = 9.8\ \text{m/s}^2$,  
- air resistance is negligible.

Answer the following:

1. How much **work** is done by the applied force?  
2. What is the **change in gravitational potential energy** of the object?  
3. What is the **final kinetic energy** of the object?  
4. How fast is the object moving at the end of the displacement?

Be careful with **units**, **signs**, and **energy accounting**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Work by a constant force:

$$
w = f_y\,\Delta y
$$

Gravitational potential energy:

$$
\Delta u_g = mg\,\Delta y
$$

Mechanical energy form of the workâ€“energy theorem:

$$
\Delta e_{\text{mech}} = w_{nc}
$$

---

**Part 1 â€” Work done by the applied force**

$$
w_{\text{app}} = (40)(2.0) = 80\ \text{J}.
$$

---

**Part 2 â€” Change in gravitational potential energy**

$$
\Delta u_g = (3.0)(9.8)(2.0) = 58.8\ \text{J}.
$$

The potential energy increases because the object moves upward.

---

**Part 3 â€” Final kinetic energy**

The object starts from rest, so

$$
k_i = 0.
$$

Mechanical energy accounting gives

$$
k_f + \Delta u_g = w_{\text{app}}.
$$

Thus,

$$
k_f = 80 - 58.8 = 21.2\ \text{J}.
$$

---

**Part 4 â€” Final speed**

Using

$$
k_f = \frac{1}{2}mv^2,
$$

we solve for $v$:

$$
v = \sqrt{\frac{2k_f}{m}}
= \sqrt{\frac{2(21.2)}{3.0}}
\approx 3.8\ \text{m/s}.
$$

---

**Final Answers**

- Work done by the applied force:  

  $$
  \boxed{w_{\text{app}} = 80\ \text{J}}
  $$

- Change in gravitational potential energy:  

  $$
  \boxed{\Delta u_g = +58.8\ \text{J}}
  $$

- Final kinetic energy:  

  $$
  \boxed{k_f = 21.2\ \text{J}}
  $$

- Final speed:  

  $$
  \boxed{v \approx 3.8\ \text{m/s}}
  $$

---

**Interpretation**

- The applied force is **non-conservative** and adds mechanical energy.  
- Gravity is **conservative** and stores part of that energy as gravitational potential energy.  
- The remaining energy appears as kinetic energy.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import math

# Given values
m = 3.0            # kg
f_y = 40.0         # N (upward applied force)
g = 9.8            # m/s^2
delta_y = 2.0      # m (upward displacement)

# Work done by the applied (non-conservative) force
w_app = f_y * delta_y

# Change in gravitational potential energy
delta_u = m * g * delta_y

# Initial kinetic energy (starts from rest)
k_i = 0.0

# Final kinetic energy from mechanical energy accounting
# k_f + delta_u = w_app
k_f = w_app - delta_u

# Final speed
v_f = math.sqrt(2 * k_f / m)

# Output results
print(f"Work done by applied force: {w_app:.1e} J")
print(f"Change in gravitational potential energy: {delta_u:.1e} J")
print(f"Final kinetic energy: {k_f:.1e} J")
print(f"Final speed: {v_f:.1e} m/s")
```

## Problem 4 â€” Mechanical Energy with Friction (1D Horizontal)

A **6.0 kg crate** moves only in the $+x$ direction along a horizontal floor.  
The crate has an initial speed of

$$
v_i = 8.0\ \text{m/s}.
$$

A constant kinetic friction force of magnitude

$$
f_{\text{fric}} = 15\ \text{N}
$$

acts on the crate opposite its motion. The crate slides a distance

$$
\Delta x = +12\ \text{m}.
$$

Assume:
- the crate does not change height, so $\Delta u_g = 0$,
- friction is the only non-conservative force acting.

Answer the following:

1. What is the work done by friction?  
2. What is the change in mechanical energy of the crate?  
3. What is the crateâ€™s final kinetic energy?  
4. How fast is the crate moving after sliding 12 m?

Be careful with **units**, **signs**, and **energy accounting**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Kinetic energy:

$$
k = \frac{1}{2}mv^2
$$

Work by a constant force:

$$
w = f_x\,\Delta x
$$

Mechanical energy form of the theorem (with $\Delta u = 0$ here):

$$
\Delta e_{\text{mech}} = w_{nc}
\quad \Rightarrow \quad
\Delta k = w_{nc}
$$

---

**Part 1 â€” Work done by friction**

Friction acts opposite the displacement, so its force is negative:

$$
f_x = -15\ \text{N}.
$$

Work by friction:

$$
w_{\text{fric}} = f_x\,\Delta x = (-15)(12) = -180\ \text{J}.
$$

**Result**

$$
\boxed{w_{\text{fric}} = -1.8\times10^2\ \text{J}}
$$

---

**Part 2 â€” Change in mechanical energy**

Since height does not change, $e_{\text{mech}} = k$ and

$$
\Delta e_{\text{mech}} = w_{nc} = w_{\text{fric}}.
$$

So

$$
\boxed{\Delta e_{\text{mech}} = -1.8\times10^2\ \text{J}}
$$

---

**Part 3 â€” Final kinetic energy**

Initial kinetic energy:

$$
k_i = \frac{1}{2}(6.0)(8.0)^2
= 3.0(64)
= 192\ \text{J}.
$$

Apply $\Delta k = w_{\text{fric}}$:

$$
k_f = k_i + w_{\text{fric}} = 192 - 180 = 12\ \text{J}.
$$

**Result**

$$
\boxed{k_f = 1.2\times10^1\ \text{J}}
$$

---

**Part 4 â€” Final speed**

$$
k_f = \frac{1}{2}mv_f^2
\quad \Rightarrow \quad
v_f = \sqrt{\frac{2k_f}{m}}
= \sqrt{\frac{2(12)}{6.0}}
= \sqrt{4}
= 2.0\ \text{m/s}.
$$

**Result**

$$
\boxed{v_f = 2.0\ \text{m/s}}
$$

---

**Interpretation**

- Friction does **negative work**, removing mechanical energy.  
- The crate slows down significantly because most of its initial kinetic energy is transferred out of mechanical energy.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import math

# Given values
m = 6.0                 # kg
v_i = 8.0               # m/s
f_fric = -15.0          # N (opposite direction of motion)
delta_x = 12.0          # m

# Initial kinetic energy
k_i = 0.5 * m * v_i**2

# Work done by friction (non-conservative work)
w_fric = f_fric * delta_x

# Change in mechanical energy (horizontal motion, no change in potential energy)
delta_e_mech = w_fric

# Final kinetic energy
k_f = k_i + w_fric

# Final speed
v_f = math.sqrt(2 * k_f / m)

# Output results
print(f"Initial kinetic energy: {k_i:.1e} J")
print(f"Work done by friction: {w_fric:.1e} J")
print(f"Change in mechanical energy: {delta_e_mech:.1e} J")
print(f"Final kinetic energy: {k_f:.1e} J")
print(f"Final speed: {v_f:.1e} m/s")
```

## Problem 5 â€” 1D Elastic Collision 

A **0.80 kg cart** moves in the $+x$ direction with speed

$$
v_{1,i} = +5.0\ \text{m/s}.
$$

It collides in a **perfectly elastic** one-dimensional collision with a second cart of mass

$$
m_2 = 1.20\ \text{kg}
$$

that is initially at rest:

$$
v_{2,i} = 0.
$$

Find the final velocities $v_{1,f}$ and $v_{2,f}$.

Be careful with **units**, **signs**, and the meaning of an **elastic collision**.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Momentum conservation:

$$
P_i = P_f
\quad \Rightarrow \quad
m_1 v_{1,i} + m_2 v_{2,i} = m_1 v_{1,f} + m_2 v_{2,f}
$$

Kinetic energy conservation (elastic collision):

$$
K_i = K_f
\quad \Rightarrow \quad
\frac{1}{2}m_1 v_{1,i}^2 + \frac{1}{2}m_2 v_{2,i}^2
=
\frac{1}{2}m_1 v_{1,f}^2 + \frac{1}{2}m_2 v_{2,f}^2
$$

---

**Given**

- $m_1 = 0.80\ \text{kg}$
- $m_2 = 1.20\ \text{kg}$
- $v_{1,i} = +5.0\ \text{m/s}$
- $v_{2,i} = 0$

---

**Step 1 â€” Momentum conservation**

$$
(0.80)(5.0) + (1.20)(0)
=
(0.80)v_{1,f} + (1.20)v_{2,f}
$$

$$
4.0 = 0.80v_{1,f} + 1.20v_{2,f}. \tag{1}
$$

---

**Step 2 â€” Kinetic energy conservation**

$$
\frac{1}{2}(0.80)(5.0)^2
=
\frac{1}{2}(0.80)v_{1,f}^2 + \frac{1}{2}(1.20)v_{2,f}^2
$$

Compute the left side:

$$
\frac{1}{2}(0.80)(25) = 10.0\ \text{J}.
$$

So

$$
10.0 = 0.40v_{1,f}^2 + 0.60v_{2,f}^2. \tag{2}
$$

---

**Step 3 â€” Solve the system**

From (1):

$$
0.80v_{1,f} = 4.0 - 1.20v_{2,f}
\quad \Rightarrow \quad
v_{1,f} = 5.0 - 1.5v_{2,f}.
$$

Substitute into (2):

$$
10.0 = 0.40(5.0 - 1.5v_{2,f})^2 + 0.60v_{2,f}^2.
$$

Solving gives

$$
v_{2,f} = 4.0\ \text{m/s}
$$

and therefore

$$
v_{1,f} = 5.0 - 1.5(4.0) = -1.0\ \text{m/s}.
$$

---

**Answers**

$$
\boxed{v_{1,f} = -1.0\ \text{m/s}}
\qquad
\boxed{v_{2,f} = +4.0\ \text{m/s}}
$$

---

**Interpretation**

- Cart 1 reverses direction after the collision.  
- Cart 2 moves forward with a larger speed.  
- Both momentum and kinetic energy are conserved, confirming the collision is elastic.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

import sympy as sp

# Given values
m1 = 0.80
m2 = 1.20
v1_i = 5.0
v2_i = 0.0

# Unknown final velocities
v1_f, v2_f = sp.symbols('v1_f v2_f')

# Initial and final momentum
p_i = m1*v1_i + m2*v2_i
p_f = m1*v1_f + m2*v2_f

# Initial and final kinetic energy
ke_i = 0.5*m1*v1_i**2 + 0.5*m2*v2_i**2
ke_f = 0.5*m1*v1_f**2 + 0.5*m2*v2_f**2

# Conservation equations
eq_momentum = sp.Eq(p_i, p_f)
eq_energy   = sp.Eq(ke_i, ke_f)

# Solve the system (as dictionaries)
solutions = sp.solve((eq_momentum, eq_energy), (v1_f, v2_f), dict=True)

# Select the physically meaningful solution (discard the trivial no-interaction solution)
physical_solution = [sol for sol in solutions if sol[v2_f] != 0][0]

# Extract numerical values
v1f_val = float(physical_solution[v1_f])
v2f_val = float(physical_solution[v2_f])

# Output results in scientific notation
print(f"Final velocity of cart 1: {v1f_val:.1e} m/s")
print(f"Final velocity of cart 2: {v2f_val:.1e} m/s")
```

## Problem D6 â€” Work and Power from Speed Change

A **1200 kg car** speeds up on a straight, horizontal road (no change in height).  
Its speed increases from

$$
v_i = 12\ \text{m/s}
$$

to

$$
v_f = 28\ \text{m/s}
$$

over a time interval of

$$
\Delta t = 6.0\ \text{s}.
$$

Assume that all the energy transfer into the carâ€™s motion is due to the net work done on the car, and ignore air resistance and rolling losses.

1. Find the **net work** done on the car during this interval.  
2. Find the **average power** delivered during this interval.

Be mindful of **units**, **significant figures**, and interpretation.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
Problem solution
</summary>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

Workâ€“energy theorem:

$$
w_{\text{net}} = \Delta k
$$

Kinetic energy:

$$
k = \frac{1}{2}mv^2
$$

Average power:

$$
P_{\text{avg}} = \frac{w}{\Delta t}
$$

---

**Part 1 â€” Net work from change in speed**

Compute the change in kinetic energy:

$$
w_{\text{net}} = \Delta k
= \frac{1}{2}m(v_f^2 - v_i^2).
$$

Substitute values:

$$
w_{\text{net}} =
\frac{1}{2}(1200)\left(28^2 - 12^2\right).
$$

Compute inside parentheses:

$$
28^2 = 784,\qquad 12^2 = 144,\qquad 784 - 144 = 640.
$$

So

$$
w_{\text{net}} = 600(640) = 3.84\times10^5\ \text{J}.
$$

With appropriate significant figures:

$$
\boxed{w_{\text{net}} \approx 3.8\times10^5\ \text{J}}
$$

---

**Part 2 â€” Average power**

$$
P_{\text{avg}} = \frac{w_{\text{net}}}{\Delta t}
= \frac{3.84\times10^5}{6.0}
= 6.40\times10^4\ \text{W}.
$$

With appropriate significant figures:

$$
\boxed{P_{\text{avg}} \approx 6.4\times10^4\ \text{W}}
$$

---

**Interpretation**

- The positive work corresponds to an increase in kinetic energy (the car speeds up).  
- Power describes how quickly that energy transfer happens.  
- A power of $6.4\times10^4\ \text{W}$ is about $64\ \text{kW}$.

</div>
</details>


```python
# DIY Cell
```

```python
%reset -f

# Problem D5.6 â€” Work from Speed Change and Average Power

# Given values
m = 1200.0        # kg
v_i = 12.0        # m/s
v_f = 28.0        # m/s
delta_t = 6.0     # s

# Change in kinetic energy (net work)
w_net = 0.5 * m * (v_f**2 - v_i**2)

# Average power
p_avg = w_net / delta_t

# Output results
print(f"Net work done on the car: {w_net:.1e} J")
print(f"Average power delivered: {p_avg:.1e} W")
```

## Problem D7 â€” Workâ€“Energy Across Two Regions (1D)

A block of mass $m$ moves along a straight track in **one dimension**.  
It passes through **two consecutive regions**:

1. **Push region:** Starting from rest, the block is pushed by a constant force $f_p$ in the direction of motion over a distance $d_1$.

m &= 8.00\ \text{kg},\\

4. Verify explicitly that the net work equals the **total change in kinetic energy**.

Be careful with **units**, **signs**, and **energy accounting**.

## Problem D8 ðŸðŸ’» â€” Workâ€“Energy Across Multiple 1D Regions (Actuator System)

In a packaging line, a **linear actuator** pushes a small carriage of mass $m$ along a straight guide rail.  
The motion is strictly **one-dimensional**.

The motion occurs in **three consecutive regions**:

1. **Drive stroke:**  
   Starting from rest, the actuator applies a constant thrust force $F_T$ in the direction of motion over a distance $d_1$.

2. **Friction zone:**  
   Immediately afterward, a constant friction (or braking) force $F_f$, opposite the direction of motion, acts over a distance $d_2$.

3. **Final trim section:**  
   Over a short distance $d_3$, a constant assistant force $F_a$ (same direction as the motion) acts so that the carriage reaches a specified final speed $v_f$ at the end of this section.

Assume:
- strictly one-dimensional motion,
- all forces are constant within their respective regions,
- no energy losses other than the stated friction,
- kinetic energy is given by $k = \tfrac12 m v^2$.

Use the **workâ€“energy theorem**

$$
w_{\text{net}} = \Delta k
$$

throughout.

**Given values (for numerical evaluation):**

$$
\begin{aligned}
m &= 2.50\ \text{kg},\\
F_T &= 180\ \text{N}, \qquad d_1 = 0.080\ \text{m},\\
F_f &= 55.0\ \text{N}, \qquad d_2 = 0.120\ \text{m},\\
d_3 &= 0.050\ \text{m}, \qquad v_f = 3.20\ \text{m/s}.
\end{aligned}
$$

Answer the following:

1. What is the **speed $v_1$** of the carriage at the end of the drive stroke?  
2. What is the **speed $v_2$** at the end of the friction zone?  
3. What **assistant force $F_a$** is required in the trim section so that the carriage reaches the desired final speed $v_f$?  
4. What is the **maximum allowable friction force $F_{f,\max}$** such that the carriage still enters the trim section with nonzero speed ($v_2 > 0$)?

Be careful with **units**, **signs**, and **energy accounting**.

**Relevant equations**

Work by a constant force:

$$
w = f_x\,\Delta x
$$

Workâ€“energy theorem:

$$
w_{\text{net}} = \Delta k
$$

Kinetic energy:

$$
k = \frac{1}{2}mv^2
$$

---

**Part 1 â€” Speed after the drive stroke**

Work done by the thrust force:

$$
w_T = F_T d_1.
$$

Starting from rest:

$$
k_1 = w_T = F_T d_1.
$$

Thus,

$$
v_1 = \sqrt{\frac{2F_T d_1}{m}}.
$$

Numerically,

$$
v_1 = \sqrt{\frac{2(180)(0.080)}{2.50}}
\approx 3.39\ \text{m/s}.
$$

---

**Part 2 â€” Speed after the friction zone**

Work done by friction:

$$
w_f = -F_f d_2.
$$

Kinetic energy after region 2:

$$
k_2 = k_1 + w_f = F_T d_1 - F_f d_2.
$$

Thus,

$$
v_2 = \sqrt{\frac{2(F_T d_1 - F_f d_2)}{m}}.
$$

Numerically,

$$
v_2 \approx 2.33\ \text{m/s}.
$$

---

**Part 3 â€” Assistant force in the trim section**

Over distance $d_3$, the work done by $F_a$ must change the kinetic energy from $k_2$ to

$$
k_f = \frac{1}{2}m v_f^2.
$$

Thus,

$$
F_a d_3 = k_f - k_2.
$$

Solving for $F_a$:

$$
F_a = \frac{\tfrac12 m(v_f^2 - v_2^2)}{d_3}.
$$

Numerically,

$$
F_a \approx 44.5\ \text{N}.
$$

---

**Part 4 â€” Maximum allowable friction**

To reach the trim section with nonzero speed, we require

$$
k_2 > 0
\quad \Rightarrow \quad
F_T d_1 - F_f d_2 > 0.
$$

Solving for the maximum friction force:

$$
F_{f,\max} = \frac{F_T d_1}{d_2}.
$$

Numerically,

$$
F_{f,\max} = \frac{(180)(0.080)}{0.120}
= 120\ \text{N}.
$$

---

**Interpretation**

- Energy is added by the actuator in the drive stroke.  
- Friction removes mechanical energy in the second region.  
- The trim section restores just enough energy to reach the desired final speed.  
- Excessive friction prevents the carriage from reaching the trim section.

```python
%reset -f

import sympy as sp
from IPython.display import display, Math

# -----------------------------
# Symbols
# -----------------------------
m, FT, d1 = sp.symbols('m F_T d1', positive=True)          # mass, thrust, drive distance
Ff, d2 = sp.symbols('F_f d2', positive=True)                # friction force, friction distance
Fa, d3 = sp.symbols('F_a d3', positive=True)                # assistant force, trim distance
v1, v2, vf = sp.symbols('v1 v2 vf', positive=True)          # speeds

# -----------------------------
# Helper: scientific-notation formatter with sig figs (exact sig figs in sci notation)
# -----------------------------
def fmt(x, sig=2): #If you donâ€™t specify sig when calling the function, it assumes 2 significant figures.
    return f"{x:.{sig-1}e}"

# -----------------------------
# Q1: Exit speed after drive stroke
# 1/2 m v1^2 = F_T d1  -> v1 = sqrt(2 F_T d1 / m)
# -----------------------------
v1_expr = sp.sqrt(2*FT*d1/m)
display(Math(r"\textbf{Q1:}\; \tfrac{1}{2} m v_1^2 - 0 = F_T d_1 \;\Rightarrow\; v_1 = " + sp.latex(v1_expr)))

# -----------------------------
# Q2: Speed after friction zone
# 1/2 m v2^2 = 1/2 m v1^2 - F_f d2
# -> v2 = sqrt( (2 F_T d1 - 2 F_f d2) / m )
# -----------------------------
v2_expr = sp.sqrt( (2*FT*d1 - 2*Ff*d2)/m )
display(Math(r"\textbf{Q2:}\; \tfrac{1}{2} m v_2^2 = \tfrac{1}{2} m v_1^2 - F_f d_2"
             r"\;\Rightarrow\; v_2 = " + sp.latex(v2_expr)))

# -----------------------------
# Q3: Required assistant force
# 1/2 m v_f^2 = 1/2 m v_2^2 + F_a d3
# -> F_a = [ (1/2) m v_f^2 - (1/2) m v_2^2 ] / d3 = m (v_f^2 - v_2^2) / (2 d3)
# -----------------------------
Fa_expr = m*(vf**2 - v2**2)/(2*d3)
display(Math(r"\textbf{Q3:}\; \tfrac{1}{2} m v_f^2 = \tfrac{1}{2} m v_2^2 + F_a d_3"
             r"\;\Rightarrow\; F_a = " + sp.latex(Fa_expr)))

# -----------------------------
# Q4: Maximum friction condition
# Require v2^2 > 0  => (2 F_T d1 - 2 F_f d2)/m > 0  => F_f < (F_T d1)/d2
# Thus F_{f,max} = (F_T d1)/d2 (limiting case v2 -> 0+)
# -----------------------------
Ff_max_expr = (FT*d1)/d2
ineq_tex = r"\textbf{Q4:}\; v_2^2 > 0 \;\Rightarrow\; F_f < \tfrac{F_T d_1}{d_2}."
display(Math(ineq_tex))
display(Math(r"\text{Hence } F_{f,\max} = " + sp.latex(Ff_max_expr)))

# -----------------------------
# Numerical values
# -----------------------------
subs_vals = {
    m: 2.50,      # kg
    FT: 180.0,    # N
    d1: 0.080,    # m
    Ff: 55.0,     # N
    d2: 0.120,    # m
    d3: 0.050,    # m
    vf: 3.20      # m/s
}

# Evaluate numerically
v1_num = float(v1_expr.subs(subs_vals))
v2_num = float(v2_expr.subs(subs_vals))
Fa_num = float(Fa_expr.subs({**subs_vals, v2: v2_num}))
Ff_max_num = float(Ff_max_expr.subs(subs_vals))

print("=== Numerical Results (scientific notation) ===")
print(f"Q1) v1 = {fmt(v1_num, sig=2)} m/s")
print(f"Q2) v2 = {fmt(v2_num, sig=2)} m/s")
print(f"Q3) F_a = {fmt(Fa_num, sig=2)} N")
print(f"Q4) F_f,max = {fmt(Ff_max_num, sig=2)} N")

# Quick check: argument for v2^2 must be positive
arg_v2 = float((2*subs_vals[FT]*subs_vals[d1] - 2*subs_vals[Ff]*subs_vals[d2]) / subs_vals[m])
print("\nSanity check: v2^2 argument =", fmt(arg_v2, sig=4), " (should be > 0)")
```

## Problem 10 ðŸðŸ’» â€” Work from a Forceâ€“Position Graph (1D Data)

A cart of mass $m$ moves along a straight track in **one dimension**.  
As it travels from $x=0$ to $x=x_{\max}$, a position-dependent force $F(x)$ acts in the **+x direction** (colinear with the motion). Assume no other energy losses.

You are given measured data for force versus position. Your job is to:

2. Estimate the work done by the force as the **area under the $F$â€“$x$ curve** using a trapezoidal sum.

| Position $x$ (m) | 0.00 | 0.10 | 0.20 | 0.30 | 0.40 | 0.50 | 0.60 |

2. Compute the work done from $0 \le x \le x_{\max}$ using a trapezoidal sum. Report $w$ in joules.  
3. If the cart starts from rest ($v_i = 0$), compute the final speed $v_f$ at $x_{\max}$.  
4. If instead the desired final speed is $v_f = 4.00\ \text{m/s}$ at $x_{\max}$, compute the required initial speed $v_i$.

<details>

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Relevant equations**

$$
w_{\text{net}} = \Delta k
$$

Kinetic energy:

$$
k = \frac{1}{2}mv^2

**Part 1 â€” Plot**

A correct plot shows the given data points with markers and a connecting line, with:
- horizontal axis labeled $x$ (m)
- vertical axis labeled $F$ (N)

**Part 2 â€” Trapezoidal estimate of work**

Here $\Delta x = 0.10\ \text{m}$ between each data point.

Compute the trapezoids:

$$
w \approx \sum_{i=0}^{5} \frac{(F_i + F_{i+1})}{2}(0.10).
$$

Using the table values:

$$
\begin{aligned}
w &\approx 0.10\left[\frac{0.0+12.0}{2}+\frac{12.0+18.0}{2}+\frac{18.0+22.0}{2}+\frac{22.0+20.0}{2}+\frac{20.0+15.0}{2}+\frac{15.0+10.0}{2}\right] \\

$$

So

$$
\boxed{w \approx 9.20\ \text{J}}.

$$
w = \Delta k = \frac{1}{2}mv_f^2.
$$

$$

= \sqrt{\frac{2(9.20)}{3.00}}
\approx 2.48\ \text{m/s}.
$$

So

$$
\boxed{v_f \approx 2.48\ \text{m/s}}.
$$

---

**Part 4 â€” Required initial speed for a desired final speed**

Use

$$
w = \frac{1}{2}m(v_f^2 - v_i^2).
$$

Solve for $v_i$:

$$
v_i = \sqrt{v_f^2 - \frac{2w}{m}}.
$$

With $v_f = 4.00\ \text{m/s}$:

$$
v_i = \sqrt{(4.00)^2 - \frac{2(9.20)}{3.00}}
= \sqrt{16.0 - 6.13}
\approx 3.14\ \text{m/s}.
$$

So

$$
\boxed{v_i \approx 3.14\ \text{m/s}}.
$$

```python
%reset -f

import numpy as np
import math
import sympy as sp
from IPython.display import display, Math
import matplotlib.pyplot as plt

# -----------------------------
# Data and parameters
# -----------------------------
x_data = np.array([0.00, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60])  # m
F_data = np.array([0.0, 12.0, 18.0, 22.0, 20.0, 15.0, 10.0])   # N
m_val = 3.00  # kg

# Scientific-notation formatter (exact sig figs using 'e' format)
def fmt(x, sig=1):
    return f"{x:.{sig-1}e}"

# -----------------------------
# Q1: Plot F vs x
# -----------------------------
plt.figure()
plt.plot(x_data, F_data, marker='o')
plt.fill_between(x_data, F_data, 0, alpha=0.2)  # visual cue for "area under the curve"
plt.xlabel("Position x (m)")
plt.ylabel("Force F (N)")
plt.title("Measured Force vs Position")
plt.grid(True)
plt.show()

# -----------------------------
# Q2: Trapezoidal work (area under Fâ€“x curve)
# -----------------------------
# Trapezoidal sum: sum over intervals i of (F_i + F_{i+1})/2 * (x_{i+1} - x_i)
W_num = np.trapezoid(F_data, x_data)  # J, since NÂ·m
print("=== Trapezoidal Work from Data ===")
print(f"W = {fmt(W_num)} J")
print(' ')

# -----------------------------
# Q3: Start from rest -> find v_f
# -----------------------------
vi_num = 0.0
vf_from_rest = math.sqrt(max(0.0, 2*W_num/m_val))
print("\n=== Q3: Final speed from rest ===")
print(f"v_f (v_i=0) = {fmt(vf_from_rest)} m/s")
print(' ')

# -----------------------------
# Q4: Given v_f, find required v_i
# -----------------------------
vf_target = 4.00  # m/s
# Rearranged: v_i = sqrt( v_f^2 - 2W/m )
under_root = vf_target**2 - 2*W_num/m_val
# Guard against small negative due to roundoff:
under_root = max(0.0, under_root)
vi_required = math.sqrt(under_root)

print("\n=== Q4: Initial speed needed to end at v_f = 4.00 m/s ===")
print(f"v_i (needed) = {fmt(vi_required)} m/s")

# Quick sanity messages
if under_root < 0:
    print("Note: The available work exceeds what is needed; the chosen target v_f cannot be reached with a real v_i.")
```

## Problem 11 ðŸðŸ’» â€” Work from Telemetry Force Data (F1 Straight, 1D)

A Formula 1 car (mass $m$) runs along a **straight** between the exit of Turn 10 at $x=0$ and the next braking board at $x=x_{\max}$. Telemetry provides the **net longitudinal force** on the chassis,

$$
F_{\text{net}}(x),
$$

as a function of position $x$ along the straight (engine minus resistive effects already combined). Motion is strictly **one-dimensional** along $+x$.

Use only the **workâ€“energy theorem**:

$$
w = \Delta k
\qquad \text{with} \qquad
k=\frac12 m v^2.
$$

**Measured data (net force vs. position):**

| Position $x$ (m) | 0   | 50  | 100 | 150 | 200 | 250 | 300 |
|------------------|-----|-----|-----|-----|-----|-----|-----|
| $F_{\text{net}}$ (N) | 4000 | 4200 | 4300 | 4200 | 4000 | 3800 | 3600 |

**Given values (for numerical evaluation):**

$$
m = 800\ \text{kg},
\qquad
x_{\max} = 300\ \text{m}.

Two cases will be considered:

- **Case A:** the car exits a slow corner with $v_i = 0$.  
- **Case B:** the required final speed is $v_f^{(\text{target})} = 85.0\ \text{m/s}$.

Answer the following:

1. Plot $F_{\text{net}}$ vs. $x$ using the table values (markers connected by a line). Label axes with units and include a title.  
2. Estimate the work done from $x=0$ to $x_{\max}$ using a trapezoidal sum (area under the $F$â€“$x$ curve). Report $w$ in joules.  
3. **Case A:** If $v_i=0$, compute the final speed $v_f$ at $x_{\max}$.  
4. **Case B:** If instead $v_f^{(\text{target})}=85.0\ \text{m/s}$ at $x_{\max}$, compute the required initial speed $v_i$.

Be careful with **units**, interpretation of **area**, and **energy accounting**.

**Relevant equations**

Work from a forceâ€“position graph (area under the curve):

$$
w \approx \sum \frac{(F_i + F_{i+1})}{2}\,\Delta x
$$

Workâ€“energy theorem:

$$
w = \Delta k = \frac12 m\left(v_f^2 - v_i^2\right)
$$

---

**Part 1 â€” Plot**

A correct plot shows the data points (markers) connected by a line, with:
- $x$ (m) on the horizontal axis  
- $F_{\text{net}}$ (N) on the vertical axis

---

**Part 2 â€” Trapezoidal estimate of work**

Here each interval has width

$$
\Delta x = 50\ \text{m}.
$$

Compute the trapezoids:

$$
w \approx \sum_{i=0}^{5} \frac{(F_i + F_{i+1})}{2}(50).
$$

Using the table values:

$$
\begin{aligned}
w &\approx 50\left[\frac{4000+4200}{2}+\frac{4200+4300}{2}+\frac{4300+4200}{2}+\frac{4200+4000}{2}+\frac{4000+3800}{2}+\frac{3800+3600}{2}\right]\\
&= 50\left[4100 + 4250 + 4250 + 4100 + 3900 + 3700\right]\\
&= 50(24300)\\
&= 1.215\times10^{6}\ \text{J}.
\end{aligned}
$$

So

$$
\boxed{w \approx 1.215\times10^{6}\ \text{J}}.
$$

---

**Part 3 â€” Case A final speed ($v_i = 0$)**

If $v_i = 0$:

$$
w = \frac12 m v_f^2
\quad \Rightarrow \quad
v_f = \sqrt{\frac{2w}{m}}.
$$

Substitute:

$$
v_f = \sqrt{\frac{2(1.215\times10^6)}{800}}
= \sqrt{3037.5}
\approx 55.1\ \text{m/s}.
$$

So

$$
\boxed{v_f \approx 55.1\ \text{m/s}}.
$$

---

**Part 4 â€” Case B required initial speed**

Use

$$
w = \frac12 m\left(v_f^2 - v_i^2\right).
$$

Solve for $v_i$:

$$
v_i = \sqrt{v_f^2 - \frac{2w}{m}}.
$$

With $v_f^{(\text{target})} = 85.0\ \text{m/s}$:

$$
v_i = \sqrt{(85.0)^2 - \frac{2(1.215\times10^6)}{800}}
= \sqrt{7225 - 3037.5}
= \sqrt{4187.5}
\approx 64.7\ \text{m/s}.
$$

So

$$
\boxed{v_i \approx 64.7\ \text{m/s}}.
$$

- The work is the area under the $F_{\text{net}}$ vs. $x$ graph.  
- Positive work increases kinetic energy.  
- The workâ€“energy theorem connects the telemetry force data directly to changes in speed.
