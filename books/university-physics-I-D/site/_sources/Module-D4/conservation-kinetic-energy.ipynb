## D4.5.1 Motivation — Why We Must Think in Terms of Systems

Conservation laws are meaningful only when we define a **system**.

For a **single particle**, the concept of kinetic energy conservation is trivial and uninformative:  
a single particle cannot exchange energy with itself. Any change in its kinetic energy must come from interactions with something else.

For this reason, discussions of energy conservation must involve **at least two interacting particles**.  
From this point forward:

- **Lowercase symbols** (e.g., $k$, $p$) refer to individual particles  
- **Uppercase symbols** (e.g., $K$, $P$) refer to the **total** quantities of a system

Our goal in this module is to understand when the **total kinetic energy of a system** remains constant during an interaction.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.5.2 Momentum Conservation — Why It Is Always Simpler

Recall why momentum conservation works so cleanly for a two-particle system.

During an interaction:
- Each particle experiences a force due to the other
- These forces are equal in magnitude and opposite in direction
- They act over the **same time interval** $\Delta t$

Because impulse depends on force acting through **time**, the impulses delivered to the two particles are equal and opposite:

$$
j_{1} = -j_{2}.
$$

As a result, the total momentum of the system,

$$
P = p_1 + p_2,
$$

satisfies

$$
\Delta P = 0
$$

whenever no external impulse acts on the system.

This is why **momentum conservation is universal** for isolated systems: the shared time interval guarantees cancellation of internal impulses.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.5.3 Kinetic Energy — A More Subtle Quantity

Kinetic energy behaves differently from momentum during interactions.

Momentum conservation follows directly from the structure of interaction forces:  
internal forces always come in equal and opposite pairs and act over the same time interval. As a result, momentum is automatically redistributed between particles, and the **total momentum of an isolated system is conserved**.

Kinetic energy does not share this automatic cancellation.

Work depends on **force acting through displacement**, and for interaction forces the relevant displacement is not just the overall motion of the objects. During an interaction, forces can act through **internal displacements** caused by deformation, compression, vibration, or sticking.

When internal forces act through these internal displacements, part of the work done does **not** remain as kinetic energy of motion. Instead, energy is transferred into internal forms such as:
- deformation,
- thermal energy,
- sound.

When this happens, the **total kinetic energy of the system decreases**, even though momentum remains conserved.

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Key Point</h3>

<strong>Momentum conservation follows from force symmetry.  
Kinetic energy conservation depends on whether internal work is transferred into non-kinetic forms.</strong>

</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


## D4.5.4 When Is Kinetic Energy Conserved?

<div style="background-color:#e6f0ff; border-left:6px solid #003366; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Conservation of Kinetic Energy</h3>

<strong>The total kinetic energy of a system is conserved only when the internal work done by the interaction forces sums to zero.</strong>

</div>

This occurs in **elastic collisions**, where:
- Interaction forces do not transform energy into thermal energy, sound, or internal deformation
- The work done on one particle is exactly balanced by the work done on the other

In **inelastic collisions**, some of the internal work changes energy into other forms, and total kinetic energy decreases—even though momentum remains conserved.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">


<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Elastic vs. Inelastic Collision and Kinetic Energy</h3>

<strong>Problem</strong>  

Two carts move along a straight, frictionless track.

- Cart 1 has mass $m$ and initial speed $v = 3.0\ \text{m/s}$.  
- Cart 2 has the same mass $m$ and is initially at rest.

Consider two different outcomes after the interaction:

**Case A:** The carts collide and separate. Cart 1 comes to rest, and Cart 2 moves away with speed $3.0\ \text{m/s}$.  

**Case B:** The carts collide and stick together, moving as a single object afterward.

For each case:
1. Compute the total kinetic energy before and after the collision.  
2. Determine whether kinetic energy is conserved.

---

<strong>Solution</strong>  

The initial kinetic energy of the system is

$$
K_i = \frac{1}{2}m(3.0)^2 = 4.5m.
$$

---

<strong>Case A: Carts Separate</strong>  

Final kinetic energy:

$$
K_f = \frac{1}{2}m(3.0)^2 = 4.5m.
$$

Since

$$
K_f = K_i,
$$

the total kinetic energy is conserved.

---

<strong>Case B: Carts Stick Together</strong>  

By momentum conservation, the combined mass $2m$ moves with speed

$$
v_f = \frac{m(3.0)}{2m} = 1.5\ \text{m/s}.
$$

Final kinetic energy:

$$
K_f = \frac{1}{2}(2m)(1.5)^2 = 2.25m.
$$

Here,

$$
K_f < K_i,
$$

so kinetic energy is not conserved.

---

<strong>Conclusion</strong>  

- **Case A** is an **elastic collision**: kinetic energy is conserved.  
- **Case B** is an **inelastic collision**: some kinetic energy is converted into internal energy (deformation, thermal energy, sound), even though momentum remains conserved.

</div>

<div style="background-color:#e0f7fa; border-left:6px solid #006a80; padding:14px; border-radius:4px;">
<h3 style="margin-top:0; color:#000000;">Example – Using Momentum and Kinetic Energy to Find Final Speeds (Elastic Collision)</h3>

<strong>Problem</strong>  

A cart of mass $m_1 = 1.0\ \text{kg}$ moves in the $+x$ direction with speed $v_{1,i} = 4.0\ \text{m/s}$.  
It undergoes a <strong>perfectly elastic</strong> collision with a second cart of mass $m_2 = 3.0\ \text{kg}$ that is initially at rest.

Assuming the collision is one-dimensional, find the final speeds $v_{1,f}$ and $v_{2,f}$.

---

<strong>Methodology</strong>  

A perfectly elastic collision satisfies <strong>both</strong>:

1. Conservation of momentum:

$$
P_i = P_f
\quad \Rightarrow \quad
m_1 v_{1,i} + m_2 v_{2,i} = m_1 v_{1,f} + m_2 v_{2,f}.
$$

2. Conservation of kinetic energy:

$$
K_i = K_f
\quad \Rightarrow \quad
\frac{1}{2}m_1 v_{1,i}^2 + \frac{1}{2}m_2 v_{2,i}^2
=
\frac{1}{2}m_1 v_{1,f}^2 + \frac{1}{2}m_2 v_{2,f}^2.
$$

Here $v_{2,i}=0$.

---

<strong>Solution</strong>  

<strong>1) Momentum equation</strong>

$$
(1.0)(4.0) = (1.0)v_{1,f} + (3.0)v_{2,f}
\quad \Rightarrow \quad
4.0 = v_{1,f} + 3v_{2,f}. \tag{1}
$$

<strong>2) Kinetic energy equation</strong>

$$
\frac{1}{2}(1.0)(4.0)^2 = \frac{1}{2}(1.0)v_{1,f}^2 + \frac{1}{2}(3.0)v_{2,f}^2.
$$

Multiply both sides by 2:

$$
16 = v_{1,f}^2 + 3v_{2,f}^2. \tag{2}
$$

<strong>3) Solve the system</strong>

From (1):

$$
v_{1,f} = 4.0 - 3v_{2,f}.
$$

Substitute into (2):

$$
16 = (4.0 - 3v_{2,f})^2 + 3v_{2,f}^2.
$$

Expand:

$$
16 = (16 - 24v_{2,f} + 9v_{2,f}^2) + 3v_{2,f}^2
= 16 - 24v_{2,f} + 12v_{2,f}^2.
$$

Subtract 16 from both sides:

$$
0 = -24v_{2,f} + 12v_{2,f}^2
= 12v_{2,f}(v_{2,f} - 2.0).
$$

So

$$
v_{2,f} = 0 \quad \text{or} \quad v_{2,f} = 2.0\ \text{m/s}.
$$

The $v_{2,f}=0$ solution corresponds to “no collision effect,” so we take:

$$
v_{2,f} = 2.0\ \text{m/s}.
$$

Then from (1):

$$
v_{1,f} = 4.0 - 3(2.0) = -2.0\ \text{m/s}.
$$

---

<strong>Answers</strong>  

$$
\boxed{v_{1,f} = -2.0\ \text{m/s}}
\qquad
\boxed{v_{2,f} = +2.0\ \text{m/s}}
$$

---

<strong>Evaluation</strong>  

- Cart 1 reverses direction after the collision (negative final velocity).  
- Cart 2 moves forward with speed $2.0\ \text{m/s}$.  
- Both momentum and kinetic energy are conserved, confirming that the collision is elastic.

<h3 style="margin-top:0; color:#000000;">Box Activity 2 – Solving an Elastic Collision with Python </h3>

In this activity, you will use **Python and SymPy** to solve an elastic collision problem by treating the conservation laws as a **system of equations**.

For a perfectly elastic, one-dimensional collision in an isolated system, both of the following must hold:

$$
P_i = P_f
\qquad \text{and} \qquad
K_i = K_f.
$$

---

**Physical Setup**

A cart of mass  

$$
m_1 = 1.5\ \text{kg}
$$

moves in the $+x$ direction with speed  

$$
v_{1,i} = 6.0\ \text{m/s}.
$$

It collides elastically with a second cart of mass  

$$
m_2 = 2.5\ \text{kg},
$$

which is initially at rest.

---

**Task**

1. Write the **momentum conservation equation** for the system.  
2. Write the **kinetic energy conservation equation** for the system.  
3. In the code cell below, use **SymPy** to solve the two equations simultaneously for:
   - $v_{1,f}$
   - $v_{2,f}$
4. Identify which mathematical solution is **physically meaningful** and explain why.

<details>
<summary style="background-color:#006633; color:white; padding:8px; border-radius:4px; cursor:pointer;">
A Possible Solution Guide

<div style="background-color:#e8f5e9; padding:10px; border-radius:4px; margin-top:6px;">

**Conservation Equations**

Momentum:
`m_1v1_i + m2*v2_i = m1*v1_f + m2*v2_f`

Kinetic Energy:
`0.5*m1*v1_i**2 + 0.5*m2*v2_i**2 = 0.5*m1*v1_f**2 + 0.5*m2*v2_f**2`

---

**SymPy Setup (example)**

See code box below for possible solution.

---

**Interpretation**

SymPy returns **two mathematical solutions**:

- One solution corresponds to  
  `v1_f = v1_i` and `v2_f = 0`,  
  which represents *no interaction* and must be rejected.

- The remaining solution gives the **physically correct post-collision velocities**.

This highlights an important lesson:

<strong>Mathematics can produce multiple solutions,  
but physics determines which one is real.</strong>

</div>

```python
%reset -f

import sympy as sp

# Given values
m1 = 1.5
m2 = 2.5
v1_i = 6.0
v2_i = 0.0

# Unknown final velocities
v1_f, v2_f = sp.symbols('v1_f v2_f')

# Initial momentum
p_i = m1*v1_i + m2*v2_i

# Final momentum
p_f = m1*v1_f + m2*v2_f

# Initial kinetic energy
ke_i = 0.5*m1*v1_i**2 + 0.5*m2*v2_i**2

# Final kinetic energy
ke_f = 0.5*m1*v1_f**2 + 0.5*m2*v2_f**2

# Conservation equations
eq_momentum = sp.Eq(p_i, p_f)
eq_energy   = sp.Eq(ke_i, ke_f)

# Solve the system (return solutions as dictionaries)
solutions = sp.solve((eq_momentum, eq_energy), (v1_f, v2_f), dict=True)

# Select the physically meaningful solution (exclude the trivial "no collision" case)
physical_solution = [sol for sol in solutions if sol[v2_f] != 0][0]

# Extract numerical values
v1f_val = float(physical_solution[v1_f])
v2f_val = float(physical_solution[v2_f])

# Output results in scientific notation
print(f"Final velocity of cart 1: {v1f_val:.1e} m/s")
print(f"Final velocity of cart 2: {v2f_val:.1e} m/s")
```
